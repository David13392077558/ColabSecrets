```python
# 运行单个模型预测示例
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
from google.colab import drive

# 1. 挂载 Google Drive (如果需要)
# drive.mount('/content/drive') # 如果已经挂载，可以注释掉

# 2. 设置数据和模型目录
base_dir = '/content/drive/MyDrive/Colab Notebooks'
data_filename = "002298.csv"
model_filename = "002298_lstm.keras"
data_path = os.path.join(base_dir, "data", data_filename)
model_path = os.path.join(base_dir, "models", model_filename)

# 3. 定义特征工程函数
def compute_rsi(series, window=14):
    """计算相对强弱指数 (RSI)"""
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    # 处理除以零的情况
    rs = gain / loss.replace(0, np.nan)
    rsi = 100 - (100 / (1 + rs))
    return rsi

def compute_atr(df, window=14):
    """计算平均真实范围 (ATR)"""
    high_low = df['High'] - df['Low']
    high_close = np.abs(df['High'] - df['Close'].shift())
    low_close = np.abs(df['Low'] - df['Close'].shift())
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    atr = tr.ewm(span=window, adjust=False).mean()
    return atr

def compute_obv(df):
    """计算能量潮 (OBV)"""
    obv = (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()
    return obv

def preprocess_data_for_prediction(df):
    """数据预处理和特征工程，为模型预测准备数据"""
    df["Date"] = pd.to_datetime(df["Date"])
    df.sort_values("Date", inplace=True)
    df.set_index("Date", inplace=True)

    # 添加衍生特征
    df["MA5"] = df["Close"].rolling(window=5).mean()
    df["MA10"] = df["Close"].rolling(window=10).mean()
    df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
    df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = df["EMA12"] - df["EMA26"]
    df["RSI"] = compute_rsi(df["Close"], window=14)
    df["Return_1d"] = df["Close"].pct_change()
    df["Volatility_5d"] = df["Close"].rolling(window=5).std()
    df["Weekday"] = df.index.weekday
    df["IsMonthEnd"] = df.index.is_month_end.astype(int)
    df["ATR"] = compute_atr(df, window=14)
    df["OBV"] = compute_obv(df)

    # 定义模型训练时使用的特征列表
    features = [
        'Open', 'Close', 'High', 'Low', 'Volume', 'Amplitude', 'ChangePct',
        'ChangeAmt', 'Turnover', 'MA5', 'MA10', 'EMA12', 'EMA26', 'MACD',
        'RSI', 'Return_1d', 'Volatility_5d', 'Weekday', 'IsMonthEnd', 'ATR', 'OBV'
    ]

    # 丢弃不需要的列和包含NaN的行
    df_processed = df.drop(columns=['Code', 'Amount']).dropna() # 假设Code和Amount不是模型输入的特征
    df_features = df_processed[features]

    # 标准化特征
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df_features)

    # 创建LSTM所需的序列数据
    time_step = 60 # 模型训练时使用的时间步长
    X_test = []
    # 调整循环范围以确保有足够的数据创建序列
    for i in range(len(scaled_data) - time_step + 1): # 修正循环范围以包含最后一个可能的序列
        X_test.append(scaled_data[i:(i + time_step)])

    X_test = np.array(X_test)

    # 返回处理后的数据，缩放器和对应的原始数据部分
    return X_test, scaler, df_processed.iloc[time_step -1:] # 返回与X_test对应的日期/索引

# 4. 定义预测和绘图函数
def predict_and_plot(model, scaled_data, scaler, window_size, stock_code, features, output_dir="./charts"):
    """使用模型进行预测并绘制结果图表"""
    # 获取用于预测的最后一个时间窗口数据
    last_window = scaled_data[-window_size:]
    # 调整形状以匹配模型输入 (样本数, 时间步长, 特征数)
    last_window = np.expand_dims(last_window, axis=0)

    # 进行预测
    prediction = model.predict(last_window)

    # 处理模型输出：如果模型输出多个值，假设第一个是收盘价预测
    if prediction.shape[-1] > 1:
        predicted_close_scaled = prediction[0, :] # 取出所有预测值 (例如未来10天的预测)
    else:
        predicted_close_scaled = prediction[0] # 取出单个预测值

    # 恢复预测价格到原始比例
    try:
        close_idx = features.index("Close")
    except ValueError:
        raise ValueError("特征列表中找不到 'Close' 列。")

    num_features = len(features)
    # 创建一个用于逆缩放的 dummy 数组
    # 数组形状为 (预测天数, 特征数)
    full_dummy = np.zeros((predicted_close_scaled.shape[-1], num_features))

    # 使用最后一个已知时间点的特征值作为参考进行逆缩放
    reference_features_scaled = scaled_data[-1, :]
    for i in range(predicted_close_scaled.shape[-1]):
        full_dummy[i, :] = reference_features_scaled
        full_dummy[i, close_idx] = predicted_close_scaled[i]

    # 逆缩放预测结果
    recovered_predictions = scaler.inverse_transform(full_dummy)[:, close_idx]

    # 创建输出目录并保存图表
    output_dir = "./charts"
    os.makedirs(output_dir, exist_ok=True)
    chart_path = os.path.join(output_dir, f"{stock_code}_{datetime.today().strftime('%Y-%m-%d')}_forecast.png")

    # 绘制图表 (添加字体设置以支持中文显示)
    plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体为黑体
    plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题

    plt.figure(figsize=(10, 4))
    plt.plot(range(1, predicted_close_scaled.shape[-1] + 1), recovered_predictions, marker='o', label="预测收盘价")
    plt.title(f"{stock_code} - 未来{predicted_close_scaled.shape[-1]}天收盘价预测")
    plt.xlabel("未来天数")
    plt.ylabel("收盘价")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path)
    plt.close()

    print(f"✅ 图表已保存：{chart_path}")
    return recovered_predictions # 返回预测结果

# 5. 执行流程
# 加载数据
try:
    df_stock = pd.read_csv(data_path)
    print(f"✅ 已成功从 {data_path} 加载数据")
except FileNotFoundError:
    print(f"❌ 找不到数据文件：{data_path}")
    exit() # 如果找不到数据文件，退出执行

# 加载模型
try:
    loaded_model = load_model(model_path)
    print(f"✅ 已成功从 {model_path} 加载模型")
except FileNotFoundError:
    print(f"❌ 找不到模型文件：{model_path}")
    exit() # 如果找不到模型文件，退出执行
except Exception as e:
    print(f"❌ 加载模型失败：{e}")
    exit() # 如果加载模型失败，退出执行


# 数据预处理 (获取用于预测的 scaled_data 和 scaler)
# 这里我们重新处理数据以获取 scaled_data 和 scaler 对象，用于预测和逆缩放
df_processed_for_scaling = df_stock.copy()
df_processed_for_scaling["Date"] = pd.to_datetime(df_processed_for_scaling["Date"])
df_processed_for_scaling.sort_values("Date", inplace=True)
df_processed_for_scaling.set_index("Date", inplace=True)

# 添加衍生特征
df_processed_for_scaling["MA5"] = df_processed_for_scaling["Close"].rolling(window=5).mean()
df_processed_for_scaling["MA10"] = df_processed_for_scaling["Close"].rolling(window=10).mean()
df_processed_for_scaling["EMA12"] = df_processed_for_scaling["Close"].ewm(span=12, adjust=False).mean()
df_processed_for_scaling["EMA26"] = df_processed_for_scaling["Close"].ewm(span=26, adjust=False).mean()
df_processed_for_scaling["MACD"] = df_processed_for_scaling["EMA12"] - df_processed_for_scaling["EMA26"]
df_processed_for_scaling["RSI"] = compute_rsi(df_processed_for_scaling["Close"], window=14)
df_processed_for_scaling["Return_1d"] = df_processed_for_scaling["Close"].pct_change()
df_processed_for_scaling["Volatility_5d"] = df_processed_for_scaling["Close"].rolling(window=5).std()
df_processed_for_scaling["Weekday"] = df_processed_for_scaling.index.weekday
df_processed_for_scaling["IsMonthEnd"] = df_processed_for_scaling.index.is_month_end.astype(int)
df_processed_for_scaling["ATR"] = compute_atr(df_processed_for_scaling, window=14)
df_processed_for_scaling["OBV"] = compute_obv(df_processed_for_scaling)

# 定义模型训练时使用的特征列表
features_used = [
    'Open', 'Close', 'High', 'Low', 'Volume', 'Amplitude', 'ChangePct',
    'ChangeAmt', 'Turnover', 'MA5', 'MA10', 'EMA12', 'EMA26', 'MACD',
    'RSI', 'Return_1d', 'Volatility_5d', 'Weekday', 'IsMonthEnd', 'ATR', 'OBV'
]

# 丢弃不需要的列和包含NaN的行
df_processed_for_scaling = df_processed_for_scaling.drop(columns=['Code', 'Amount']).dropna()
df_features_for_scaling = df_processed_for_scaling[features_used]

# 标准化特征 (这里 fit_transform 是为了获取 scaled_data 和 scaler)
scaler_for_plotting = MinMaxScaler(feature_range=(0, 1))
scaled_data_for_plotting = scaler_for_plotting.fit_transform(df_features_for_scaling)


# 使用predict_and_plot函数进行预测和绘图
print("\n正在进行预测和绘图...")
predicted_prices = predict_and_plot(
    model=loaded_model,
    scaled_data=scaled_data_for_plotting,
    scaler=scaler_for_plotting,
    window_size=60, # 模型训练时使用的时间步长
    stock_code="000625", # 股票代码
    features=features_used # 特征列表
)

print("\n预测完成。")
print("未来预测收盘价：", predicted_prices)
```

```python
import pandas as pd # 导入 pandas 库，用于数据处理和分析
import numpy as np # 导入 numpy 库，用于科学计算和数组操作
import os # 导入 os 库，用于操作系统相关的功能，比如创建文件夹
import matplotlib.pyplot as plt # 导入 matplotlib.pyplot，用于绘制图表
from datetime import datetime # 导入 datetime 模块，用于处理日期和时间
from sklearn.preprocessing import MinMaxScaler # 导入 MinMaxScaler，用于数据标准化（缩放到一定范围）
from tensorflow.keras.models import load_model # 导入 load_model 函数，用于加载训练好的 Keras 模型

# 定义一个类 StockForecaster，用于单个股票的预测
class StockForecaster:
    # 类的初始化方法，设置预测所需的参数和加载模型
    def __init__(self, stock_code, model_dir, features, window_size=60, forecast_horizon=10, chart_dir="./charts"):
        self.stock_code = stock_code # 股票代码
        self.model_path = os.path.join(model_dir, f"{stock_code}_lstm.keras") # 构建模型文件的完整路径
        self.features = features # 用于模型输入的特征列表
        self.window_size = window_size # 时间窗口大小，即用过去多少天的数据来预测未来
        self.forecast_horizon = forecast_horizon # 预测未来多少天
        self.scaler = MinMaxScaler() # 初始化一个数据标准化器
        self.scaled_data = None # 用于存储标准化后的数据
        self.chart_dir = chart_dir # 图表保存目录
        os.makedirs(self.chart_dir, exist_ok=True) # 如果图表目录不存在，则创建它

        try:
            self.model = load_model(self.model_path) # 尝试加载训练好的模型
            print(f"✅ 已成功加载模型：{self.model_path}") # 加载成功打印信息
        except Exception as e: # 如果加载失败，捕获异常并抛出 FileNotFoundError
            raise FileNotFoundError(f"❌ 无法加载模型文件：{self.model_path}\n错误信息：{e}")

    # 计算相对强弱指数（RSI）的函数
    def compute_rsi(self, series, window=14):
        delta = series.diff() # 计算价格差异
        gain = (delta.where(delta > 0, 0)).rolling(window=window).mean() # 计算上涨的平均值
        loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean() # 计算下跌的平均值
        rs = gain / loss.replace(0, np.nan) # 计算相对强度，处理除以零的情况
        return 100 - (100 / (1 + rs)) # 计算 RSI

    # 计算平均真实范围（ATR）的函数
    def compute_atr(self, df, window=14):
        high_low = df['High'] - df['Low'] # 计算最高价和最低价的差值
        high_close = np.abs(df['High'] - df['Close'].shift()) # 计算最高价和前一天收盘价的差值的绝对值
        low_close = np.abs(df['Low'] - df['Close'].shift()) # 计算最低价和前一天收盘价的差值的绝对值
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1) # 计算真实范围（TR），取三者中的最大值
        return tr.ewm(span=window, adjust=False).mean() # 计算 ATR，使用指数加权移动平均

    # 计算能量潮（OBV）的函数
    def compute_obv(self, df):
        # 根据收盘价的变化方向，乘以成交量，然后累加
        return (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()

    # 数据预处理和特征工程函数
    def preprocess(self, df):
        df = df.copy() # 复制原始 DataFrame，避免修改原始数据
        df["Date"] = pd.to_datetime(df["Date"]) # 将日期列转换为 datetime 对象
        df.sort_values("Date", inplace=True) # 按日期排序
        df.set_index("Date", inplace=True) # 将日期列设置为索引

        # 计算技术指标作为新特征
        df["MA5"] = df["Close"].rolling(window=5).mean() # 5日移动平均线
        df["MA10"] = df["Close"].rolling(window=10).mean() # 10日移动平均线
        df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean() # 12日指数移动平均线
        df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean() # 26日指数移动平均线
        df["MACD"] = df["EMA12"] - df["EMA26"] # MACD 线
        df["RSI"] = self.compute_rsi(df["Close"]) # RSI 指标
        df["Return_1d"] = df["Close"].pct_change() # 1日收益率
        df["Volatility_5d"] = df["Close"].rolling(window=5).std() # 5日波动率
        df["Weekday"] = df.index.weekday # 星期几
        df["IsMonthEnd"] = df.index.is_month_end.astype(int) # 是否是月末
        df["ATR"] = self.compute_atr(df) # ATR 指标
        df["OBV"] = self.compute_obv(df) # OBV 指标

        # 丢弃不需要的列（如股票代码和成交额）和包含 NaN 值的行
        df = df.drop(columns=['Code', 'Amount'], errors='ignore').dropna()
        df_features = df[self.features] # 选取用于模型的特征列

        # 标准化特征数据
        self.scaled_data = self.scaler.fit_transform(df_features) # 拟合标准化器并转换数据

    # 进行滚动预测的函数
    def rolling_forecast(self):
        predictions = [] # 存储预测结果的列表
        # 获取用于初始预测的最后一个时间窗口数据
        current_window = self.scaled_data[-self.window_size:].copy()

        # 循环进行未来 horizon 天的预测
        for step in range(self.forecast_horizon):
            input_window = np.expand_dims(current_window, axis=0) # 扩展维度以匹配模型输入形状
            pred_scaled = self.model.predict(input_window)[0] # 进行预测，获取标准化后的预测结果

            # 逆缩放预测结果
            close_idx = self.features.index("Close") # 获取 'Close' 特征在特征列表中的索引
            num_features = len(self.features)
            # 创建一个用于逆缩放的 dummy 数组
            # 数组形状为 (1, 特征数)
            dummy = np.zeros((1, num_features))
            # 使用最后一个已知时间点的特征值作为参考进行逆缩放
            reference_features_scaled = current_window[-1, :]
            dummy[0, :] = reference_features_scaled
            dummy[0, close_idx] = pred_scaled[0]


            recovered = self.scaler.inverse_transform(dummy)[0, close_idx] # 逆缩放预测结果
            predictions.append(recovered) # 将逆缩放后的预测价格添加到结果列表

            # 更新时间窗口，用于下一次预测（滚动预测）
            next_step = current_window[1:].copy() # 移除最旧的一天数据
            new_day = current_window[-1].copy() # 复制最新一天的数据
            new_day[close_idx] = pred_scaled[0] # 将预测的收盘价（标准化后）作为新的一天数据
            current_window = np.vstack([next_step, new_day]) # 将新的一天数据添加到窗口末尾

        self.plot_forecast(predictions) # 绘制预测图表
        return predictions # 返回预测结果

    # 绘制预测结果图表的函数
    def plot_forecast(self, predictions):
        # 设置 Matplotlib 字体以支持中文显示
        plt.rcParams['font.sans-serif'] = ['SimHei']
        plt.rcParams['axes.unicode_minus'] = False

        plt.figure(figsize=(10, 4)) # 创建图表
        # 绘制预测收盘价曲线
        plt.plot(range(1, len(predictions) + 1), predictions, marker='o', label="预测收盘价")
        plt.title(f"{self.stock_code} - 未来{len(predictions)}天收盘价预测") # 设置图表标题
        plt.xlabel("未来天数") # 设置 X 轴标签
        plt.ylabel("收盘价") # 设置 Y 轴标签
        plt.grid(True) # 显示网格
        plt.legend() # 显示图例
        plt.tight_layout() # 自动调整布局

        # 保存图表到文件
        chart_path = os.path.join(self.chart_dir, f"{self.stock_code}_{datetime.today().strftime('%Y-%m-%d')}_forecast.png")
        plt.savefig(chart_path)
        plt.close() # 关闭图表，释放内存
        print(f"✅ 图表已保存：{chart_path}")

# 定义 BatchForecaster 类，用于批量预测多支股票
class BatchForecaster:
    # 类的初始化方法
    def __init__(self, stock_codes, model_dir, data_dir, features, window_size=60, forecast_horizon=10, chart_dir="./charts"):
        self.stock_codes = stock_codes # 股票代码列表
        self.model_dir = model_dir # 模型文件目录
        self.data_dir = data_dir # 数据文件目录
        self.features = features # 特征列表
        self.window_size = window_size # 时间窗口大小
        self.forecast_horizon = forecast_horizon # 预测天数
        self.chart_dir = chart_dir # 图表保存目录
        self.results = {} # 存储所有股票预测结果的字典
        os.makedirs(self.chart_dir, exist_ok=True) # 确保图表目录存在
        self.scaler = MinMaxScaler() # 初始化一个数据标准化器，用于整体标准化

    # 计算相对强弱指数（RSI）的函数
    def compute_rsi(self, series, window=14):
        delta = series.diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
        rs = gain / loss.replace(0, np.nan)
        return 100 - (100 / (1 + rs))

    # 计算平均真实范围（ATR）的函数
    def compute_atr(self, df, window=14):
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        return tr.ewm(span=window, adjust=False).mean()

    # 计算能量潮（OBV）的函数
    def compute_obv(self, df):
        return (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()


    # 数据预处理和特征工程函数
    def preprocess_batch(self):
        all_data = []
        for code in self.stock_codes:
            data_path = os.path.join(self.data_dir, f"{code}.csv")
            try:
                df = pd.read_csv(data_path)
                df["Date"] = pd.to_datetime(df["Date"])
                df.sort_values("Date", inplace=True)
                df.set_index("Date", inplace=True)

                # 添加衍生特征
                df["MA5"] = df["Close"].rolling(window=5).mean()
                df["MA10"] = df["Close"].rolling(window=10).mean()
                df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
                df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
                df["MACD"] = df["EMA12"] - df["EMA26"]
                df["RSI"] = self.compute_rsi(df["Close"])
                df["Return_1d"] = df["Close"].pct_change()
                df["Volatility_5d"] = df["Close"].rolling(window=5).std()
                df["Weekday"] = df.index.weekday
                df["IsMonthEnd"] = df.index.is_month_end.astype(int)
                df["ATR"] = self.compute_atr(df)
                df["OBV"] = self.compute_obv(df)

                df = df.drop(columns=['Code', 'Amount'], errors='ignore').dropna()
                df_features = df[self.features].copy()
                df_features['Code'] = code # 添加股票代码列用于后续分组
                all_data.append(df_features)

            except FileNotFoundError:
                print(f"❌ 找不到数据文件：{data_path}")
            except Exception as e:
                print(f"❌ 处理股票 {code} 数据失败：{e}")

        if not all_data:
            raise ValueError("没有加载到任何股票数据，无法进行标准化。")

        combined_df = pd.concat(all_data)
        # 对所有股票数据进行整体标准化
        self.scaled_data = self.scaler.fit_transform(combined_df[self.features])
        combined_df[self.features] = self.scaled_data # 将标准化后的数据放回 DataFrame
        return combined_df


    # 进行批量预测的函数
    def run(self):
        try:
            combined_df = self.preprocess_batch() # 预处理所有股票数据并整体标准化
            print("✅ 所有股票数据已预处理并整体标准化。")
        except ValueError as e:
            print(f"❌ 批量预测失败：{e}")
            return

        for code in self.stock_codes: # 遍历股票代码列表
            print(f"\n📈 正在处理股票：{code}") # 打印当前处理的股票代码
            try:
                # 获取当前股票的标准化数据
                stock_scaled_data = combined_df[combined_df['Code'] == code][self.features].values

                if len(stock_scaled_data) < self.window_size:
                    print(f"⚠️ 股票 {code} 数据量不足 ({len(stock_scaled_data)} 天)，无法进行预测。")
                    self.results[code] = None
                    continue

                # 获取用于初始预测的最后一个时间窗口数据
                current_window = stock_scaled_data[-self.window_size:].copy()

                # 加载当前股票的模型
                model_path = os.path.join(self.model_dir, f"{code}_lstm.keras")
                try:
                    model = load_model(model_path)
                    print(f"✅ 已成功加载模型：{model_path}")
                except Exception as e:
                    print(f"❌ 无法加载股票 {code} 的模型文件：{model_path}\n错误信息：{e}")
                    self.results[code] = None
                    continue

                predictions = [] # 存储预测结果的列表
                # 循环进行未来 horizon 天的预测
                for step in range(self.forecast_horizon):
                    input_window = np.expand_dims(current_window, axis=0) # 扩展维度以匹配模型输入形状
                    pred_scaled = model.predict(input_window, verbose=0)[0] # 进行预测，获取标准化后的预测结果

                    # 逆缩放预测结果，使用整体的标准化器
                    close_idx = self.features.index("Close") # 获取 'Close' 特征在特征列表中的索引
                    num_features = len(self.features)
                    # Create a dummy array for inverse scaling
                    # The shape of the array is (1, num_features)
                    dummy = np.zeros((1, num_features))
                    # Use the feature values from the last known timestep as a reference for inverse scaling
                    reference_features_scaled = current_window[-1, :]
                    dummy[0, :] = reference_features_scaled
                    dummy[0, close_idx] = pred_scaled[0]


                    recovered = self.scaler.inverse_transform(dummy)[0, close_idx] # 逆缩放预测结果
                    predictions.append(recovered) # 将逆缩放后的预测价格添加到结果列表

                    # Update the time window for the next prediction (rolling forecast)
                    next_step = current_window[1:].copy() # Remove the oldest day's data
                    new_day = current_window[-1].copy() # Copy the latest day's data
                    new_day[close_idx] = pred_scaled[0] # Use the predicted close price (scaled) as the new day's data
                    current_window = np.vstack([next_step, new_day]) # Add the new day's data to the end of the window


                self.plot_forecast(predictions, code) # 绘制预测图表
                self.results[code] = predictions # 将预测结果存储到结果字典中

            except Exception as e: # If processing fails, catch the exception and print the error message
                print(f"❌ 股票 {code} 处理失败：{e}")
                self.results[code] = None # Set the prediction result for this stock to None

    # 绘制预测结果图表的函数
    def plot_forecast(self, predictions, stock_code):
        # 设置 Matplotlib 字体以支持中文显示
        plt.rcParams['font.sans-serif'] = ['SimHei']
        plt.rcParams['axes.unicode_minus'] = False

        plt.figure(figsize=(10, 4)) # 创建图表
        # 绘制预测收盘价曲线
        plt.plot(range(1, len(predictions) + 1), predictions, marker='o', label="预测收盘价")
        plt.title(f"{stock_code} - 未来{len(predictions)}天收盘价预测") # 设置图表标题
        plt.xlabel("未来天数") # 设置 X 轴标签
        plt.ylabel("收盘价") # 设置 Y 轴标签
        plt.grid(True) # 显示网格
        plt.legend() # 显示图例
        plt.tight_layout() # 自动调整布局

        # 保存图表到文件
        chart_path = os.path.join(self.chart_dir, f"{stock_code}_{datetime.today().strftime('%Y-%m-%d')}_forecast.png")
        plt.savefig(chart_path)
        plt.close() # 关闭图表，释放内存
        print(f"✅ 图表已保存：{chart_path}")


    # 获取预测结果的函数
    def get_results(self):
        return self.results # 返回存储所有预测结果的字典

    # 将预测结果保存到 CSV 文件的函数
    def save_to_csv(self, output_path="./batch_predictions.csv"):
        # 创建输出目录（如果不存在）
        output_dir = os.path.dirname(output_path)
        if output_dir and not os.path.exists(output_dir):
            os.makedirs(output_dir, exist_ok=True)

        rows = [] # 存储 CSV 行数据的列表
        for code, preds in self.results.items(): # 遍历预测结果字典
            if preds is None: # 如果某支股票预测失败，则跳过
                continue
            for i, price in enumerate(preds, start=1): # 遍历每支股票的预测结果
                # 将预测结果格式化为字典，方便创建 DataFrame
                rows.append({
                    "StockCode": code,
                    "Day": f"Day {i}",
                    "PredictedClose": round(price, 4) # 保留 4 位小数
                })

        df_output = pd.DataFrame(rows) # 从行数据列表创建 DataFrame
        # 将 DataFrame 保存为 CSV 文件，不包含索引，使用 utf-8-sig 编码以确保中文显示正常
        df_output.to_csv(output_path, index=False, encoding="utf-8-sig")
        print(f"✅ 所有预测结果已保存为 CSV：{output_path}") # 打印保存成功信息

# 主执行块
if __name__ == "__main__":
    # 定义用于模型训练和预测的特征列表
    features_used = [
        'Open', 'Close', 'High', 'Low', 'Volume', 'Amplitude', 'ChangePct',
        'ChangeAmt', 'Turnover', 'MA5', 'MA10', 'EMA12', 'EMA26', 'MACD',
        'RSI', 'Return_1d', 'Volatility_5d', 'Weekday', 'IsMonthEnd', 'ATR', 'OBV'
    ]

    # 定义需要进行批量预测的股票代码列表
    stock_codes = ['002269','603920','601138','002272','002456','000070']

    # 创建 BatchForecaster 实例
    batch = BatchForecaster(
        stock_codes=stock_codes,
        model_dir="/content/drive/MyDrive/Colab Notebooks/models", # 模型所在的目录
        data_dir="/content/drive/MyDrive/Colab Notebooks/data", # 数据所在的目录
        features=features_used, # 使用的特征
        window_size=60, # 时间窗口大小
        forecast_horizon=10, # 预测未来天数
        chart_dir="./charts" # 图表保存目录
    )

    batch.run() # 运行批量预测
    # 保存预测结果到 CSV 文件在 Google Drive 中
    batch.save_to_csv(output_path="/content/drive/MyDrive/Colab Notebooks/results/batch_predictions.csv")
```
