```python
from google.colab import drive
drive.mount('/content/drive')
import torch
import os
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# 指定模型路径
model_path = '/content/drive/MyDrive/AI/models/gpt2-xl'  # 替换为您的模型路径

# 检查是否有可用的 GPU
if torch.cuda.is_available():
    device = torch.device("cuda")  # 使用 GPU
    print("使用 GPU")
else:
    device = torch.device("cpu")  # 使用 CPU
    print("使用 CPU")

try:
    # 加载模型和分词器
    model = GPT2LMHeadModel.from_pretrained(model_path)
    tokenizer = GPT2Tokenizer.from_pretrained(model_path)

    # 将模型移动到 GPU
    model.to(device)

    # 设置模型为评估模式
    model.eval()

    # 输入文本
    input_text = "Once upon a time"

    # 编码输入文本
    input_ids = tokenizer.encode(input_text, return_tensors='pt').to(device)  # 将输入数据移动到 GPU

    # 明确设置 pad token
    tokenizer.pad_token = tokenizer.eos_token

    # 创建 attention mask
    attention_mask = input_ids.ne(tokenizer.pad_token_id).long().to(device)  # 将 attention_mask 移动到 GPU

    # 生成文本
    output = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=50,
        num_return_sequences=1,
        do_sample=True,  # 使用采样生成
        temperature=0.7,  # 控制生成文本的随机性
        top_p=0.9,      # 进行 nucleus sampling
    )

    # 解码生成的文本
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    print("模型加载成功！")
    print("生成的文本：", generated_text)

except Exception as e:
    print(f"模型加载失败：{e}")
```

```python
import torch # 导入torch库，用于GPU操作
import os
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# ... (像之前一样加载你的模型和tokenizer) ...

# 1. 确定设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # 判断是否有可用的GPU，有则使用GPU，否则使用CPU
print(f"使用设备: {device}") # 打印当前使用的设备

# 2. 在加载任何数据*之前*，将模型移动到设备
model.to(device) # 将模型加载到指定的设备

# 3. 输入文本和编码
input_text = "Once upon a time"
input_ids = tokenizer.encode(input_text, return_tensors='pt').to(device) # 在创建input_ids时就将其移动到指定设备

# 4. Attention mask (处理可能的None pad_token_id)
if tokenizer.pad_token_id is not None:
    attention_mask = input_ids.ne(tokenizer.pad_token_id).long().to(device) # 在创建attention_mask时就将其移动到指定设备
else: # 如果pad_token_id是None，则全部设为1
    attention_mask = torch.ones(input_ids.shape, dtype=torch.long).to(device)

# 5. 生成文本
output = model.generate(
    input_ids=input_ids,  # 明确指定input_ids参数名，增加代码可读性
    attention_mask=attention_mask,
    max_length=50,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
