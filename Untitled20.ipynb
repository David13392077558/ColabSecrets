```python
# ✅ 导入库
import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import zipfile
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, MultiHeadAttention, LayerNormalization, Dense, Dropout, TimeDistributed, Lambda
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.metrics import MeanAbsoluteError, BinaryAccuracy


# ✅ 股票列表
stock_codes = ['002456','601138','601288','002269']
start_date = "20210924" # Modified start_date
end_date = datetime.today().strftime("%Y%m%d")
save_path = "./data/"
os.makedirs(save_path, exist_ok=True)

# ✅ 标准化字段
def standardize_columns(df):
    column_mapping = {
        "日期": "Date", "股票代码": "Code", "开盘": "Open", "收盘": "Close",
        "最高": "High", "最低": "Low", "成交量": "Volume", "成交额": "Amount",
        "振幅": "Amplitude", "涨跌幅": "ChangePct", "涨跌额": "ChangeAmt", "换手率": "Turnover",
        # Ensure 'Date' is always included in the mapping if it exists
    }
    # Add existing columns to the mapping to ensure they are kept
    for col in df.columns:
        if col not in column_mapping:
            column_mapping[col] = col

    # Filter out mappings where the source column doesn't exist in the DataFrame
    valid_column_mapping = {old_col: new_col for old_col, new_col in column_mapping.items() if old_col in df.columns}
    df.rename(columns=valid_column_mapping, inplace=True)
    return df

# ✅ RSI计算
def compute_rsi(series, window=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# ✅ 数据处理 (Updated to use the new loading function and add feature engineering)
def process_features(df):
    if df is None or df.empty:
        return None, None, None, None, None, None

    # Ensure '涨跌幅' column exists, if not, create it or handle its absence
    # Prioritize 'ChangePct' if it exists after standardization, otherwise check for '涨跌幅'
    if "ChangePct" in df.columns:
        df["涨跌幅"] = df["ChangePct"]
    elif "涨跌幅" not in df.columns:
        df["涨跌幅"] = 0 # Or np.nan, depending on how you want to handle missing data
        print(f"⚠️ Column '涨跌幅' not found. Created with default value 0.")

    # Create the classification target: 1 for price increase (涨), 0 otherwise (跌或平)
    # Based on the '涨跌幅' column
    df["涨跌方向"] = (df["涨跌幅"] > 0).astype(int)


    # Derivative Features
    df["MA5"] = df["Close"].rolling(window=5).mean()
    df["MA10"] = df["Close"].rolling(window=10).mean()
    df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
    df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = df["EMA12"] - df["EMA26"]
    df["RSI"] = compute_rsi(df["Close"], window=14)
    df["Return_1d"] = df["Close"].pct_change()
    df["Volatility_5d"] = df["Close"].rolling(window=5).std()
    df["Weekday"] = df.index.weekday
    df["IsMonthEnd"] = df.index.is_month_end.astype(int)
    # Handle potential division by zero for MoneyFlow
    df["MoneyFlow"] = df.apply(lambda row: row["Amount"] / row["Volume"] if row["Volume"] != 0 else 0, axis=1)


    # Fund flow fields - ensure these exist after merging or are initialized
    # Moved initialization logic to load_daily_and_fundflow
    fundflow_features = ["主力净流入", "主力净流入比率"]

    df.dropna(inplace=True)

    features = [
        "Open", "High", "Low", "Close", "Volume", "Amount", "Turnover",
        "MA5", "MA10", "EMA12", "EMA26", "MACD", "RSI",
        "Return_1d", "Volatility_5d", "Weekday", "IsMonthEnd", "MoneyFlow"
    ] + [col for col in fundflow_features if col in df.columns] # Only include fund flow features if they exist

    # Updated targets based on user request
    targets_regression = ["Close", "主力净流入", "涨跌幅"] # Removed "超大单净流入", "大单净流入"
    target_classification = "涨跌方向" # New classification target

    # Filter targets to only include columns that exist in the dataframe
    targets_regression = [target for target in targets_regression if target in df.columns]
    if "涨跌幅" not in targets_regression and "ChangePct" in df.columns:
         targets_regression.append("ChangePct") # Use ChangePct if 涨跌幅 is not available and ChangePct exists
    elif "ChangePct" not in targets_regression and "涨跌幅" in df.columns:
         targets_regression.append("涨跌幅")


    # Check if essential targets are present
    if "Close" not in df.columns:
         raise ValueError("Required target column 'Close' is missing from the data.")
    else:
        # Re-filter targets based on available columns before scaling
        # Ensure '涨跌幅' is in targets if it was created
        targets_regression = [target for target in ["Close", "主力净流入", "涨跌幅", "ChangePct"] if target in df.columns or (target == "涨跌幅" and "涨跌幅" in df.columns)]

        # Ensure '涨跌方向' is also in the dataframe for the classification target
        if target_classification not in df.columns:
             raise ValueError(f"Required classification target column '{target_classification}' is missing from the data.")


    # Combine features and regression targets for scaling
    cols_to_scale = [col for col in features + targets_regression if col in df.columns and df[col].dtype in ['int64', 'float64']]
    scaler = MinMaxScaler()
    df_scaled_regression_features = scaler.fit_transform(df[cols_to_scale])

    # The classification target is not scaled
    target_classification_data = df[target_classification].values

    # Need to return the actual target columns data for the scaler
    # Ensure the order of targets matches the order in cols_to_scale for correct inverse transformation later
    target_columns_in_scaled_order = [col for col in cols_to_scale if col in targets_regression]
    target_regression_data = df[target_columns_in_scaled_order].values

    # Return the list of actual regression target names used for scaling
    actual_regression_target_names = [col for col in cols_to_scale if col in targets_regression]


    return df, df_scaled_regression_features, target_regression_data, target_classification_data, scaler, actual_regression_target_names

# ✅ 加载指定股票的日线数据和资金流数据，并进行合并与预处理。
# 返回：标准化后的 DataFrame（用于建模）、收盘价序列、scaler 对象
def load_daily_and_fundflow(code, backtest_start, backtest_end):
    """
    加载指定股票的日线数据和资金流数据，并进行合并与预processing.
    返回：标准化后的 DataFrame（用于建模）、收盘价序列、scaler 对象
    """
    prefix = code[-6:]  # 提取标准股票代码
    market_prefix = "sz" if code.startswith("0") else "sh"

    # 📁 日线数据路径（来自 kline 文件夹）
    # Adjusted daily_path construction
    daily_path = f"/content/drive/MyDrive/Colab Notebooks/kline/{prefix}_daily.csv"


    # 📁 资金流数据路径（来自 jiqi_do 文件夹）
    # Adjusted fundflow_path construction
    fundflow_path = f"/content/drive/MyDrive/Colab Notebooks/jiqi_do/jiqi_do/{market_prefix}{prefix}.csv"

    if not os.path.exists(daily_path):
        print(f"⚠️ 缺失日线数据文件：{daily_path}")
        return None, None, None, None, None, None

    try:
        # ✅ 加载日线数据
        df = pd.read_csv(daily_path)
        print(f"📈 原始日线数据（前5行） for {code}:")
        display(df.head()) # Display original daily data head

        # Apply standardization after reading
        df = standardize_columns(df.copy()) # Use .copy() to avoid modifying the original df in place unexpectedly

        # Added check after standardization
        if "Date" not in df.columns:
            print(f"❌ 'Date' column not found in {daily_path} AFTER standardization for {code}. Columns: {df.columns.tolist()}")
            return None, None, None, None, None, None
        else:
            print(f"✅ 'Date' column found after standardization for {code}.")


        # Ensure 'Date' is datetime and set as index, then sort
        df["Date"] = pd.to_datetime(df["Date"])
        df.set_index("Date", inplace=True)
        df.sort_index(inplace=True)

        # Handle duplicate index values by keeping the last entry for each date
        if not df.index.is_unique:
            print(f"⚠️ Duplicate dates found in index for {code}. Keeping the last entry for each date.")
            df = df[~df.index.duplicated(keep='last')].copy()


        # Debugging prints for slicing issue
        if code == '601288':
            print(f"📊 Index before slicing for {code}: {df.index.tolist()[:5]} ... {df.index.tolist()[-5:]}")
            print(f"📊 Index is monotonic increasing: {df.index.is_monotonic_increasing}")
            print(f"📊 Index min/max for {code}: {df.index.min()}, {df.index.max()}")
            print(f"📊 Slice start_date: {backtest_start}, end_date: {backtest_end}")
            # Check if start and end dates are in the index
            print(f"📊 start_date ('{backtest_start}') in index for {code}: {pd.to_datetime(backtest_start) in df.index}")
            print(f"📊 end_date ('{backtest_end}') in index for {code}: {pd.to_datetime(backtest_end) in df.index}")

        # Convert string dates to datetime objects for slicing
        start_date_dt = pd.to_datetime(backtest_start)
        end_date_dt = pd.to_datetime(backtest_end)

        # Use .loc for slicing after ensuring index is datetime and sorted
        # This is the most straightforward approach and should work if duplicates are handled
        df = df.loc[start_date_dt:end_date_dt].copy() # Use .copy() to avoid SettingWithCopyWarning

        if df.empty:
             print(f"⚠️ DataFrame is empty after slicing for {code} with dates {backtest_start} to {backtest_end}.")
             return None, None, None, None, None, None


        print(f"📈 股票 {code} 日线数据加载成功，共 {len(df)} 条记录")

        # ✅ 合并资金流数据（如果存在）
        if os.path.exists(fundflow_path):
            df_fund = pd.read_csv(fundflow_path)
            # Updated fund flow column mapping
            fundflow_column_mapping = {
                "opendate": "Date", # Corrected column name
                "netamount": "主力净流入",
                "ratioamount": "主力净流入比率",
                # Removed "trade" as per user request
            }
            df_fund.rename(columns=fundflow_column_mapping, inplace=True)
            if "turnover" in df_fund.columns:
                df_fund.drop(columns=["turnover"], inplace=True)
            # Drop the 'trade' column if it exists after renaming (as it's not needed)
            if "大单净流入" in df_fund.columns:
                 df_fund.drop(columns=["大单净流入"], inplace=True)
            # Drop '超大单净流入' if it exists (based on previous request to not predict it, also not in fundflow_column_mapping anymore)
            if "超大单净流入" in df_fund.columns:
                 df_fund.drop(columns=["超大单净流入"], inplace=True)

            if "Date" not in df_fund.columns:
                 print(f"❌ 'Date' column not found in fund flow data {fundflow_path} after renaming for {code}. Columns: {df_fund.columns.tolist()}")
                 # Continue without merging fund flow if Date is missing
            else:
                df_fund["Date"] = pd.to_datetime(df_fund["Date"])
                df_fund.set_index("Date", inplace=True)
                df = df.join(df_fund, how="left")
                print(f"💰 资金流数据合并成功 for {code}")
                print(f"📊 合并后的数据（前5行） for {code}:")
                display(df.head()) # Display merged data head

        # ✅ Initialize fund flow columns if they don't exist after merging (in case the file was missing)
        # Moved this logic here to ensure columns exist before process_features
        fundflow_cols_to_check = ["主力净流入", "主力净流入比率"]
        for col in fundflow_cols_to_check:
             if col not in df.columns:
                 df[col] = 0 # Initialize with 0 if column is missing
                 print(f"⚠️ Column '{col}' not found after merging, initialized with 0 for {code}.")


        else:
            print(f"⚠️ 缺失资金流数据文件：{fundflow_path} for {code}")
            # Ensure fund flow columns are initialized even if fundflow_path does not exist
            fundflow_cols_to_check = ["主力净流入", "主力净流入比率"]
            for col in fundflow_cols_to_check:
                 if col not in df.columns:
                     df[col] = 0 # Initialize with 0 if column is missing
                     print(f"⚠️ Column '{col}' not found (fund flow file missing), initialized with 0 for {code}.")


        # ✅ 调用特征工程函数
        # Pass df to process_features
        df, scaled_data, target_regression_data, target_classification_data, scaler, actual_regression_target_names = process_features(df)

        # Return df as well
        return df, scaled_data, target_regression_data, target_classification_data, scaler, actual_regression_target_names

    except Exception as e:
        print(f"❌ 数据处理失败：{code} → {e}")
        # Return None for df as well in case of error
        return None, None, None, None, None, None


# ✅ 构造样本 (Updated to return both regression and classification targets)
def create_sequences(data_scaled, target_regression_data, target_classification_data, window_size=60, pred_size=15):
    X, y_regression, y_classification = [], [], []
    # Ensure enough data exists for the window and prediction size
    if len(data_scaled) < window_size + pred_size:
        print(f"⚠️ Not enough data for sequence creation. Need at least {window_size + pred_size} data points, but only have {len(data_scaled)}.")
        return np.array(X), np.array(y_regression), np.array(y_classification)

    for i in range(len(data_scaled) - window_size - pred_size + 1):
        X.append(data_scaled[i:i+window_size])
        y_regression.append(target_regression_data[i+window_size:i+window_size+pred_size])
        # For classification, we predict the direction for the *next* day
        y_classification.append(target_classification_data[i+window_size:i+window_size+pred_size]) # Predict next 'pred_size' days direction


    return np.array(X), np.array(y_regression), np.array(y_classification)

# ✅ Attention模型 (Updated for multiple outputs)
def build_transformer_model(input_shape, num_regression_targets, pred_size=15):
    inputs = Input(shape=input_shape)
    attn = MultiHeadAttention(num_heads=4, key_dim=32)(inputs, inputs)
    attn = LayerNormalization()(attn)
    x = Dense(64, activation='relu')(attn)
    x = Dropout(0.2)(x)
    x = LayerNormalization()(x)

    # Regression output branch
    regression_output = TimeDistributed(Dense(num_regression_targets))(x)
    regression_output = Lambda(lambda t: t[:, -pred_size:, :], name="regression_output")(regression_output) # Use Lambda to preserve name

    # Classification output branch
    # Predict the direction for each of the next `pred_size` days
    classification_output = TimeDistributed(Dense(1, activation='sigmoid'))(x)
    classification_output = Lambda(lambda t: t[:, -pred_size:, :], name="classification_output")(classification_output) # Use Lambda to preserve name


    # Define the model with multiple outputs
    model = Model(inputs=inputs, outputs=[regression_output, classification_output])

    # Compile with appropriate loss functions and metrics for each output
    model.compile(optimizer=Adam(),
                  loss={'regression_output': MeanSquaredError(), 'classification_output': BinaryCrossentropy()},
                  metrics={'regression_output': MeanAbsoluteError(), 'classification_output': BinaryAccuracy()})

    return model

# ✅ 训练模型 (Updated for multiple outputs)
def train_transformer_model(X_train, y_train_regression, y_train_classification, input_shape, model_path):
    num_regression_targets = y_train_regression.shape[-1] if y_train_regression.shape[0] > 0 else 3 # Determine output size dynamically
    pred_size = y_train_regression.shape[1] if y_train_regression.shape[0] > 0 else 15 # Determine pred size dynamically


    model = build_transformer_model(input_shape, num_regression_targets=num_regression_targets, pred_size=pred_size)

    # Define callbacks for saving the best model based on the total loss or a specific output's loss
    # Here, saving based on total loss
    checkpoint = ModelCheckpoint(model_path, monitor='loss', save_best_only=True, mode='min', verbose=0) # Set verbose to 0 to reduce output

    # Train the model with multiple outputs
    # Pass a dictionary of target arrays corresponding to the model's named outputs
    model.fit(X_train, {"regression_output": y_train_regression, "classification_output": y_train_classification},
              epochs=50,
              batch_size=32,
              callbacks=[checkpoint],
              verbose=0) # Set verbose to 0 to reduce output

    return model


# ✅ 预测并绘图 (Updated for multiple outputs)
# Added df as an argument
def predict_and_plot_targets(model, df, scaled_data, scaler, window_size, stock_code, actual_regression_target_names, output_dir="./charts"):
    # Ensure enough data exists for the last window
    if len(scaled_data) < window_size:
        print(f"⚠️ Not enough data to create the last window for prediction. Need {window_size} data points, but only have {len(scaled_data)}.")
        return

    last_window = scaled_data[-window_size:]
    # Predict returns a list of predictions, one for each output
    predictions = model.predict(np.expand_dims(last_window, axis=0))
    regression_prediction = predictions[0][0] # First output is regression, take the first (and only) sample
    classification_prediction = predictions[1][0] # Second output is classification

    num_regression_targets = regression_prediction.shape[-1]
    pred_size = regression_prediction.shape[0]

    # Inverse transform the regression predictions
    # Need to create a dummy array with the same number of features + regression targets as used during scaling
    num_features_plus_regression_targets = scaled_data.shape[-1]
    dummy_inverse_transform_data = np.zeros((pred_size, num_features_plus_regression_targets))

    # Place the regression predictions in the columns corresponding to the regression targets
    # This assumes the regression targets were the last columns in the scaled data during training
    dummy_inverse_transform_data[:, -num_regression_targets:] = regression_prediction

    recovered_full = scaler.inverse_transform(dummy_inverse_transform_data)
    recovered_regression_targets = recovered_full[:, -num_regression_targets:] # Extract the inverse transformed regression targets


    os.makedirs(output_dir, exist_ok=True)
    today = datetime.today().strftime('%Y-%m-%d')

    # Plot Regression Targets
    for i, name in enumerate(actual_regression_target_names): # Iterate based on actual regression target names
        plt.figure(figsize=(10, 4))
        plt.plot(range(1, pred_size + 1), recovered_regression_targets[:, i], marker='o', label=f"预测{name}")
        plt.title(f"{stock_code} - 未来{pred_size}天{name}预测 (回归)")
        plt.xlabel("未来天数")
        plt.ylabel(name)
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        chart_path = os.path.join(output_dir, f"{stock_code}_{today}_{name}_regression.png")
        plt.savefig(chart_path)
        plt.close()
        print(f"✅ 回归预测图表已保存：{chart_path}")

    # Plot Classification Target (涨跌方向)
    plt.figure(figsize=(10, 4))
    # Convert probabilities to binary predictions (0 or 1)
    predicted_direction = (classification_prediction > 0.5).astype(int)
    # Map 0 to '跌/平' and 1 to '涨' for plotting
    direction_labels = ['跌/平' if d == 0 else '涨' for d in predicted_direction]

    plt.plot(range(1, pred_size + 1), predicted_direction, marker='o', linestyle='-', label="预测涨跌方向")
    # Add text labels for direction
    for i, label in enumerate(direction_labels):
        plt.text(i + 1, predicted_direction[i], label, ha='center', va='bottom')

    plt.title(f"{stock_code} - 未来{pred_size}天涨跌方向预测 (分类)")
    plt.xlabel("未来天数")
    plt.ylabel("方向 (0: 跌/平, 1: 涨)")
    plt.yticks([0, 1], ['跌/平', '涨']) # Set y-ticks to show labels
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    chart_path = os.path.join(output_dir, f"{stock_code}_{today}_direction_classification.png")
    plt.savefig(chart_path)
    plt.close()
    print(f"✅ 涨跌方向预测图表已保存：{chart_path}")

    # ✅ Save predictions to CSV
    # Use df.index.max() from the passed df
    prediction_dates = pd.date_range(start=df.index.max() + pd.Timedelta(days=1), periods=pred_size, freq='B') # Assuming business days for prediction

    # Regression predictions
    regression_df = pd.DataFrame(recovered_regression_targets, index=prediction_dates, columns=actual_regression_target_names)
    regression_csv_path = os.path.join(output_dir, f"{stock_code}_{today}_regression_predictions.csv")
    regression_df.to_csv(regression_csv_path)
    print(f"✅ 回归预测数据已保存：{regression_csv_path}")

    # Classification predictions
    classification_df = pd.DataFrame(predicted_direction, index=prediction_dates, columns=["涨跌方向"])
    classification_csv_path = os.path.join(output_dir, f"{stock_code}_{today}_classification_predictions.csv")
    classification_df.to_csv(classification_csv_path)
    print(f"✅ 涨跌方向预测数据已保存：{classification_csv_path}")


# ✅ 打包图表和CSV文件
def zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip"):
    today_str = datetime.today().strftime("%Y-%m-%d")
    os.makedirs(chart_dir, exist_ok=True) # Ensure chart directory exists before listing
    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for file in os.listdir(chart_dir):
            if (file.endswith(".png") or file.endswith(".csv")) and today_str in file:
                zipf.write(os.path.join(chart_dir, file), arcname=file)
    print(f"📦 今日图表和CSV文件已打包为：{zip_name}")

# ✅ 主流程：加载、处理、训练并预测所有股票
os.makedirs("./models", exist_ok=True)
chart_dir = "./charts"
stock_data = {} # Initialize stock_data here to store processed data

for code in stock_codes:
    # Use the new loading function and capture df
    df, scaled_data, target_regression_data, target_classification_data, scaler, actual_regression_target_names = load_daily_and_fundflow(code, start_date, end_date)

    if scaled_data is not None:
        # Store df along with other data
        stock_data[code] = (df, scaled_data, target_regression_data, target_classification_data, scaler, actual_regression_target_names)
        print(f"✅ 成功加载和处理数据：{code}")
    else:
        print(f"❌ 数据加载或处理失败：{code}")

print(f"\nStock data loaded and processed for: {list(stock_data.keys())}\n")


for code in stock_codes:
    if code not in stock_data:
        print(f"⚠️ 数据未加载或处理，跳过训练：{code}")
        continue

    # Unpack df along with other data
    df, scaled_data, target_regression_data, target_classification_data, scaler, actual_regression_target_names = stock_data[code]
    # Create sequences for both regression and classification targets
    X, y_regression, y_classification = create_sequences(scaled_data, target_regression_data, target_classification_data, window_size=60, pred_size=15)

    print(f"➡️ Processing {code}: X shape = {X.shape}, y_regression shape = {y_regression.shape}, y_classification shape = {y_classification.shape}")

    if X.shape[0] == 0:
        print(f"⚠️ 样本数量不足，跳过训练：{code}")
        continue

    model_path = f"./models/{code}_attention.keras" # Modified model_path
    print(f"✨ 开始训练模型：{code}")
    # Train the model with multiple outputs
    model = train_transformer_model(X, y_regression, y_classification, input_shape=X.shape[1:], model_path=model_path)
    print(f"✅ 模型训练完成：{code}")

    print(f"📈 开始预测和绘图：{code}")
    # Predict and plot both regression and classification targets
    # Pass df to predict_and_plot_targets
    predict_and_plot_targets(model, df, scaled_data, scaler, window_size=60, stock_code=code, actual_regression_target_names=actual_regression_target_names, output_dir=chart_dir)
    print(f"✅ 预测和绘图完成：{code}\n")

# ✅ 打包图表和CSV文件
zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip")
```
