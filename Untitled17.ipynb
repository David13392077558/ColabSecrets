```python
# ✅ 挂载 Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ✅ 导入库
import os, pandas as pd, numpy as np, torch, torch.nn as nn
from sklearn.model_selection import train_test_split

# ✅ 模型结构：双层 LSTM + Dropout + 多目标输出
class MultiTargetLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, forecast_horizon, dropout=0.3):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.dropout = nn.Dropout(dropout)
        self.fc_fundflow = nn.Linear(hidden_size, forecast_horizon)
        self.fc_direction = nn.Linear(hidden_size, forecast_horizon)
        self.fc_pct = nn.Linear(hidden_size, forecast_horizon)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.fc_fundflow(h), self.fc_direction(h), self.fc_pct(h)

# ✅ 配置参数
class Config:
    input_size = 20
    hidden_size = 64
    num_layers = 2
    forecast_horizon = 8
    dropout = 0.3
    lr = 1e-3
    epochs = 100
    window_size = 50
    model_path = "best_lstm_model.pt"

def align_stock_data(minute_df, fundflow_df, daily_df):
    print(f"Minute df initial shape: {minute_df.shape}")
    print(f"Fundflow df initial shape: {fundflow_df.shape}")
    print(f"Daily df initial shape: {daily_df.shape}")

    # ✅ 清洗 minute_df（分钟数据：主力净流入等）
    minute_df = minute_df.dropna(subset=['date', 'time'])
    minute_df['date'] = pd.to_datetime(minute_df['date'], errors='coerce').dt.date
    minute_df['time'] = minute_df['time'].astype(str).str.zfill(4)
    minute_df['datetime'] = pd.to_datetime(minute_df['date'].astype(str) + ' ' + minute_df['time'], format='%Y-%m-%d %H%M', errors='coerce')
    minute_df = minute_df.dropna(subset=['datetime'])
    minute_df['date'] = minute_df['datetime'].dt.date

    # ✅ 聚合分钟数据为日频
    minute_daily = minute_df.groupby('date').agg({
        'main_inflow': 'sum',
        'super_inflow': 'sum',
        'large_inflow': 'sum',
        'total_amount': 'sum'
    }).reset_index()
    print(f"Minute daily shape: {minute_daily.shape}")
    print(f"Minute daily dates: {minute_daily['date'].min()} → {minute_daily['date'].max()}")

    # ✅ 清洗 fundflow_df（资金流数据：r0~r3）
    fundflow_df = fundflow_df.dropna(subset=['date'])
    fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], errors='coerce').dt.date
    fundflow_df = fundflow_df.dropna(subset=['date'])
    print(f"Fundflow df dates: {fundflow_df['date'].min()} → {fundflow_df['date'].max()}")

    # ✅ 清洗 daily_df（日线数据）
    daily_df = daily_df.dropna(subset=['date'])
    daily_df['date'] = pd.to_datetime(daily_df['date'], errors='coerce').dt.date
    daily_df = daily_df.dropna(subset=['date'])
    print(f"Daily df dates: {daily_df['date'].min()} → {daily_df['date'].max()}")

    # ✅ 打印共有日期交集
    common_dates = set(minute_daily['date']) & set(fundflow_df['date']) & set(daily_df['date'])
    print(f"共有交易日数量：{len(common_dates)}")
    if len(common_dates) == 0:
        print("⚠️ 三个数据源没有交集日期，请检查日期范围或格式")

    # ✅ 合并三类数据（按日期）
    df = minute_daily.merge(fundflow_df, on='date', how='inner').merge(daily_df, on='date', how='inner')
    df = df.sort_values('date').reset_index(drop=True)
    print(f"Aligned df shape: {df.shape}")
    return df

# ✅ 构建训练样本
def build_lstm_samples(df, window_size=50, forecast_horizon=8):
    X, y_fundflow, y_direction, y_pct = [], [], [], []
    for i in range(len(df) - window_size - forecast_horizon + 1):
        window_end = i + window_size
        forecast_end = window_end + forecast_horizon

        dates_in_window = df.iloc[i:window_end]['date']
        dates_in_forecast = df.iloc[window_end:forecast_end]['date']

        if len(dates_in_window) == window_size and len(dates_in_forecast) == forecast_horizon and \
           (pd.to_datetime(dates_in_window).diff().dropna().dt.days == 1).all() and \
           (pd.to_datetime(dates_in_forecast).diff().dropna().dt.days == 1).all():

            X.append(df.iloc[i:window_end].drop(columns=['date']).values)
            future = df.iloc[window_end:forecast_end]
            y_fundflow.append(future['主力净流入'].values)
            y_direction.append((future['ChangePct'] > 0).astype(int).values)
            y_pct.append(future['ChangePct'].values)
    return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

# ✅ 训练函数
def train_model(X, y1, y2, y3, config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.lr)

    if len(X) == 0:
        raise ValueError(f"数据不足，无法构建训练样本. 需要至少 {config.window_size + config.forecast_horizon} 个连续的日期数据.")

    X_train, X_val, y1_train, y1_val, y2_train, y2_val, y3_train, y3_val = train_test_split(
        X, y1, y2, y3, test_size=0.2, shuffle=False
    )

    best_val_loss = float('inf')
    for epoch in range(config.epochs):
        model.train()
        inputs = torch.tensor(X_train).float().to(device)
        targets1 = torch.tensor(y1_train).float().to(device)
        targets2 = torch.tensor(y2_train).float().to(device)
        targets3 = torch.tensor(y3_train).float().to(device)

        pred1, pred2, pred3 = model(inputs)
        loss1 = nn.MSELoss()(pred1, targets1)
        loss2 = nn.BCEWithLogitsLoss()(pred2, targets2)
        loss3 = nn.MSELoss()(pred3, targets3)
        loss = loss1 + loss2 + loss3

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            val_inputs = torch.tensor(X_val).float().to(device)
            val_targets1 = torch.tensor(y1_val).float().to(device)
            val_targets2 = torch.tensor(y2_val).float().to(device)
            val_targets3 = torch.tensor(y3_val).float().to(device)

            val_pred1, val_pred2, val_pred3 = model(val_inputs)
            val_loss = (
                nn.MSELoss()(val_pred1, val_targets1) +
                nn.BCEWithLogitsLoss()(val_pred2, val_targets2) +
                nn.MSELoss()(val_pred3, val_targets3)
            )

        print(f"Epoch {epoch}: TrainLoss={loss.item():.4f}, ValLoss={val_loss.item():.4f}")
        if val_loss.item() < best_val_loss:
            best_val_loss = val_loss.item()
            torch.save(model.state_dict(), config.model_path)
            print(f"✅ 保存最优模型于 epoch {epoch}")
```

```python
# 从原始代码块复制必要的函数和配置
from google.colab import drive
import os, pandas as pd, numpy as np, torch, torch.nn as nn
from sklearn.model_selection import train_test_split

# 确保 Drive 已挂载
# drive.mount('/content/drive') # 这行在之前的单元格已经运行过，避免重复挂载

# ✅ 模型结构：双层 LSTM + Dropout + 多目标输出 (从原单元格复制)
class MultiTargetLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, forecast_horizon, dropout=0.3):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.dropout = nn.Dropout(dropout)
        self.fc_fundflow = nn.Linear(hidden_size, forecast_horizon)
        self.fc_direction = nn.Linear(hidden_size, forecast_horizon)
        self.fc_pct = nn.Linear(hidden_size, forecast_horizon)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.fc_fundflow(h), self.fc_direction(h), self.fc_pct(h)

# ✅ 配置参数 (从原单元格复制)
class Config:
    input_size = 20
    hidden_size = 64
    num_layers = 2
    forecast_horizon = 8
    dropout = 0.3
    lr = 1e-3
    epochs = 100
    window_size = 50
    model_path = "best_lstm_model.pt"

# ✅ 对齐股票数据函数 (修复重命名逻辑，并添加更多打印信息)
def align_stock_data(minute_df, fundflow_df, daily_df):
    print(f"Minute df initial shape: {minute_df.shape}")
    print(f"Fundflow df initial shape: {fundflow_df.shape}")
    print(f"Daily df initial shape: {daily_df.shape}")

    # ✅ 清洗 minute_df（分钟数据：主力净流入等）
    minute_df = minute_df.dropna(subset=['date', 'time'])

    # ✅ 打印原始日期和时间样本 (调试用)
    print("✅ 原始分钟数据样本（日期和时间）：")
    print(minute_df[['date', 'time']].head())
    print("-" * 30) # 分隔线方便查看


    # 日期字段是 '2023/9/25' → 转换为 datetime.date
    minute_df['date'] = pd.to_datetime(minute_df['date'], format='%Y/%m/%d', errors='coerce').dt.date

    # 时间字段是 '09:30' → 保持原样
    minute_df['time'] = minute_df['time'].astype(str).str.strip()

    # 拼接为完整时间戳
    minute_df['datetime'] = pd.to_datetime(
        minute_df['date'].dt.strftime('%Y-%m-%d') + ' ' + minute_df['time'],
        format='%Y-%m-%d %H:%M',
        errors='coerce'
    )

    # 丢弃解析失败的行
    minute_df = minute_df.dropna(subset=['datetime'])

    # 提取标准日期字段用于 merge
    minute_df['date'] = minute_df['datetime'].dt.date
    print("✅ 有效分钟数据样本：")
    print(minute_df[['datetime', 'main_inflow']].head())
    print("✅ 有效日期范围：", minute_df['date'].min(), "→", minute_df['date'].max())


    print(f"Minute df after cleaning and datetime conversion: {minute_df.shape}")
    print(f"Minute df dates after cleaning: {minute_df['date'].min()} → {minute_df['date'].max()}")


    # ✅ 聚合分钟数据为日频
    # 确保聚合所需的列存在
    agg_cols = ['main_inflow', 'super_inflow', 'large_inflow', 'total_amount']
    if not all(col in minute_df.columns for col in agg_cols):
         print(f"❌ minute_df 缺少聚合所需的列: {', '.join([col for col in agg_cols if col not in minute_df.columns])}")
         return pd.DataFrame() # 返回空 DataFrame

    minute_daily = minute_df.groupby('date').agg({
        'main_inflow': 'sum',
        'super_inflow': 'sum',
        'large_inflow': 'sum',
        'total_amount': 'sum'
    }).reset_index()
    print(f"Minute daily shape: {minute_daily.shape}")
    print(f"Minute daily dates: {minute_daily['date'].min()} → {minute_daily['date'].max()}")

    # ✅ 清洗 fundflow_df（资金流数据：r0~r3）
    if 'date' not in fundflow_df.columns:
        print("❌ fundflow_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    fundflow_df = fundflow_df.dropna(subset=['date'])
    fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], errors='coerce').dt.date
    fundflow_df = fundflow_df.dropna(subset=['date'])
    print(f"Fundflow df shape after cleaning: {fundflow_df.shape}")
    print(f"Fundflow df dates: {fundflow_df['date'].min()} → {fundflow_df['date'].max()}")

    # ✅ 清洗 daily_df（日线数据）
    if 'date' not in daily_df.columns:
        print("❌ daily_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    daily_df = daily_df.dropna(subset=['date'])
    daily_df['date'] = pd.to_datetime(daily_df['date'], errors='coerce').dt.date
    daily_df = daily_df.dropna(subset=['date'])
    print(f"Daily df shape after cleaning: {daily_df.shape}")
    print(f"Daily df dates: {daily_df['date'].min()} → {daily_df['date'].max()}")

    # ✅ 打印共有日期交集
    common_dates = set(minute_daily['date']) & set(fundflow_df['date']) & set(daily_df['date'])
    print(f"共有交易日数量：{len(common_dates)}")
    if len(common_dates) == 0:
        print("⚠️ 三个数据源没有交集日期，请检查日期范围或格式")
        return pd.DataFrame() # 返回空 DataFrame

    # ✅ 合并三类数据（按日期）
    # 先合并 minute_daily 和 fundflow_df
    df_merged_1 = minute_daily.merge(fundflow_df, on='date', how='inner')
    print(f"Shape after merging minute_daily and fundflow_df: {df_merged_1.shape}")

    # 再合并 df_merged_1 和 daily_df
    df = df_merged_1.merge(daily_df, on='date', how='inner')

    df = df.sort_values('date').reset_index(drop=True)
    print(f"Aligned df shape: {df.shape}")
    print(f"Aligned df columns: {df.columns.tolist()}") # 打印最终合并后的列名
    return df

# ✅ 构建训练样本 (从原单元格复制)
def build_lstm_samples(df, window_size=50, forecast_horizon=8):
    X, y_fundflow, y_direction, y_pct = [], [], [], []
    # 确保 df 不为空且有足够的行
    if len(df) < window_size + forecast_horizon:
        print(f"⚠️ 数据不足，无法构建训练样本. 需要至少 {window_size + forecast_horizon} 行数据，但只有 {len(df)} 行.")
        return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

    # 检查构建样本所需的列是否存在
    required_cols_for_samples = ['主力净流入', 'ChangePct']
    if not all(col in df.columns for col in required_cols_for_samples):
         print(f"❌ 构建样本所需的列缺失: {', '.join([col for col in required_cols_for_samples if col not in df.columns])}")
         return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


    for i in range(len(df) - window_size - forecast_horizon + 1):
        window_end = i + window_size
        forecast_end = window_end + forecast_horizon

        # 检查窗口和预测期内的数据是否完整
        if forecast_end > len(df):
            break # 确保不会超出数据范围

        window_dates = df.iloc[i:window_end]['date']
        forecast_dates = df.iloc[window_end:forecast_end]['date']

        # 检查日期是否连续 (虽然您之前有这个逻辑，但可能会因为非交易日导致不连续，这里先简化检查长度)
        if len(window_dates) == window_size and len(forecast_dates) == forecast_horizon:

            # 删除 date 列之前，确保其他列是数值型
            X_window = df.iloc[i:window_end].drop(columns=['date']).values
            if not np.issubdtype(X_window.dtype, np.number):
                print(f"⚠️ 窗口数据包含非数值类型，请检查数据清洗和列选择: {X_window.dtype}")
                # 可以打印出非数值列的名称
                non_numeric_cols = df.iloc[i:window_end].drop(columns=['date']).select_dtypes(exclude=np.number).columns.tolist()
                print(f"非数值列: {non_numeric_cols}")
                continue # 跳过当前样本

            X.append(X_window)

            future = df.iloc[window_end:forecast_end]

            # 确保未来数据的列存在且是数值型
            if not all(col in future.columns for col in ['主力净流入', 'ChangePct']):
                 print(f"❌ 未来数据所需列缺失: {', '.join([col for col in ['主力净流入', 'ChangePct'] if col not in future.columns])}")
                 continue

            if not np.issubdtype(future['主力净流入'].dtype, np.number) or \
               not np.issubdtype(future['ChangePct'].dtype, np.number):
               print(f"⚠️ 未来数据包含非数值类型，请检查数据清洗: 主力净流入({future['主力净流入'].dtype}), ChangePct({future['ChangePct'].dtype})")
               continue


            y_fundflow.append(future['主力净流入'].values)
            y_direction.append((future['ChangePct'] > 0).astype(int).values)
            y_pct.append(future['ChangePct'].values)

    print(f"成功构建样本数量: {len(X)}")
    return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


# ✅ 训练函数 (从原单元格复制)
def train_model(X, y1, y2, y3, config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    # input_size 需要根据实际对齐后的 DataFrame 列数确定，这里需要动态获取
    # Assuming X is already shaped as (num_samples, window_size, input_size)
    if X.shape[0] == 0:
        raise ValueError(f"数据不足，无法构建训练样本. 需要至少 {config.window_size + config.forecast_horizon} 个连续的日期数据.")

    current_input_size = X.shape[2]
    print(f"模型输入的特征数量 (input_size): {current_input_size}")

    # 更新 Config 中的 input_size
    config.input_size = current_input_size

    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.lr)

    X_train, X_val, y1_train, y1_val, y2_train, y2_val, y3_train, y3_val = train_test_split(
        X, y1, y2, y3, test_test=0.2, shuffle=False
    )

    best_val_loss = float('inf')
    for epoch in range(config.epochs):
        model.train()
        inputs = torch.tensor(X_train).float().to(device)
        targets1 = torch.tensor(y1_train).float().to(device)
        targets2 = torch.tensor(y2_train).float().to(device)
        targets3 = torch.tensor(y3_train).float().to(device)

        pred1, pred2, pred3 = model(inputs)
        loss1 = nn.MSELoss()(pred1, targets1)
        loss2 = nn.BCEWithLogitsLoss()(pred2, targets2.float()) # BCEWithLogitsLoss 需要 float 类型的 target
        loss3 = nn.MSELoss()(pred3, targets3)
        loss = loss1 + loss2 + loss3

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            val_inputs = torch.tensor(X_val).float().to(device)
            val_targets1 = torch.tensor(y1_val).float().to(device)
            val_targets2 = torch.tensor(y2_val).float().to(device)
            val_targets3 = torch.tensor(y3_val).float().to(device)

            val_pred1, val_pred2, val_pred3 = model(val_inputs)
            val_loss = (
                nn.MSELoss()(val_pred1, val_targets1) +
                nn.BCEWithLogitsLoss()(val_pred2, val_targets2.float()) + # BCEWithLogitsLoss 需要 float 类型的 target
                nn.MSELoss()(val_pred3, val_targets3)
            )

        print(f"Epoch {epoch}: TrainLoss={loss.item():.4f}, ValLoss={val_loss.item():.4f}")
        if val_loss.item() < best_val_loss:
            best_val_loss = val_loss.item()
            torch.save(model.state_dict(), config.model_path)
            print(f"✅ 保存最优模型于 epoch {epoch}")

# ✅ 加载模型 (从原单元格复制)
def load_model(config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
     # input_size 需要根据实际对齐后的 DataFrame 列数确定
    # Assuming a model was trained and saved, we need to know the input_size used during training.
    # A robust way is to save config along with the model state dict, or infer from data.
    # For now, assuming config.input_size is set based on aligned data before calling train_model
    # If the model is loaded for prediction after training in the same run, config.input_size will be correct.
    # If loading in a new session, need to ensure config.input_size matches the trained model.
    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    model.load_state_dict(torch.load(config.model_path, map_location=device))
    model.eval()
    return model

# ✅ 构建最新样本（最近50天） (从原单元格复制，添加数据检查)
def prepare_latest_sample(df, window_size=50):
    if len(df) < window_size:
        print(f"⚠️ 数据不足，无法构建最新预测样本. 需要至少 {window_size} 行数据，但只有 {len(df)} 行.")
        return None # 返回 None 表示无法构建样本

    latest = df.tail(window_size).drop(columns=['date'])

    # 检查是否包含非数值列
    if not np.issubdtype(latest.values.dtype, np.number):
        print(f"⚠️ 最新样本数据包含非数值类型，请检查数据清洗和列选择: {latest.values.dtype}")
        non_numeric_cols = latest.select_dtypes(exclude=np.number).columns.tolist()
        print(f"非数值列: {non_numeric_cols}")
        return None # 返回 None 表示无法构建样本


    X = latest.values.reshape(1, window_size, -1)
    return torch.tensor(X).float()

# ✅ 预测未来8天 (从原单元格复制)
def predict_future(model, X, config):
    if X is None: # 检查样本是否成功构建
        return None, None, None

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    X = X.to(device)
    with torch.no_grad():
        pred_fundflow, pred_direction, pred_pct = model(X)
        pred_fundflow = pred_fundflow.cpu().numpy().flatten()
        # 应用 sigmoid 并进行阈值判断
        pred_direction = (torch.sigmoid(pred_direction) > 0.5).int().cpu().numpy().flatten()
        pred_pct = pred_pct.cpu().numpy().flatten()
    return pred_fundflow, pred_direction, pred_pct

# ✅ 打印预测结果 (从原单元格复制)
def print_predictions(pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # 检查预测是否成功
        print("❌ 无法进行预测，请检查数据和样本构建.")
        return
    print("📊 未来8天预测结果：")
    for i in range(8):
        direction = "📈 涨" if pred_direction[i] == 1 else "📉 跌"
        print(f"Day {i+1}: 主力净流入={pred_fundflow[i]:.2f}, 涨跌方向={direction}, 涨跌幅={pred_pct[i]:.2%}")

# ✅ 保存预测结果为 CSV (从原单元格复制)
def save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # 检查预测是否成功
        print("❌ 无法保存预测结果，因为预测失败.")
        return
    df_pred = pd.DataFrame({
        'Day': [f'Day {i+1}' for i in range(8)],
        '主力净流入': pred_fundflow,
        '涨跌方向': ['涨' if d == 1 else '跌' for d in pred_direction],
        '涨跌幅': pred_pct
    })
    save_path = f'/content/drive/MyDrive/davidliang/{prefix}_预测结果.csv'
    df_pred.to_csv(save_path, index=False)
    print(f"✅ 预测完成，结果已保存为 CSV 文件: {save_path}")


# ✅ 批量处理多个个股 (从原单元格复制，并使用修复后的 align_stock_data)
stock_prefixes = ['002364_中恒电气']

# ✅ 字段统一为英文（重命名字典，从原单元格复制）
# ✅ 分钟数据（002364_中恒电气_合并.csv）
minute_rename = {
    '日期': 'date',
    '时间': 'time',
    '主力净流入': 'main_inflow',
    '超大单净流入': 'super_inflow',
    '大单净流入': 'large_inflow',
    '总成交额': 'total_amount'
}
# ✅ 资金流数据（sz002364.csv）
fundflow_rename = {
    'opendate': 'date',
    'trade': 'time',
    'changeratio': 'change_ratio',
    'turnover': 'turnover',
    'netamount': 'net_amount',
    'ratioamount': 'ratio_amount',
    'r0': 'r0',
    'r1': 'r1',
    'r2': 'r2',
    'r3': 'r3',
    'r0_net': 'r0_net',
    'r1_net': 'r1_net',
    'r2_net': 'r2_net',
    'r3_net': 'r3_net'
}
# ✅  daily_df（日线数据）
daily_rename = {
    'Date': 'date',
    '股票代码': 'stock_code',
    'Open': 'open',
    'Close': 'close',
    'High': 'high',
    'Low': 'low',
    'Volume': 'volume',
    'Amount': 'amount',
    'Amplitude': 'amplitude',
    'ChangePct': 'change_pct',
    'ChangeAmt': 'change_amt',
    'Turnover': 'turnover_rate',
    'Code': 'code'
}


for prefix in stock_prefixes:
    try:
        print(f"\n🚀 开始处理：{prefix}")

        # 构建文件路径 - ✅ 修正变量命名：确保数据角色正确
        fundflow_path = f'/content/drive/MyDrive/davidliang/sz{prefix[0:6]}.csv'  # 资金流数据（r0~r3）
        minute_path = f'/content/drive/MyDrive/davidliang/{prefix}_合并.csv'  # 分钟数据（主力净流入）
        daily_path = f'/content/drive/MyDrive/davidliang/{prefix}_daily.csv'


        print(f"加载文件: {minute_path}")
        minute_df_raw = pd.read_csv(minute_path)
        print(f"加载文件: {fundflow_path}")
        fundflow_df_raw = pd.read_csv(fundflow_path)
        print(f"加载文件: {daily_path}")
        daily_df_raw = pd.read_csv(daily_path)

        # 应用重命名 - ✅ **这里是之前修改的地方，确保对应关系正确**
        minute_df = minute_df_raw.rename(columns=minute_rename)
        fundflow_df = fundflow_df_raw.rename(columns=fundflow_rename)
        daily_df = daily_df_raw.rename(columns=daily_rename)


        # 字段检查
        print("原始分钟数据字段：", minute_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后分钟数据字段：", minute_df.columns.tolist()) # 打印重命名后列名
        print("原始资金流数据字段：", fundflow_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后资金流数据字段：", fundflow_df.columns.tolist()) # 打印重命名后列名
        print("原始日线数据字段：", daily_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后日线数据字段：", daily_df.columns.tolist()) # 打印重命名后列名


        # 数据对齐与样本构建
        print("\n--- 开始数据对齐 ---")
        # ✅ 将重命名后的 DataFrame 传递给 align_stock_data
        df_aligned = align_stock_data(minute_df, fundflow_df, daily_df)
        print("--- 数据对齐结束 ---")


        # 检查对齐后的数据
        if df_aligned.empty:
            print(f"❌ 数据对齐失败，跳过后续步骤处理 {prefix}")
            continue # 跳过当前股票的处理

        required_rows = 50 + 8 # window_size + forecast_horizon
        if len(df_aligned) < required_rows:
            raise ValueError(f"数据不足，无法构建训练样本. 需要至少 {required_rows} 行数据，但只有 {len(df_aligned)} 行.")

        print("\n✅ 正在构建训练样本...")
        X, y1, y2, y3 = build_lstm_samples(df_aligned, window_size=50, forecast_horizon=8)

        if X.shape[0] == 0:
            print(f"❌ 样本构建失败，跳过后续步骤处理 {prefix}")
            continue # 跳过当前股票的处理

        # 训练模型
        config = Config()
        config.model_path = f"/content/drive/MyDrive/davidliang/{prefix}_best_model.pt"
        print("✅ 开始训练模型...")
        train_model(X, y1, y2, y3, config)
        print("✅ 模型训练完成，开始预测...")


        # 预测并输出
        print("\n✅ 正在加载模型并进行预测...")
        model = load_model(config)
        X_latest = prepare_latest_sample(df_aligned, window_size=50)

        pred_fundflow, pred_direction, pred_pct = predict_future(model, X_latest, config)

        print_predictions(pred_fundflow, pred_direction, pred_pct)

        # 保存结果
        save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct)


    except Exception as e:
        print(f"❌ 处理 {prefix} 时出错：{e}")
```

```python
# 从头开始，包含所有必要的导入、函数定义和主逻辑

# ✅ 导入库
from google.colab import drive
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split

# ✅ 挂载 Google Drive
# 请确保您已经挂载了 Google Drive，如果之前已经运行成功，可以注释掉下一行
# drive.mount('/content/drive')

# ✅ 模型结构：双层 LSTM + Dropout + 多目标输出
class MultiTargetLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, forecast_horizon, dropout=0.3):
        super().__init__()
        # 确保 dropout 参数在 LSTM 层只应用于非最后一层
        lstm_dropout = dropout if num_layers > 1 else 0
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=lstm_dropout)
        self.dropout = nn.Dropout(dropout) # 在 LSTM 输出后应用 dropout
        self.fc_fundflow = nn.Linear(hidden_size, forecast_horizon)
        self.fc_direction = nn.Linear(hidden_size, forecast_horizon)
        self.fc_pct = nn.Linear(hidden_size, forecast_horizon)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.fc_fundflow(h), self.fc_direction(h), self.fc_pct(h)


# ✅ 配置参数
class Config:
    input_size = 20 # 这个会在 train_model 中根据数据列数动态更新
    hidden_size = 64
    num_layers = 2
    forecast_horizon = 8
    dropout = 0.3
    lr = 1e-3
    epochs = 100
    window_size = 50
    model_path = "best_lstm_model.pt" # 会根据股票前缀更新

# ✅ 对齐股票数据函数
def align_stock_data(minute_df, fundflow_df, daily_df):
    print(f"Minute df initial shape: {minute_df.shape}")
    print(f"Fundflow df initial shape: {fundflow_df.shape}")
    print(f"Daily df initial shape: {daily_df.shape}")

    # ✅ 清洗 minute_df（分钟数据：主力净流入等）
    minute_df = minute_df.dropna(subset=['date', 'time']).copy() # 添加 .copy()

    # ✅ 打印原始日期和时间样本 (调试用)
    print("✅ 原始分钟数据样本（日期和时间）：")
    print(minute_df[['date', 'time']].head())
    print("-" * 30) # 分隔线方便查看

    # ✅ 更稳健的日期和时间分别解析和组合
    try:

        # 根据之前打印的原始样本，日期是 YYYY-MM-DD 格式
        minute_df['date_parsed'] = pd.to_datetime(minute_df['date'], errors='coerce', format='%Y-%m-%d')
        print("✅ 日期列初步解析成功")


        # 需要先转为字符串，并确保没有异常值或非时间格式导致 to_timedelta 失败
        minute_df['time_str'] = minute_df['time'].astype(str).str.strip()

        # 将时间字符串转换为 timedelta

        # 然后提取其 time 部分
        minute_df['time_parsed_dt'] = pd.to_datetime(minute_df['time_str'], format='%H:%M', errors='coerce').dt.time
        print("✅ 时间列初步解析成功")


        # 组合日期 (datetime64[ns]) 和 时间 (datetime.time 对象)


        # 使用 apply + combine 组合，对每一行操作
        # 需要先处理解析失败的 NaT/NaN
        valid_indices = minute_df['date_parsed'].notna() & minute_df['time_parsed_dt'].notna()
        minute_df_valid = minute_df.loc[valid_indices].copy() # 只处理有效行

        if not minute_df_valid.empty:
             # 使用 apply 和 pd.Timestamp.combine
             minute_df.loc[valid_indices, 'datetime'] = minute_df_valid.apply(
                 lambda row: pd.Timestamp.combine(row['date_parsed'].date(), row['time_parsed_dt']),
                 axis=1
             )
             print(f"✅ 成功组合 {len(minute_df_valid)} 行的日期和时间")
        else:
            print("⚠️ 没有成功解析的日期或时间对，无法组合 datetime.")
            minute_df['datetime'] = pd.NaT # 如果没有有效行，datetime 列全为 NaT


    except Exception as e:
        print(f"❌ 日期或时间解析组合失败：{e}")

        return pd.DataFrame() # 解析失败，返回空 DataFrame

    # 丢弃解析失败的行 (datetime 为 NaT 的行)
    minute_df = minute_df.dropna(subset=['datetime']).copy() # 添加 .copy()

    # 提取标准日期字段用于 merge (从解析成功的 datetime 中提取)
    minute_df['date'] = minute_df['datetime'].dt.date

    # ✅ 验证清洗和解析结果
    print("✅ 有效分钟数据样本：")
    if not minute_df.empty:
        print(minute_df[['datetime', 'main_inflow']].head())
        print("✅ 有效日期范围：", minute_df['date'].min(), "→", minute_df['date'].max())
    else:
        print("⚠️ 清洗和日期时间转换后，分钟数据 DataFrame 为空.")

    print(f"Minute df after cleaning and datetime conversion: {minute_df.shape}")
    if not minute_df.empty:
        print(f"Minute df dates after cleaning: {minute_df['date'].min()} → {minute_df['date'].max()}")
    else:
         print(f"Minute df dates after cleaning: 无效日期范围")

    # 清理不再需要的中间列
    minute_df = minute_df.drop(columns=['date_parsed', 'time_str', 'time_parsed_dt'])

    # ✅ 聚合分钟数据为日频 (这部分保持不变，因为输入是 minute_df)
    # 确保聚合所需的列存在且分钟数据不为空
    agg_cols = ['main_inflow', 'super_inflow', 'large_inflow', 'total_amount']
    # 检查 minute_df 在进行聚合前是否为空或者缺少列
    if minute_df.empty or not all(col in minute_df.columns for col in agg_cols):
         print(f"❌ 无法聚合分钟数据，DataFrame 为空或缺少聚合所需的列: {', '.join([col for col in agg_cols if col not in minute_df.columns])}")
         # 检查 minute_df 的当前列名，以防是重命名问题或其他列丢失
         print("当前 minute_df 列名:", minute_df.columns.tolist())
         return pd.DataFrame() # 返回空 DataFrame


    minute_daily = minute_df.groupby('date').agg({
        'main_inflow': 'sum',
        'super_inflow': 'sum',
        'large_inflow': 'sum',
        'total_amount': 'sum'
    }).reset_index()
    print(f"Minute daily shape: {minute_daily.shape}")
    if not minute_daily.empty:
        print(f"Minute daily dates: {minute_daily['date'].min()} → {minute_daily['date'].max()}")
    else:
        print(f"Minute daily dates: 无效日期范围")


    # ✅ 清洗 fundflow_df（资金流数据：r0~r3）
    if 'date' not in fundflow_df.columns:
        print("❌ fundflow_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    fundflow_df = fundflow_df.dropna(subset=['date']).copy() # 添加 .copy()

    try:
        fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], format='%Y-%m-%d', errors='coerce').dt.date
        print("✅ Fundflow df 日期用 %Y-%m-%d 格式解析成功")
    except Exception as e:
         print(f"⚠️ Fundflow df 日期用 %Y-%m-%d 格式解析失败: {e}. 尝试推断格式.")
         fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], errors='coerce').dt.date # 依赖推断
         print("✅ Fundflow df 日期用推断格式解析完成 (可能包含 NaT)")

    fundflow_df = fundflow_df.dropna(subset=['date']).copy() # 添加 .copy()
    print(f"Fundflow df shape after cleaning: {fundflow_df.shape}")
    if not fundflow_df.empty:
        print(f"Fundflow df dates: {fundflow_df['date'].min()} → {fundflow_df['date'].max()}")
    else:
        print(f"Fundflow df dates: 无效日期范围")


    # ✅ 清洗 daily_df（日线数据）
    if 'date' not in daily_df.columns:
        print("❌ daily_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    daily_df = daily_df.dropna(subset=['date']).copy() # 添加 .copy()
    # 尝试用多种可能的格式解析 daily_df 的日期
    try:
        daily_df['date'] = pd.to_datetime(daily_df['date'], format='%Y-%m-%d', errors='coerce').dt.date
        print("✅ Daily df 日期用 %Y-%m-%d 格式解析成功")
    except Exception as e:
        print(f"⚠️ Daily df 日期用 %Y-%m-%d 格式解析失败: {e}. 尝试推断格式.")
        daily_df['date'] = pd.to_datetime(daily_df['date'], errors='coerce').dt.date # 依赖推断
        print("✅ Daily df 日期用推断格式解析完成 (可能包含 NaT)")

    daily_df = daily_df.dropna(subset=['date']).copy() # 添加 .copy()
    print(f"Daily df shape after cleaning: {daily_df.shape}")
    if not daily_df.empty:
        print(f"Daily df dates: {daily_df['date'].min()} → {daily_df['date'].max()}")
    else:
         print(f"Daily df dates: 无效日期范围")


    # ✅ 打印共有日期交集
    common_dates = set()
    # 确保 DataFrame 不为空再尝试获取 date 列
    if not minute_daily.empty and not fundflow_df.empty and not daily_df.empty:
         # 确保 date 列是 date/datetime 类型，并且不包含 NaT
         if pd.api.types.is_datetime64_any_dtype(minute_daily['date']) or isinstance(minute_daily['date'].iloc[0], pd._libs.tslibs.timestamps.Timestamp) or isinstance(minute_daily['date'].iloc[0], dt.date):
             minute_dates = set(minute_daily['date'].dropna()) # 移除 NaT
         else:
             print("⚠️ minute_daily 的 date 列类型异常，无法获取日期集合.")
             minute_dates = set()

         if pd.api.types.is_datetime64_any_dtype(fundflow_df['date']) or isinstance(fundflow_df['date'].iloc[0], pd._libs.tslibs.timestamps.Timestamp) or isinstance(fundflow_df['date'].iloc[0], dt.date):
              fundflow_dates = set(fundflow_df['date'].dropna()) # 移除 NaT
         else:
             print("⚠️ fundflow_df 的 date 列类型异常，无法获取日期集合.")
             fundflow_dates = set()

         if pd.api.types.is_datetime64_any_dtype(daily_df['date']) or isinstance(daily_df['date'].iloc[0], pd._libs.tslibs.timestamps.Timestamp) or isinstance(daily_df['date'].iloc[0], dt.date):
              daily_dates = set(daily_df['date'].dropna()) # 移除 NaT
         else:
             print("⚠️ daily_df 的 date 列类型异常，无法获取日期集合.")
             daily_dates = set()

         common_dates = minute_dates & fundflow_dates & daily_dates


    print(f"共有交易日数量：{len(common_dates)}")
    if len(common_dates) == 0:
        print("⚠️ 三个数据源没有交集日期，请检查日期范围或格式")
        return pd.DataFrame() # 返回空 DataFrame

    # ✅ 合并三类数据（按日期）
    # 先合并 minute_daily 和 fundflow_df
    df_merged_1 = minute_daily.merge(fundflow_df, on='date', how='inner')
    print(f"Shape after merging minute_daily and fundflow_df: {df_merged_1.shape}")

    # 再合并 df_merged_1 和 daily_df
    df = df_merged_1.merge(daily_df, on='date', how='inner')

    df = df.sort_values('date').reset_index(drop=True)
    print(f"Aligned df shape: {df.shape}")
    if not df.empty:
        print(f"Aligned df columns: {df.columns.tolist()}") # 打印最终合并后的列名
    else:
        print("Aligned df columns: 合并后 DataFrame 为空")

    return df

# ✅ 构建训练样本
def build_lstm_samples(df, window_size=50, forecast_horizon=8):
    X, y_fundflow, y_direction, y_pct = [], [], [], []
    # 确保 df 不为空且有足够的行
    if df.empty or len(df) < window_size + forecast_horizon:
        print(f"⚠️ 数据不足，无法构建训练样本. 需要至少 {window_size + forecast_horizon} 行数据，但只有 {len(df)} 行.")
        return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

    # 检查构建样本所需的列是否存在
    required_cols_for_samples = ['main_inflow', 'change_pct'] # 使用重命名后的英文列名
    if not all(col in df.columns for col in required_cols_for_samples):
         print(f"❌ 构建样本所需的列缺失: {', '.join([col for col in required_cols_for_samples if col not in df.columns])}")
         return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


    for i in range(len(df) - window_size - forecast_horizon + 1):
        window_end = i + window_size
        forecast_end = window_end + forecast_horizon

        # 检查窗口和预测期内的数据是否完整
        if forecast_end > len(df):
            break # 确保不会超出数据范围

        window_dates = df.iloc[i:window_end]['date']
        forecast_dates = df.iloc[window_end:forecast_end]['date']



        if len(window_dates) == window_size and len(forecast_dates) == forecast_horizon:

            # 删除 date 列之前，确保其他列是数值型
            X_window = df.iloc[i:window_end].drop(columns=['date', 'datetime', 'time']).values # 排除非数值列
            if not np.issubdtype(X_window.dtype, np.number):
                print(f"⚠️ 窗口数据包含非数值类型，请检查数据清洗和列选择: {X_window.dtype}")
                non_numeric_cols = df.iloc[i:window_end].drop(columns=['date', 'datetime', 'time']).select_dtypes(exclude=np.number).columns.tolist()
                print(f"非数值列: {non_numeric_cols}")
                continue # 跳过当前样本

            X.append(X_window)

            future = df.iloc[window_end:forecast_end]

            # 确保未来数据的列存在且是数值型
            if not all(col in future.columns for col in required_cols_for_samples):
                 print(f"❌ 未来数据所需列缺失: {', '.join([col for col in required_cols_for_samples if col not in future.columns])}")
                 continue

            if not np.issubdtype(future['main_inflow'].dtype, np.number) or \
               not np.issubdtype(future['change_pct'].dtype, np.number): # 使用重命名后的列名
               print(f"⚠️ 未来数据包含非数值类型，请检查数据清洗: main_inflow({future['main_inflow'].dtype}), change_pct({future['change_pct'].dtype})")
               continue


            y_fundflow.append(future['main_inflow'].values) # 使用重命名后的列名
            y_direction.append((future['change_pct'] > 0).astype(int).values) # 使用重命名后的列名
            y_pct.append(future['change_pct'].values) # 使用重命名后的列名

    print(f"成功构建样本数量: {len(X)}")
    # 检查 input_size 是否匹配样本的特征数量
    if len(X) > 0 and X[0].shape[1] != df.shape[1] - 3: # 减去 date, datetime, time 三列
        print(f"❌ 构建的样本特征数量 ({X[0].shape[1]}) 与预期数量 ({df.shape[1] - 3}) 不匹配. 请检查 build_lstm_samples 中的列选择.")


    return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


# ✅ 训练函数
def train_model(X, y1, y2, y3, config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    # input_size 需要根据实际对齐后的 DataFrame 列数确定，这里需要动态获取

    if X.shape[0] == 0:
        raise ValueError(f"数据不足，无法构建训练样本. 需要至少 {config.window_size + config.forecast_horizon} 个连续的日期数据.")

    current_input_size = X.shape[2]
    print(f"模型输入的特征数量 (input_size): {current_input_size}")

    # 更新 Config 中的 input_size
    config.input_size = current_input_size

    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.lr)

    # 确保 y1, y2, y3 也有足够的样本数量
    if len(X) != len(y1) or len(X) != len(y2) or len(X) != len(y3):
         print(f"❌ 训练样本数量 ({len(X)}) 与目标样本数量 ({len(y1)}, {len(y2)}, {len(y3)}) 不匹配. 请检查 build_lstm_samples 函数.")
         raise ValueError("训练样本数量与目标样本数量不匹配")

    X_train, X_val, y1_train, y1_val, y2_train, y2_val, y3_train, y3_val = train_test_split(
        X, y1, y2, y3, test_size=0.2, shuffle=False
    )

    best_val_loss = float('inf')
    for epoch in range(config.epochs):
        model.train()
        inputs = torch.tensor(X_train).float().to(device)
        targets1 = torch.tensor(y1_train).float().to(device)
        targets2 = torch.tensor(y2_train).float().to(device)
        targets3 = torch.tensor(y3_train).float().to(device)

        pred1, pred2, pred3 = model(inputs)
        loss1 = nn.MSELoss()(pred1, targets1)
        loss2 = nn.BCEWithLogitsLoss()(pred2, targets2.float()) # BCEWithLogitsLoss 需要 float 类型的 target
        loss3 = nn.MSELoss()(pred3, targets3)
        loss = loss1 + loss2 + loss3

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            val_inputs = torch.tensor(X_val).float().to(device)
            val_targets1 = torch.tensor(y1_val).float().to(device)
            val_targets2 = torch.tensor(y2_val).float().to(device)
            val_targets3 = torch.tensor(y3_val).float().to(device)

            val_pred1, val_pred2, val_pred3 = model(val_inputs)
            val_loss = (
                nn.MSELoss()(val_pred1, val_targets1) +
                nn.BCEWithLogitsLoss()(val_pred2, val_targets2.float()) + # BCEWithLogitsLoss 需要 float 类型的 target
                nn.MSELoss()(val_pred3, val_targets3)
            )

        print(f"Epoch {epoch}: TrainLoss={loss.item():.4f}, ValLoss={val_loss.item():.4f}")
        if val_loss.item() < best_val_loss:
            best_val_loss = val_loss.item()
            torch.save(model.state_dict(), config.model_path)
            print(f"✅ 保存最优模型于 epoch {epoch}")

# ✅ 加载模型
def load_model(config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
     # input_size 需要根据实际对齐后的 DataFrame 列数确定

    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    # 检查模型文件是否存在
    if not os.path.exists(config.model_path):
        print(f"❌ 模型文件不存在: {config.model_path}. 请先训练模型.")
        return None # 返回 None 表示加载失败

    model.load_state_dict(torch.load(config.model_path, map_location=device))
    model.eval()
    return model

# ✅ 构建最新样本（最近50天）
def prepare_latest_sample(df, window_size=50):
    if df.empty or len(df) < window_size:
        print(f"⚠️ 数据不足，无法构建最新预测样本. 需要至少 {window_size} 行数据，但只有 {len(df)} 行.")
        return None # 返回 None 表示无法构建样本

    # 删除 date, datetime, time 列 before converting to values
    latest = df.tail(window_size).drop(columns=['date', 'datetime', 'time'])

    # 检查是否包含非数值列
    if not np.issubdtype(latest.values.dtype, np.number):
        print(f"⚠️ 最新样本数据包含非数值类型，请检查数据清洗和列选择: {latest.values.dtype}")
        non_numeric_cols = latest.select_dtypes(exclude=np.number).columns.tolist()
        print(f"非数值列: {non_numeric_cols}")
        return None # 返回 None 表示无法构建样本


    X = latest.values.reshape(1, window_size, -1)
    return torch.tensor(X).float()

# ✅ 预测未来8天
def predict_future(model, X, config):
    if model is None or X is None: # 检查模型和样本是否成功构建
        print("❌ 无法进行预测，模型或样本构建失败.")
        return None, None, None

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    X = X.to(device)
    with torch.no_grad():
        pred_fundflow, pred_direction, pred_pct = model(X)
        pred_fundflow = pred_fundflow.cpu().numpy().flatten()
        pred_direction = (torch.sigmoid(pred_direction) > 0.5).int().cpu().numpy().flatten()
        pred_pct = pred_pct.cpu().numpy().flatten()
    return pred_fundflow, pred_direction, pred_pct

# ✅ 打印预测结果
def print_predictions(pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # 检查预测是否成功
        print("❌ 无法进行预测，请检查数据和样本构建.")
        return
    print("📊 未来8天预测结果：")
    for i in range(8):
        direction = "📈 涨" if pred_direction[i] == 1 else "📉 跌"
        print(f"Day {i+1}: 主力净流入={pred_fundflow[i]:.2f}, 涨跌方向={direction}, 涨跌幅={pred_pct[i]:.2%}")

# ✅ 保存预测结果为 CSV
def save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # 检查预测是否成功
        print("❌ 无法保存预测结果，因为预测失败.")
        return
    df_pred = pd.DataFrame({
        'Day': [f'Day {i+1}' for i in range(8)],
        '主力净流入': pred_fundflow,
        '涨跌方向': ['涨' if d == 1 else '跌' for d in pred_direction],
        '涨跌幅': pred_pct
    })
    save_path = f'/content/drive/MyDrive/davidliang/{prefix}_预测结果.csv'
    try:
        df_pred.to_csv(save_path, index=False)
        print(f"✅ 预测完成，结果已保存为 CSV 文件: {save_path}")
    except Exception as e:
        print(f"❌ 保存预测结果到 CSV 文件失败: {e}")


# ✅ 批量处理多个个股
stock_prefixes = ['002364_中恒电气'] # 您可以在这里添加更多股票前缀

# ✅ 分钟数据（002364_中恒电气_合并.csv）
minute_rename = {
    '日期': 'date',
    '时间': 'time',
    '主力净流入': 'main_inflow',
    '超大单净流入': 'super_inflow',
    '大单净流入': 'large_inflow',
    '总成交额': 'total_amount'
}
# ✅ 资金流数据（sz002364.csv）
fundflow_rename = {
    'opendate': 'date',
    'trade': 'time', # 注意：这里 time 列在 fundflow_df 中可能代表其他含义，合并后需要确认其用途或丢弃
    'changeratio': 'change_ratio',
    'turnover': 'turnover',
    'netamount': 'net_amount',
    'ratioamount': 'ratio_amount',
    'r0': 'r0',
    'r1': 'r1',
    'r2': 'r2', 'r3': 'r3',
    'r0_net': 'r0_net',
    'r1_net': 'r1_net',
    'r2_net': 'r2_net',
    'r3_net': 'r3_net'
}
# ✅  daily_df（日线数据）
daily_rename = {
    'Date': 'date',
    '股票代码': 'stock_code',
    'Open': 'open',
    'Close': 'close',
    'High': 'high',
    'Low': 'low',
    'Volume': 'volume',
    'Amount': 'amount',
    'Amplitude': 'amplitude',
    'ChangePct': 'change_pct',
    'ChangeAmt': 'change_amt',
    'Turnover': 'turnover_rate',
    'Code': 'code'
}


# 主处理循环
for prefix in stock_prefixes:
    try:
        print(f"\n🚀 开始处理：{prefix}")

        # 构建文件路径 - ✅ 确保数据角色正确

        daily_path = f'/content/drive/MyDrive/davidliang/{prefix}_daily.csv'
        minute_path = f'/content/drive/MyDrive/davidliang/{prefix}_合并.csv'
        fundflow_path = f'/content/drive/MyDrive/davidliang/sz{prefix[0:6]}.csv'


        print(f"加载文件: {minute_path}")
        minute_df_raw = pd.read_csv(minute_path)
        print(f"加载文件: {fundflow_path}")
        fundflow_df_raw = pd.read_csv(fundflow_path)
        print(f"加载文件: {daily_path}")
        daily_df_raw = pd.read_csv(daily_path)

        # ✅ 应用重命名
        minute_df = minute_df_raw.rename(columns=minute_rename)
        fundflow_df = fundflow_df_raw.rename(columns=fundflow_rename)
        daily_df = daily_df_raw.rename(columns=daily_rename)

        # ✅ 打印重命名后的前几行和字段 (调试用)
        print("\n✅ 重命名后分钟数据样本:")
        display(minute_df.head())
        print("\n✅ 重命名后资金流数据样本:")
        display(fundflow_df.head())
        print("\n✅ 重命名后日线数据样本:")
        display(daily_df.head())
        print("-" * 30) # 分隔线

        # 字段检查
        print("原始分钟数据字段：", minute_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后分钟数据字段：", minute_df.columns.tolist()) # 打印重命名后列名
        print("原始资金流数据字段：", fundflow_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后资金流数据字段：", fundflow_df.columns.tolist()) # 打印重命名后列名
        print("原始日线数据字段：", daily_df_raw.columns.tolist()) # 打印原始列名
        print("重命名后日线数据字段：", daily_df.columns.tolist()) # 打印重命名后列名


        # 数据对齐与样本构建
        print("\n--- 开始数据对齐 ---")
        # ✅ 将重命名后的 DataFrame 传递给 align_stock_data
        df_aligned = align_stock_data(minute_df, fundflow_df, daily_df)
        print("--- 数据对齐结束 ---")


        # 检查对齐后的数据
        if df_aligned.empty:
            print(f"❌ 数据对齐失败，跳过后续步骤处理 {prefix}")
            continue # 跳过当前股票的处理

        required_rows = Config.window_size + Config.forecast_horizon # 从 Config 获取参数
        if len(df_aligned) < required_rows:
            print(f"❌ 数据不足，无法构建训练样本. 需要至少 {required_rows} 行数据，但只有 {len(df_aligned)} 行. 跳过处理 {prefix}")
            continue # 跳过当前股票的处理


        print("\n✅ 正在构建训练样本...")
        # build_lstm_samples 函数中会根据 df_aligned 的列数确定 input_size
        X, y1, y2, y3 = build_lstm_samples(df_aligned, window_size=Config.window_size, forecast_horizon=Config.forecast_horizon)

        if X.shape[0] == 0:
            print(f"❌ 样本构建失败，跳过后续步骤处理 {prefix}")
            continue # 跳过当前股票的处理

        # 训练模型
        config = Config() # 创建新的 Config 实例
        # input_size 会在 train_model 中根据 X 的形状更新
        config.model_path = f"/content/drive/MyDrive/davidliang/{prefix}_best_model.pt"
        print("✅ 开始训练模型...")
        # 传递 config实例
        train_model(X, y1, y2, y3, config)
        print("✅ 模型训练完成，开始预测...")


        # 预测并输出
        print("\n✅ 正在加载模型并进行预测...")

        if 'df_aligned' in locals() and not df_aligned.empty:
            # 减去 date, datetime, time 列
            config.input_size = df_aligned.drop(columns=['date', 'datetime', 'time']).shape[1]
            print(f"ℹ️ 加载模型前，根据对齐数据更新 config.input_size: {config.input_size}")
        else:
             print("⚠️ 无法在加载模型前根据对齐数据更新 input_size. 使用 Config 默认值或之前设置的值.")
             # 如果无法从数据获取，确保 Config 中的 input_size 匹配 trained model

        model = load_model(config)

        if model is None: # 检查模型是否成功加载
            print(f"❌ 模型加载失败，跳过预测 {prefix}")
            continue


        X_latest = prepare_latest_sample(df_aligned, window_size=Config.window_size) # 从 Config 获取参数

        pred_fundflow, pred_direction, pred_pct = predict_future(model, X_latest, config)

        print_predictions(pred_fundflow, pred_direction, pred_pct)

        # 保存结果
        save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct)


    except Exception as e:
        print(f"❌ 处理 {prefix} 时出错：{e}")
```

```python
# ✅ 加载模型
def load_model(config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    model.load_state_dict(torch.load(config.model_path, map_location=device))
    model.eval()
    return model

# ✅ 构建最新样本（最近50天）
def prepare_latest_sample(df, window_size=50):
    latest = df.tail(window_size).drop(columns=['date'])
    X = latest.values.reshape(1, window_size, -1)
    return torch.tensor(X).float()

# ✅ 预测未来8天
def predict_future(model, X, config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    X = X.to(device)
    with torch.no_grad():
        pred_fundflow, pred_direction, pred_pct = model(X)
        pred_fundflow = pred_fundflow.cpu().numpy().flatten()
        pred_direction = (torch.sigmoid(pred_direction) > 0.5).int().cpu().numpy().flatten()
        pred_pct = pred_pct.cpu().numpy().flatten()
    return pred_fundflow, pred_direction, pred_pct

# ✅ 打印预测结果
def print_predictions(pred_fundflow, pred_direction, pred_pct):
    print("📊 未来8天预测结果：")
    for i in range(8):
        direction = "📈 涨" if pred_direction[i] == 1 else "📉 跌"
        print(f"Day {i+1}: 主力净流入={pred_fundflow[i]:.2f}, 涨跌方向={direction}, 涨跌幅={pred_pct[i]:.2%}")

# ✅ 保存预测结果为 CSV
def save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct):
    df_pred = pd.DataFrame({
        'Day': [f'Day {i+1}' for i in range(8)],
        '主力净流入': pred_fundflow,
        '涨跌方向': ['涨' if d == 1 else '跌' for d in pred_direction],
        '涨跌幅': pred_pct
    })
    df_pred.to_csv(f'/content/drive/MyDrive/davidliang/{prefix}_预测结果.csv', index=False)

# ✅ 批量处理多个个股
stock_prefixes = ['002364_中恒电气']

for prefix in stock_prefixes:
    try:
        print(f"\n🚀 开始处理：{prefix}")

        # 构建文件路径
        minute_path = f'/content/drive/MyDrive/davidliang/sz{prefix[0:6]}.csv'
        fundflow_path = f'/content/drive/MyDrive/davidliang/{prefix}_合并.csv'
        daily_path = f'/content/drive/MyDrive/davidliang/{prefix}_daily.csv'

        # 读取数据
        minute_df = pd.read_csv(minute_path)
        fundflow_df = pd.read_csv(fundflow_path)
        daily_df = pd.read_csv(daily_path)
        # ✅ 字段统一为英文（重命名）
        # ✅ 分钟数据（002364_中恒电气_合并.csv）
        minute_rename = {
            '日期': 'date',
            '时间': 'time',
            '主力净流入': 'main_inflow',
            '超大单净流入': 'super_inflow',
            '大单净流入': 'large_inflow',
            '总成交额': 'total_amount'
        }
        # ✅ 资金流数据（sz002364.csv）
        fundflow_rename = {
            'opendate': 'date',
            'trade': 'time',
            'changeratio': 'change_ratio',
            'turnover': 'turnover',
            'netamount': 'net_amount',
            'ratioamount': 'ratio_amount',
            'r0': 'r0',
            'r1': 'r1',
            'r2': 'r2',
            'r3': 'r3',
            'r0_net': 'r0_net',
            'r1_net': 'r1_net',
            'r2_net': 'r2_net',
            'r3_net': 'r3_net'
        }
        # ✅  daily_df（日线数据）
        daily_rename = {
            'Date': 'date',
            '股票代码': 'stock_code',
            'Open': 'open',
            'Close': 'close',
            'High': 'high',
            'Low': 'low',
            'Volume': 'volume',
            'Amount': 'amount',
            'Amplitude': 'amplitude',
            'ChangePct': 'change_pct',
            'ChangeAmt': 'change_amt',
            'Turnover': 'turnover_rate',
            'Code': 'code'
        }

        # ✅ 应用重命名
        minute_df.rename(columns=minute_rename, inplace=True)  # minute_df 是分钟数据
        fundflow_df.rename(columns=fundflow_rename, inplace=True)  # fundflow_df 是资金流数据
        daily_df.rename(columns=daily_rename, inplace=True)


        # 字段检查
        print("分钟数据字段：", minute_df.columns.tolist())
        print("资金流数据字段：", fundflow_df.columns.tolist())
        print("日线数据字段：", daily_df.columns.tolist())

        # 数据对齐与样本构建
        df_aligned = align_stock_data(fundflow_df, minute_df, daily_df)

        required_rows = 50 + 8 # window_size + forecast_horizon
        if len(df_aligned) < required_rows:
            raise ValueError(f"数据不足，无法构建训练样本. 需要至少 {required_rows} 行数据，但只有 {len(df_aligned)} 行.")
        X, y1, y2, y3 = build_lstm_samples(df_aligned, window_size=50, forecast_horizon=8)

        # 训练模型
        config = Config()
        config.model_path = f"/content/drive/MyDrive/davidliang/{prefix}_best_model.pt"
        train_model(X, y1, y2, y3, config)
        print("✅ 正在构建训练样本...")
        print("✅ 开始训练模型...")
        print("✅ 模型训练完成，开始预测...")
        print("✅ 预测完成，保存结果中...")


        # 预测并输出
        model = load_model(config)
        X_latest = prepare_latest_sample(df_aligned, window_size=50)
        pred_fundflow, pred_direction, pred_pct = predict_future(model, X_latest, config)
        print_predictions(pred_fundflow, pred_direction, pred_pct)

        # 保存结果
        save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct)
        print(f"✅ {prefix} 预测完成，结果已保存为 CSV")

    except Exception as e:
        print(f"❌ 处理 {prefix} 时出错：{e}")
```
