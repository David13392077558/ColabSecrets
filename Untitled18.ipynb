```python
import akshare as ak
import pandas as pd
import numpy as np
import os
from google.colab import drive
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import zipfile
import matplotlib.font_manager as fm

# 配置matplotlib支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# 股票列表
stock_codes = ['002456','601138','601288','002269','002364']
print(f"🔥 已加载股票代码列表，共 {len(stock_codes)} 支股票")

# 时间范围
start_date = "20210924"
end_date = datetime.today().strftime("%Y%m%d")

# 保存路径
save_path = "./data/"
os.makedirs(save_path, exist_ok=True)

# 字段标准化
def standardize_columns(df):
    column_mapping = {
        "日期": "Date", "股票代码": "Code", "开盘": "Open", "收盘": "Close",
        "最高": "High", "最低": "Low", "成交量": "Volume", "成交额": "Amount",
        "振幅": "Amplitude", "涨跌幅": "ChangePct", "涨跌额": "ChangeAmt", "换手率": "Turnover"
    }
    df.rename(columns=column_mapping, inplace=True)
    return df

# RSI计算
def compute_rsi(series, window=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# 数据加载与特征工程
def load_and_prepare_data(file_path, fundflow_path=None):
    df = pd.read_csv(file_path)
    df["Date"] = pd.to_datetime(df["Date"])
    df.sort_values("Date", inplace=True)
    df.set_index("Date", inplace=True)
    print("原始数据前5行:")
    display(df.head())

    # 合并资金流数据
    if fundflow_path and os.path.exists(fundflow_path):
        df_fund = pd.read_csv(fundflow_path)
        # Map 'opendate' to 'Date' and standardize other fund flow columns
        fundflow_column_mapping = {
            "opendate": "Date",
            "netamount": "主力净流入",
            "ratioamount": "净流入率",
            # "trade": "资金流收盘价" # Mapping 'trade' to '资金流收盘价' based on user feedback
        }
        df_fund.rename(columns=fundflow_column_mapping, inplace=True)

        # Drop the redundant 'turnover' column if it exists in the fund flow data
        if "turnover" in df_fund.columns:
            df_fund.drop(columns=["turnover"], inplace=True)
            print("已删除资金流数据中的 'turnover' 列")

        df_fund["Date"] = pd.to_datetime(df_fund["Date"])
        df_fund.sort_values("Date", inplace=True) # Sort fund flow data by date
        df_fund.set_index("Date", inplace=True)
        df = df.join(df_fund, how="left")
        print("合并资金流数据后前5行:")
        display(df.head())


    # 衍生特征
    df["MA5"] = df["Close"].rolling(window=5).mean()
    df["MA10"] = df["Close"].rolling(window=10).mean()
    df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
    df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = df["EMA12"] - df["EMA26"]
    df["RSI"] = compute_rsi(df["Close"], window=14)
    df["Return_1d"] = df["Close"].pct_change()
    df["Volatility_5d"] = df["Close"].rolling(window=5).std()
    df["Weekday"] = df.index.weekday
    df["IsMonthEnd"] = df.index.is_month_end.astype(int)
    df["MoneyFlow"] = df["Amount"] / df["Volume"]

    # 资金流字段（根据你的文件结构调整）
    # Updated fundflow_features based on user feedback
    fundflow_features = ["主力净流入", "净流入率"]
    for col in fundflow_features:
        if col in df.columns:
            df[col] = df[col].fillna(0)
        else:
            df[col] = 0


    df.dropna(inplace=True)

    features = [
        "Open", "High", "Low", "Close", "Volume", "Amount", "Turnover",
        "MA5", "MA10", "EMA12", "EMA26", "MACD", "RSI",
        "Return_1d", "Volatility_5d", "Weekday", "IsMonthEnd", "MoneyFlow"
    ] + fundflow_features


    scaler = MinMaxScaler()
    df_scaled = scaler.fit_transform(df[features])

    return df_scaled, df["Close"].values, scaler

# 构造时间序列样本
def create_sequences(data_scaled, close_prices, window_size=60, pred_size=15):
    X, y = [], []
    for i in range(len(data_scaled) - window_size - pred_size + 1):
        X.append(data_scaled[i:i+window_size])
        y.append(close_prices[i+window_size:i+window_size+pred_size])
    return np.array(X), np.array(y)

# 构建LSTM模型
def build_lstm_model(input_shape, output_size=15):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(32))
    model.add(Dense(output_size))
    model.compile(optimizer='adam', loss='mse')
    return model

# 模型训练
def train_lstm_model(X, y, input_shape, model_path):
    model = build_lstm_model(input_shape)
    checkpoint = ModelCheckpoint(
        model_path.replace(".h5", ".keras"),
        monitor='loss',
        save_best_only=True,
        verbose=1
    )
    history = model.fit(X, y, epochs=100, batch_size=32, callbacks=[checkpoint], verbose=0)
    for epoch, loss in enumerate(history.history['loss']):
        print(f"Epoch {epoch+1}: loss = {loss:.4f}")
    return model

# 预测并绘图
def predict_and_plot(model, scaled_data, scaler, window_size, stock_code, output_dir="./charts"):
    last_window = scaled_data[-window_size:]
    prediction = model.predict(np.expand_dims(last_window, axis=0))[0]

    reference = scaled_data[-1]
    dummy = np.tile(reference, (15, 1))
    full_dummy = np.zeros((15, dummy.shape[1]))
    full_dummy[:, :-1] = dummy[:, :-1]
    full_dummy[:, -1] = prediction

    recovered = scaler.inverse_transform(full_dummy)[:, -1]

    os.makedirs(output_dir, exist_ok=True)
    chart_path = os.path.join(output_dir, f"{stock_code}_{datetime.today().strftime('%Y-%m-%d')}_forecast.png")

    plt.figure(figsize=(10, 4))
    plt.plot(range(1, 16), recovered, marker='o', label="预测收盘价")
    plt.title(f"{stock_code} - 未来15天收盘价预测")
    plt.xlabel("未来天数")
    plt.ylabel("收盘价")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path)
    plt.close()

    print(f"✅ 图表已保存：{chart_path}")
    return recovered

# 打包图表
def zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip"):
    today_str = datetime.today().strftime("%Y-%m-%d")
    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for file in os.listdir(chart_dir):
            if file.endswith(".png") and today_str in file:
                zipf.write(os.path.join(chart_dir, file), arcname=file)
    print(f"📦 今日图表已打包为：{zip_name}")

# 主流程
scaled_stock_data = {}
for code in stock_codes:
    symbol = code[-6:]
    print(f"📈 正在处理：{code} → {symbol}")
    try:
        df = ak.stock_zh_a_hist(symbol=symbol, period="daily", start_date=start_date, end_date=end_date, adjust="qfq")
        if df is not None and not df.empty:
            df["股票代码"] = code
            df = standardize_columns(df)
            print(f"✅ {code} 获取成功，共 {len(df)} 条记录")
            df.to_csv(f"{save_path}{code}.csv", index=False)
        else:
            print(f"⚠️ {code} 无数据，跳过保存")
    except Exception as e:
        print(f"❌ {code} 抓取失败：{e}")

# 加载并处理数据
for code in stock_codes:
    file_path = f"{save_path}{code}.csv"
    prefix = code[-6:]

    # 判断市场类型
    market_prefix = "sz" if code.startswith("0") else "sh"
    fundflow_path = f"/content/drive/MyDrive/Colab Notebooks/jiqi_do/jiqi_do/{market_prefix}{prefix}.csv"

    if os.path.exists(file_path):
        try:
            scaled_data, close_prices, scaler = load_and_prepare_data(file_path, fundflow_path)
            df_scaled = pd.DataFrame(scaled_data)
            df_scaled["Close"] = close_prices
            scaled_stock_data[code] = (df_scaled, scaler)
        except Exception as e:
            print(f"❌ 数据处理失败：{code} → {e}")
    else:
        print(f"⚠️ 缺失数据文件：{file_path}")


# 模型训练与预测
os.makedirs("./models", exist_ok=True)
chart_dir = "./charts"

for code in stock_codes:
    if code not in scaled_stock_data:
        continue

    df_scaled, scaler = scaled_stock_data[code]
    scaled_data = df_scaled.drop(columns=["Close"]).values
    close_prices = df_scaled["Close"].values

    # 🔁 构造训练样本并训练模型
    X, y = create_sequences(scaled_data, close_prices, window_size=60, pred_size=15)
    if X.shape[0] == 0:
        print(f"⚠️ 样本数量不足，跳过训练：{code}")
        continue

    model_path = f"./models/{code}_lstm.h5"
    model = train_lstm_model(X, y, input_shape=X.shape[1:], model_path=model_path)
    predict_and_plot(model, scaled_data, scaler, window_size=60, stock_code=code, output_dir=chart_dir)
# 📦 打包所有图表
zip_today_charts(chart_dir)
```
