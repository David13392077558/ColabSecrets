```python
import openpyxl

wb = openpyxl.load_workbook('工作表 在 致胜科技报价单5.14.xlsx')
ws = wb.active

html_rows = []
for row in ws.iter_rows():
    html_cells = []
    for cell in row:
        value = cell.value if cell.value else ""
        html_cells.append(f'<td style="border:1px solid #ccc;padding:6px;">{value}</td>')
    html_rows.append(f"<tr>{''.join(html_cells)}</tr>")

html_table = f"<table style='border-collapse:collapse;'>{''.join(html_rows)}</table>"
html_content = f"<html><body>{html_table}</body></html>"

with open('excel_preview.html', 'w', encoding='utf-8') as f:
    f.write(html_content)
```

```python
!pip install pyppeteer nest_asyncio

import asyncio
import nest_asyncio
from pyppeteer import launch
import os

nest_asyncio.apply()

async def html_to_image():
    browser = await launch(args=['--no-sandbox', '--disable-setuid-sandbox'])
    page = await browser.newPage()
    await page.goto('file://' + os.path.abspath('excel_preview.html'))
    await page.screenshot({'path': 'excel_image.png', 'fullPage': True})
    await browser.close()

loop = asyncio.get_event_loop()
loop.run_until_complete(html_to_image())
```

```python
import requests

url = 'http://127.0.0.1:5000/convert_excel'
files = {'file': open('your_excel_file.xlsx', 'rb')}

response = requests.post(url, files=files)

if response.status_code == 200:
    with open('converted_output.pdf', 'wb') as f:
        f.write(response.content)
    print("Conversion successful. PDF saved as converted_output.pdf")
else:
    print(f"Conversion failed with status code {response.status_code}: {response.text}")
```
