```python
# ✅ 导入库
import akshare as ak
import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import zipfile
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, MultiHeadAttention, LayerNormalization, Dense, Dropout, TimeDistributed
from tensorflow.keras.callbacks import ModelCheckpoint

# ✅ 股票列表
stock_codes = ['002456','601138','601288','002269','002364']
start_date = "20210924"
end_date = datetime.today().strftime("%Y%m%d")
save_path = "./data/"
os.makedirs(save_path, exist_ok=True)

# Add checks for directory and files
print(f"Checking if directory exists: {save_path}")
if os.path.exists(save_path):
    print("✅ Directory exists.")
    print(f"Checking for stock data files in {save_path}:")
    found_files = os.listdir(save_path)
    print(f"Files found: {found_files}")
    missing_files = [f"{code}.csv" for code in stock_codes if f"{code}.csv" not in found_files]
    if missing_files:
        print(f"❌ Missing stock data files: {missing_files}")
    else:
        print("✅ All expected stock data files found.")
else:
    print("❌ Directory does not exist.")

# ✅ 标准化字段
def standardize_columns(df):
    column_mapping = {
        "日期": "Date", "股票代码": "Code", "开盘": "Open", "收盘": "Close",
        "最高": "High", "最低": "Low", "成交量": "Volume", "成交额": "Amount",
        "振幅": "Amplitude", "涨跌幅": "ChangePct", "涨跌额": "ChangeAmt", "换手率": "Turnover"
    }
    df.rename(columns=column_mapping, inplace=True)
    return df

# ✅ RSI计算
def compute_rsi(series, window=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# ✅ 数据处理
def load_and_prepare_data(file_path, fundflow_path=None):
    try:
        print(f"Attempting to load data from: {file_path}") # Debug print

        # Try reading with different parameters if the default fails
        try:
            df = pd.read_csv(file_path)
            print(f"✅ Successfully read CSV with default settings. Columns: {df.columns.tolist()}") # Debug print
        except Exception as e:
            print(f"❌ Failed to read CSV with default settings: {e}")
            # Attempt with a different encoding
            try:
                df = pd.read_csv(file_path, encoding='gbk')
                print(f"✅ Successfully read CSV with GBK encoding. Columns: {df.columns.tolist()}") # Debug print
            except Exception as e:
                print(f"❌ Failed to read CSV with GBK encoding: {e}")
                # Attempt with another common encoding
                try:
                    df = pd.read_csv(file_path, encoding='utf-8')
                    print(f"✅ Successfully read CSV with UTF-8 encoding. Columns: {df.columns.tolist()}") # Debug print
                except Exception as e:
                    print(f"❌ Failed to read CSV with UTF-8 encoding: {e}")
                    return None, None, None # Return None if unable to read the file


        # Standardize columns immediately after reading
        df = standardize_columns(df)

        if "Date" not in df.columns:
            print(f"❌ 'Date' column not found in {file_path} after standardization.")
            return None, None, None

        df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
        df.dropna(subset=['Date'], inplace=True)
        df.sort_values("Date", inplace=True)
        df.set_index("Date", inplace=True)

        # 合并资金流
        if fundflow_path and os.path.exists(fundflow_path):
            print(f"Attempting to load fund flow data from: {fundflow_path}") # Debug print
            try:
                df_fund = pd.read_csv(fundflow_path)
            except Exception as e:
                 print(f"❌ Failed to read fund flow CSV with default settings: {e}")
                 try:
                     df_fund = pd.read_csv(fundflow_path, encoding='gbk')
                     print("✅ Successfully read fund flow CSV with GBK encoding.")
                 except Exception as e:
                     print(f"❌ Failed to read fund flow CSV with GBK encoding: {e}")
                     try:
                         df_fund = pd.read_csv(fundflow_path, encoding='utf-8')
                         print("✅ Successfully read fund flow CSV with UTF-8 encoding.")
                     except Exception as e:
                         print(f"❌ Failed to read fund flow CSV with UTF-8 encoding: {e}")
                         df_fund = pd.DataFrame() # Continue without fund flow if cannot read


            # Map 'pendate' to 'Date' and standardize other fund flow columns
            fundflow_column_mapping = {
                "opendate": "Date",
                "netamount": "主力净流入",
                "ratioamount": "主力净流入率",
                #  "trade": "大单净流入" # Assuming 'trade' might be related to '大单净流入' based on the user's data sample, adjust if needed
            }
            df_fund.rename(columns=fundflow_column_mapping, inplace=True)

            if "Date" in df_fund.columns: # Check if 'Date' column exists after renaming
                df_fund["Date"] = pd.to_datetime(df_fund["Date"], errors='coerce')
                df_fund.dropna(subset=['Date'], inplace=True)
                df_fund.sort_values("Date", inplace=True) # Ensure fund flow data is also sorted by date
                df_fund.set_index("Date", inplace=True)

                # Select specific columns to join, ensuring they exist in df_fund
                cols_to_join = [col for col in ["主力净流入", "超大单净流入", "大单净流入"] if col in df_fund.columns]
                if cols_to_join:
                     df = df.join(df_fund[cols_to_join], how="left")
                else:
                    print("⚠️ No fund flow columns to join after reading and renaming.")

            else:
                print("❌ 'Date' column not found in fund flow data after renaming.")


        # Ensure '涨跌幅' column exists, if not, create it or handle its absence
        if "涨跌幅" not in df.columns and "ChangePct" in df.columns:
             df["涨跌幅"] = df["ChangePct"]
        elif "涨跌幅" not in df.columns:
            # Option 1: Create the column with default value (e.g., 0 or NaN)
            df["涨跌幅"] = 0 # Or np.nan, depending on how you want to handle missing data
            print(f"⚠️ Column '涨跌幅' not found in {file_path}. Created with default value 0.")
            # Option 2: Or, if it's not critical, you might remove it from targets later

        # Derivative Features
        df["MA5"] = df["Close"].rolling(window=5).mean()
        df["MA10"] = df["Close"].rolling(window=10).mean()
        df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
        df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
        df["MACD"] = df["EMA12"] - df["EMA26"]
        df["RSI"] = compute_rsi(df["Close"], window=14)
        df["Return_1d"] = df["Close"].pct_change()
        df["Volatility_5d"] = df["Close"].rolling(window=5).std()
        df["Weekday"] = df.index.weekday
        df["IsMonthEnd"] = df.index.is_month_end.astype(int)
        df["MoneyFlow"] = df["Amount"] / df["Volume"]

        # Fund flow fields
        fundflow_features = ["主力净流入", "超大单净流入", "大单净流入"]
        for col in fundflow_features:
            df[col] = df[col].fillna(0) if col in df.columns else 0
            if col not in df.columns:
                 df[col] = 0 # Ensure fund flow columns exist, even if not in the fundflow file


        df.dropna(inplace=True)

        features = [
            "Open", "High", "Low", "Close", "Volume", "Amount", "Turnover",
            "MA5", "MA10", "EMA12", "EMA26", "MACD", "RSI",
            "Return_1d", "Volatility_5d", "Weekday", "IsMonthEnd", "MoneyFlow"
        ] + [col for col in fundflow_features if col in df.columns] # Only include fund flow features if they exist

        targets = ["Close", "主力净流入", "涨跌幅", "超大单净流入"]
        # Filter targets to only include columns that exist in the dataframe
        targets = [target for target in targets if target in df.columns]
        if "涨跌幅" not in targets and "ChangePct" in df.columns:
             targets.append("ChangePct") # Use ChangePct if 涨跌幅 is not available and ChangePct exists

        # If after all checks, the necessary targets are not present, raise an error or handle
        if not all(target in df.columns for target in ["Close", "主力净流入", "超大单净流入"]) and not ("涨跌幅" in df.columns or "ChangePct" in df.columns):
             # This condition might be too strict, depending on which targets are truly essential
             # For now, let's allow processing as long as 'Close' is present
             if "Close" not in df.columns:
                  raise ValueError("Required target column 'Close' is missing from the data.")
             else:
                 print("⚠️ Some target columns are missing, proceeding with available targets.")
                 # Re-filter targets based on available columns before scaling
                 targets = [target for target in ["Close", "主力净流入", "涨跌幅", "超大单净流入"] if target in df.columns]


        scaler = MinMaxScaler()
        # Ensure only existing columns are selected for scaling
        cols_to_scale = [col for col in features + targets if col in df.columns and df[col].dtype in ['int64', 'float64']] # Ensure only numeric columns are selected for scaling
        df_scaled = scaler.fit_transform(df[cols_to_scale])

        # Need to return the actual target columns data for the scaler
        # Ensure the order of targets matches the order in cols_to_scale for correct inverse transformation later
        target_columns_in_scaled_order = [col for col in cols_to_scale if col in targets]
        target_columns_data = df[target_columns_in_scaled_order].values


        return df_scaled, target_columns_data, scaler

    except Exception as e:
        print(f"❌ 数据处理失败：{file_path} → {e}")
        return None, None, None


# ✅ 构造样本
def create_sequences(data_scaled, target_series, window_size=60, pred_size=15):
    X, y = [], []
    # Ensure enough data exists for the window and prediction size
    if len(data_scaled) < window_size + pred_size:
        print(f"⚠️ Not enough data for sequence creation. Need at least {window_size + pred_size} data points, but only have {len(data_scaled)}.")
        return np.array(X), np.array(y)

    for i in range(len(data_scaled) - window_size - pred_size + 1):
        X.append(data_scaled[i:i+window_size])
        y.append(target_series[i+window_size:i+window_size+pred_size])
    return np.array(X), np.array(y)

# ✅ Attention模型
def build_transformer_model(input_shape, output_size=4, pred_size=15):
    inputs = Input(shape=input_shape)
    attn = MultiHeadAttention(num_heads=4, key_dim=32)(inputs, inputs)
    attn = LayerNormalization()(attn)
    x = Dense(64, activation='relu')(attn)
    x = Dropout(0.2)(x)
    x = LayerNormalization()(x)
    x = TimeDistributed(Dense(output_size))(x)
    x = x[:, -pred_size:, :]
    model = Model(inputs, x)
    model.compile(optimizer='adam', loss='mse')
    return model

# ✅ 训练模型
def train_transformer_model(X, y, input_shape, model_path):
    model = build_transformer_model(input_shape, output_size=y.shape[-1]) # Adjust output size based on actual targets
    checkpoint = ModelCheckpoint(model_path.replace(".h5", ".keras"), monitor='loss', save_best_only=True, verbose=1)
    history = model.fit(X, y, epochs=200, batch_size=32, callbacks=[checkpoint], verbose=0)
    for epoch, loss in enumerate(history.history['loss']):
        print(f"Epoch {epoch+1}: loss = {loss:.4f}")
    return model

# ✅ 预测并绘图
def predict_and_plot_targets(model, scaled_data, scaler, window_size, stock_code, output_dir="./charts"):
    # Ensure enough data exists for the last window
    if len(scaled_data) < window_size:
        print(f"⚠️ Not enough data to create the last window for prediction. Need {window_size} data points, but only have {len(scaled_data)}.")
        return

    last_window = scaled_data[-window_size:]
    prediction = model.predict(np.expand_dims(last_window, axis=0))[0]
    # Need to correctly inverse transform based on the columns used for scaling
    # This part needs careful consideration based on which columns were scaled and which are targets
    # Assuming the last 'output_size' columns of the scaled data correspond to the targets
    num_targets = model.output_shape[-1]
    reference_for_inverse_transform = np.zeros((15, scaled_data.shape[-1]))
    # Place the prediction in the columns corresponding to the targets in the original scaled data
    reference_for_inverse_transform[:, -num_targets:] = prediction

    # To correctly inverse transform, we need a dummy array with the same structure as the one used for fitting the scaler
    # We can't easily recreate the exact structure without knowing which columns were used for features vs targets during scaling
    # A safer approach is to scale features and targets separately or ensure the order is consistent

    # For now, let's assume the last 'num_targets' columns of the original dataframe were the targets
    # We need to create a dummy array with the correct number of columns that were scaled
    dummy_inverse_transform_data = np.zeros((15, scaled_data.shape[-1]))
    dummy_inverse_transform_data[:, -num_targets:] = prediction # Place predictions in the target columns

    recovered_full = scaler.inverse_transform(dummy_inverse_transform_data)
    recovered_targets = recovered_full[:, -num_targets:] # Extract the inverse transformed targets

    # Ensure target_names matches the actual targets being predicted
    # This requires knowing which targets were successfully loaded and used
    # Based on the load_and_prepare_data function, targets are ["Close", "主力净流入", "涨跌幅", "超大单净流入"] or a subset
    # Need to pass the actual list of targets used for scaling/prediction
    # For now, let's use a placeholder or the original list and hope it aligns
    target_names = ["Close", "主力净流入", "涨跌幅", "超大单净流入"] # This might need adjustment

    os.makedirs(output_dir, exist_ok=True)
    today = datetime.today().strftime('%Y-%m-%d')

    # Generate CSV for 主力净流入 and 涨跌幅
    csv_data = {}
    csv_target_names = ["主力净流入", "涨跌幅"]
    for i, name in enumerate(target_names[:num_targets]):
        if name in csv_target_names:
            csv_data[f"预测{name}"] = recovered_targets[:, i]
        elif name == "Close":
            # Keep plotting for Close price
            plt.figure(figsize=(10, 4))
            plt.plot(range(1, 16), recovered_targets[:, i], marker='o', label=f"预测{name}")
            plt.title(f"{stock_code} - 未来15天{name}预测")
            plt.xlabel("未来天数")
            plt.ylabel(name)
            plt.grid(True)
            plt.legend()
            plt.tight_layout()
            chart_path = os.path.join(output_dir, f"{stock_code}_{today}_{name}.png")
            plt.savefig(chart_path)
            plt.close()
            print(f"✅ 图表已保存：{chart_path}")

    if csv_data:
        csv_df = pd.DataFrame(csv_data)
        csv_path = os.path.join(output_dir, f"{stock_code}_{today}_预测明细.csv")
        csv_df.to_csv(csv_path, index=False)
        print(f"✅ 预测明细已保存：{csv_path}")


# ✅ 打包图表
def zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip"):
    today_str = datetime.today().strftime("%Y-%m-%d")
    # Check if directory exists before listing files
    if not os.path.exists(chart_dir):
        print(f"⚠️ Chart directory does not exist: {chart_dir}")
        return

    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for file in os.listdir(chart_dir):
            if file.endswith(".png") and today_str in file:
                zipf.write(os.path.join(chart_dir, file), arcname=file)
    print(f"📦 今日图表已打包为：{zip_name}")

# ✅ 主流程：加载、处理、训练并预测所有股票
os.makedirs("./models", exist_ok=True)
chart_dir = "./charts"
scaled_stock_data = {} # Initialize scaled_stock_data here
fundflow_data_path = "/content/drive/MyDrive/Colab Notebooks/jiqi_do/jiqi_do/" # Base path for fund flow data

for code in stock_codes:
    file_path = os.path.join(save_path, f"{code}.csv")

    # Determine market prefix for fund flow data
    market_prefix = "sz" if code.startswith("0") else "sh"
    fundflow_file_name = f"{market_prefix}{code}.csv" # Assuming the fund flow file name is market_prefix + code + .csv
    fundflow_path = os.path.join(fundflow_data_path, fundflow_file_name)


    if not os.path.exists(file_path):
        print(f"⚠️ 股票数据文件不存在：{file_path}")
        continue

    scaled_data, target_series, scaler = load_and_prepare_data(file_path, fundflow_path)

    if scaled_data is not None:
        scaled_stock_data[code] = (scaled_data, target_series, scaler)
        print(f"✅ 成功加载和处理数据：{code}")
    else:
        print(f"❌ 数据加载或处理失败：{code}")

print(f"\nScaled stock data loaded for: {list(scaled_stock_data.keys())}\n")


for code in stock_codes:
    if code not in scaled_stock_data:
        print(f"⚠️ 数据未加载或处理，跳过训练：{code}")
        continue

    scaled_data, target_series, scaler = scaled_stock_data[code]
    X, y = create_sequences(scaled_data, target_series, window_size=60, pred_size=15)

    print(f"➡️ Processing {code}: X shape = {X.shape}, y shape = {y.shape}")

    if X.shape[0] == 0:
        print(f"⚠️ 样本数量不足，跳过训练：{code}")
        continue

    model_path = f"./models/{code}_attention.h5"
    print(f"✨ 开始训练模型：{code}")
    # Pass the actual number of targets to the model building function
    model = train_transformer_model(X, y, input_shape=X.shape[1:], model_path=model_path)
    print(f"✅ 模型训练完成：{code}")

    print(f"📈 开始预测和绘图：{code}")
    # Pass the actual scaler used for this stock's data
    predict_and_plot_targets(model, scaled_data, scaler, window_size=60, stock_code=code, output_dir=chart_dir)
    print(f"✅ 预测和绘图完成：{code}\n")

# ✅ 打包图表
zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip")
```

```python
import akshare as ak
import os
from datetime import datetime

stock_codes = ['002456','601138','601288','002269','002364']
start_date = "20210924"
end_date = datetime.today().strftime("%Y%m%d")
save_path = "./data/"

os.makedirs(save_path, exist_ok=True)

print(f"开始下载股票数据到 {save_path}...")

for code in stock_codes:
    file_path = os.path.join(save_path, f"{code}.csv")
    print(f"正在下载 {code} 数据...")
    try:
        # Using stock_zh_a_hist API for historical data
        stock_df = ak.stock_zh_a_hist(symbol=code, period="daily", start_date=start_date, end_date=end_date, adjust="qfq")
        if not stock_df.empty:
            stock_df.to_csv(file_path, index=False)
            print(f"✅ {code} 数据下载并保存成功: {file_path}")
        else:
            print(f"❌ {code} 数据下载失败或为空")
    except Exception as e:
        print(f"❌ 下载 {code} 数据时发生错误: {e}")

print("股票数据下载完成。")
```
