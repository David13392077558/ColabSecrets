```python
# 从头开始，包含所有必要的导入、函数定义和主逻辑

# ✅ 导入库
from google.colab import drive
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
import datetime as dt # Import the datetime module and alias it as dt
from torch.utils.data import DataLoader, TensorDataset # Import DataLoader and TensorDataset
from sklearn.preprocessing import StandardScaler # Import StandardScaler

# ✅ 挂载 Google Drive
# 请确保您已经挂载了 Google Drive，如果之前已经运行成功，可以注释掉下一行
# drive.mount('/content/drive')

# ✅ 模型结构：双层 LSTM + Dropout + 多目标输出
class MultiTargetLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, forecast_horizon, dropout=0.3):
        super().__init__()
        # 确保 dropout 参数在 LSTM 层只应用于非最后一层
        lstm_dropout = dropout if num_layers > 1 else 0
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=lstm_dropout)
        self.dropout = nn.Dropout(dropout) # 在 LSTM 输出后应用 dropout
        self.fc_fundflow = nn.Linear(hidden_size, forecast_horizon)
        self.fc_direction = nn.Linear(hidden_size, forecast_horizon)
        self.fc_pct = nn.Linear(hidden_size, forecast_horizon)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.fc_fundflow(h), self.fc_direction(h), self.fc_pct(h)


# ✅ 配置参数
class Config:
    input_size = 20 # 这个会在 train_model 中根据数据列数动态更新
    hidden_size = 64
    num_layers = 2
    forecast_horizon = 8
    dropout = 0.3
    lr = 1e-4 # Reduced learning rate
    epochs = 200
    window_size = 30
    model_path = "best_lstm_model.pt" # 会根据股票前缀更新
    batch_size = 512 # Add batch size to config

# ✅ 数值异常清洗函数
def clean_numeric_columns(df):
    # Identify numeric columns
    numeric_cols = df.select_dtypes(include=np.number).columns

    # Replace Inf with NaN
    df[numeric_cols] = df[numeric_cols].replace([np.inf, -np.inf], np.nan)

    # Fill NaN with mean
    # Use a loop or apply directly to the numeric columns
    for col in numeric_cols:
        if df[col].isnull().any(): # Check if there are any NaN values before filling
            mean_val = df[col].mean()
            df[col].fillna(mean_val, inplace=True)

    return df

# ✅ 对数压缩资金流数据函数
def compress_fundflow(df, cols=['main_inflow', 'super_inflow', 'large_inflow', 'total_amount', 'net_amount', 'ratio_amount', 'r0', 'r1', 'r2', 'r3', 'r0_net', 'r1_net', 'r2_net', 'r3_net']):
    for col in cols:
        if col in df.columns:
            # Apply log1p to the absolute value and then multiply by the sign
            df[col] = np.log1p(np.abs(df[col])) * np.sign(df[col])
    return df


# ✅ 对齐股票数据函数
def align_stock_data(minute_df, fundflow_df, daily_df):
    print(f"Minute df initial shape: {minute_df.shape}")
    print(f"Fundflow df initial shape: {fundflow_df.shape}")
    print(f"Daily df initial shape: {daily_df.shape}")

    # ✅ 清洗 minute_df（分钟数据：主力净流入等）
    minute_df = minute_df.dropna(subset=['date', 'time']).copy() # 添加 .copy()

    # ✅ 打印原始日期和时间样本 (调试用)
    print("✅ 原始分钟数据样本（日期和时间）：")
    print(minute_df[['date', 'time']].head())
    print("-" * 30) # 分隔线方便查看

    # ✅ 更稳健的日期和时间分别解析和组合
    try:

        # 根据之前打印的原始样本，日期是 YYYY-MM-DD 格式
        minute_df['date_parsed'] = pd.to_datetime(minute_df['date'], errors='coerce', format='%Y-%m-%d')
        print("✅ 日期列初步解析成功")


        # 需要先转为字符串，并确保没有异常值或非时间格式导致 to_timedelta failed
        minute_df['time_str'] = minute_df['time'].astype(str).str.strip()

        # 将时间字符串转换为 timedelta

        # 然后提取其 time 部分
        minute_df['time_parsed_dt'] = pd.to_datetime(minute_df['time_str'], format='%H:%M', errors='coerce').dt.time
        print("✅ 时间列初步解析成功")


        # 组合日期 (datetime64[ns]) 和 时间 (datetime.time 对象)


        # 使用 apply + combine 组合，对每一行操作
        # 需要先处理解析失败的 NaT/NaN
        valid_indices = minute_df['date_parsed'].notna() & minute_df['time_parsed_dt'].notna()
        minute_df_valid = minute_df.loc[valid_indices].copy() # 只处理有效行

        if not minute_df_valid.empty:
             # 使用 apply 和 pd.Timestamp.combine
             minute_df.loc[valid_indices, 'datetime'] = minute_df_valid.apply(
                 lambda row: pd.Timestamp.combine(row['date_parsed'].date(), row['time_parsed_dt']),
                 axis=1
             )
             print(f"✅ 成功组合 {len(minute_df_valid)} 行的日期和时间")
        else:
            print("⚠️ 没有成功解析的日期或时间对，无法组合 datetime.")
            minute_df['datetime'] = pd.NaT # 如果没有有效行，datetime 列全为 NaT


    except Exception as e:
        print(f"❌ 日期或时间解析组合失败：{e}")

        return pd.DataFrame() # 解析失败，返回空 DataFrame

    # 丢弃解析失败的行 (datetime 为 NaT 的行)
    minute_df = minute_df.dropna(subset=['datetime']).copy() # 添加 .copy()

    # 提取标准日期字段用于 merge (从解析成功的 datetime 中提取)
    minute_df['date'] = minute_df['datetime'].dt.date

    # ✅ 验证清洗和解析结果
    print("✅ 有效分钟数据样本：")
    if not minute_df.empty:
        print(minute_df[['datetime', 'main_inflow']].head())
        print("✅ 有效日期范围：", minute_df['date'].min(), "→", minute_df['date'].max())
    else:
        print("⚠️ 清洗和日期时间转换后，分钟数据 DataFrame 为空.")

    print(f"Minute df after cleaning and datetime conversion: {minute_df.shape}")
    if not minute_df.empty:
        print(f"Minute df dates after cleaning: {minute_df['date'].min()} → {minute_df['date'].max()}")
    else:
         print(f"Minute df dates after cleaning: 无效日期范围")

    # 清理不再需要的中间列
    minute_df = minute_df.drop(columns=['date_parsed', 'time_str', 'time_parsed_dt'])

    # ✅ 清洗 fundflow_df（资金流数据：r0~r3）
    if 'date' not in fundflow_df.columns:
        print("❌ fundflow_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    fundflow_df = fundflow_df.dropna(subset=['date']).copy() # 添加 .copy()

    try:
        fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], format='%Y-%m-%d', errors='coerce').dt.date
        print("✅ Fundflow df 日期用 %Y-%m-%d 格式解析成功")
    except Exception as e:
         print(f"⚠️ Fundflow df 日期用 %Y-%m-%d 格式解析失败: {e}. 尝试推断格式.")
         fundflow_df['date'] = pd.to_datetime(fundflow_df['date'], errors='coerce').dt.date # 依赖推断
         print("✅ Fundflow df 日期用推断格式解析完成 (可能包含 NaT)")
    # 1. 去除无效日期
    fundflow_df = fundflow_df.dropna(subset=['date']).copy()

    # 2. 排序并重置索引
    fundflow_df.sort_values(by='date', inplace=True)
    fundflow_df.reset_index(drop=True, inplace=True)

    # 3. 打印信息
    print(f"Fundflow df shape after cleaning: {fundflow_df.shape}")
    if not fundflow_df.empty:
        print(f"Fundflow df dates: {fundflow_df['date'].min()} → {fundflow_df['date'].max()}")
    else:
        print(f"Fundflow df dates: 无效日期范围")



    # ✅ 清洗 daily_df（日线数据）
    if 'date' not in daily_df.columns:
        print("❌ daily_df 缺少 'date' 列")
        return pd.DataFrame() # 返回空 DataFrame
    daily_df = daily_df.dropna(subset=['date']).copy() # 添加 .copy()
    # 尝试用多种可能的格式解析 daily_df 的日期
    try:
        daily_df['date'] = pd.to_datetime(daily_df['date'], format='%Y-%m-%d', errors='coerce').dt.date
        print("✅ Daily df 日期用 %Y-%m-%d 格式解析成功")
    except Exception as e:
        print(f"⚠️ Daily df 日期用 %Y-%m-%d 格式解析失败: {e}. 尝试推断格式.")
        daily_df['date'] = pd.to_datetime(daily_df['date'], errors='coerce').dt.date # 依赖推断
        print("✅ Daily df 日期用推断格式解析完成 (可能包含 NaT)")

    daily_df = daily_df.dropna(subset=['date']).copy() # 添加 .copy()
    print(f"Daily df shape after cleaning: {daily_df.shape}")
    if not daily_df.empty:
        print(f"Daily df dates: {daily_df['date'].min()} → {daily_df['date'].max()}")
    else:
         print(f"Daily df dates: 无效日期范围")


    # ✅ Print date ranges before calculating common dates
    print("\n--- Date ranges before checking for common dates ---")
    if not minute_df.empty:
        print(f"Minute df date range: {minute_df['date'].min()} to {minute_df['date'].max()}")
    else:
        print("Minute df is empty.")
    if not fundflow_df.empty:
        print(f"Fundflow df date range: {fundflow_df['date'].min()} to {fundflow_df['date'].max()}")
    else:
        print("Fundflow df is empty.")
    if not daily_df.empty:
        print(f"Daily df date range: {daily_df['date'].min()} to {daily_df['date'].max()}")
    else:
        print("Daily df is empty.")
    print("--------------------------------------------------")


    # ✅ 打印共有日期交集
    common_dates = set()
    # 确保 DataFrame 不为空再尝试获取 date 列
    if not minute_df.empty and not fundflow_df.empty and not daily_df.empty:
         # 确保 date 列是 date/datetime type, and does not contain NaT
        minute_dates = set(minute_df['date'].dropna()) # Remove NaT
        fundflow_dates = set(fundflow_df['date'].dropna()) # Remove NaT
        daily_dates = set(daily_df['date'].dropna()) # Remove NaT

        common_dates = minute_dates & fundflow_dates & daily_dates


    print(f"共有交易日数量：{len(common_dates)}")
    if len(common_dates) == 0:
        print("⚠️ 三个数据源没有交集日期，请检查日期范围或格式")
        return pd.DataFrame() # 返回空 DataFrame


    # ✅ 合并三类数据（按日期） - Implementing the suggested merge strategy
    # Sort dataframes by date before merging
    minute_df.sort_values(by='date', inplace=True)
    daily_df.sort_values(by='date', inplace=True)
    fundflow_df.sort_values(by='date', inplace=True)


    # Merge minute_df with daily_df and fundflow_df on 'date'
    # Use a left merge to keep all minute entries
    # Explicitly drop 'time' from daily_df and fundflow_df before merging to avoid duplicate 'time' columns
    cols_to_drop_daily = ['time'] if 'time' in daily_df.columns else []
    cols_to_drop_fundflow = ['time'] if 'time' in fundflow_df.columns else []

    df_aligned = minute_df.merge(daily_df.drop(columns=cols_to_drop_daily, errors='ignore'), on='date', how='left', suffixes=('_minute', '_daily'))
    df_aligned = df_aligned.merge(fundflow_df.drop(columns=cols_to_drop_fundflow, errors='ignore'), on='date', how='left', suffixes=('_aligned', '_fundflow'))


    df_aligned = df_aligned.sort_values('datetime').reset_index(drop=True) # Sort by the minute-level datetime
    print(f"Aligned df shape: {df_aligned.shape}")
    if not df_aligned.empty:
        print(f"Aligned df columns: {df_aligned.columns.tolist()}") # 打印最终合并后的列名
    else:
        print("Aligned df columns: 合并后 DataFrame 为空")

    # ✅ Print last few rows of df_aligned for verification
    print("\ndf_aligned 最后几行日期：")
    if not df_aligned.empty:
        # Ensure 'open' and 'main_inflow' columns exist before printing
        cols_to_print = ['date', 'open', 'main_inflow']
        existing_cols_to_print = [col for col in cols_to_print if col in df_aligned.columns]
        if existing_cols_to_print:
            print(df_aligned.tail(5)[existing_cols_to_print])
        else:
            print("⚠️ 无法打印 'date', 'open', 或 'main_inflow' 列，它们在 df_aligned 中不存在。")
    else:
        print("⚠️ df_aligned 为空，无法打印最后几行。")


    return df_aligned

# ✅ 构建训练样本
def build_lstm_samples(df, window_size=50, forecast_horizon=8):
    X, y_fundflow, y_direction, y_pct = [], [], [], []
    # 确保 df 不为空且有足够的行
    if df.empty or len(df) < window_size + forecast_horizon:
        print(f"⚠️ 数据不足，无法构建训练样本. 需要至少 {window_size + forecast_horizon} 行数据，但只有 {len(df)} 行.")
        return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

    # Determine features to use dynamically
    # Exclude date, time, and datetime columns
    cols_to_drop_for_features = ['date', 'time_minute', 'time_fundflow', 'datetime', 'time']
    cols_to_drop_for_features = [col for col in cols_to_drop_for_features if col in df.columns]
    feature_columns = [col for col in df.columns if col not in cols_to_drop_for_features]

    if not feature_columns:
        print("❌ 没有可用的数值特征列来构建样本.")
        return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

    # ✅ Identify and filter out columns with all NaN values in the entire DataFrame before building samples
    # This ensures consistency between training and prediction features
    initial_feature_columns_count = len(feature_columns)
    valid_feature_columns = [col for col in feature_columns if df[col].notnull().any()]
    invalid_cols_build = [col for col in feature_columns if col not in valid_feature_columns]

    if invalid_cols_build:
        print(f"⚠️ 在构建训练样本前，发现以下字段在整个数据集中整列缺失或全为 NaN，将从特征中剔除：{invalid_cols_build}")
        feature_columns = valid_feature_columns # Update feature columns to use only valid ones

    if not feature_columns:
        print("❌ 剔除全 NaN 列后，没有可用的数值特征列来构建样本.")
        return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)

    # Check if target columns exist (ensure they are not in the invalid_cols_build list if they were features)
    required_target_cols = ['main_inflow', 'change_pct']
    # Ensure target columns are still available after filtering invalid features
    for target_col in required_target_cols:
         if target_col not in df.columns or target_col in invalid_cols_build:
              print(f"❌ 构建样本所需的目标列 '{target_col}' 缺失或在整个数据集中全为 NaN，无法构建样本.")
              return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


    # Modify the loop step to 5
    for i in range(0, len(df) - window_size - forecast_horizon + 1, 5):
        window_end = i + window_size
        forecast_end = window_end + forecast_horizon

        # 检查窗口和预测期内的数据是否完整
        if forecast_end > len(df):
            break # 确保不会超出数据范围

        window_dates = df.iloc[i:window_end]['date']
        forecast_dates = df.iloc[window_end:forecast_end]['date']



        if len(window_dates) == window_size and len(forecast_dates) == forecast_horizon:

            # Use the filtered feature_columns for the window
            X_window = df.iloc[i:window_end][feature_columns].values
            if not np.issubdtype(X_window.dtype, np.number):
                print(f"⚠️ 窗口数据包含非数值类型，请检查数据清洗和列选择: {X_window.dtype}")
                non_numeric_cols = df.iloc[i:window_end][feature_columns].select_dtypes(exclude=np.number).columns.tolist()
                print(f"非数值列: {non_numeric_cols}")
                continue # Skip current sample

            # ✅ Add NaN/Inf check for the window data
            if np.isnan(X_window).any() or np.isinf(X_window).any():
                print(f"⚠️ 窗口数据存在 NaN 或 Inf，跳过样本 i={i}")
                continue


            X.append(X_window)

            future = df.iloc[window_end:forecast_end].copy() # Create a copy to avoid SettingWithCopyWarning

            # Ensure future data has the required columns and they are numeric
            if not all(col in future.columns for col in required_target_cols):
                 print(f"❌ 未来数据所需列缺失: {', '.join([col for col in required_target_cols if col not in future.columns])}")
                 continue

            if not np.issubdtype(future['main_inflow'].dtype, np.number) or \
               not np.issubdtype(future['change_pct'].dtype, np.number): # Use renamed column names
               print(f"⚠️ 未来数据包含非数值类型，请检查数据清洗: main_inflow({future['main_inflow'].dtype}), change_pct({future['change_pct'].dtype})")
               continue

            # ✅ Handle NaN/Inf in target variables
            future['main_inflow'].replace([np.inf, -np.inf], np.nan, inplace=True)
            future['change_pct'].replace([np.inf, -np.inf], np.nan, inplace=True)

            # Use mean for filling NaN in target variables
            future['main_inflow'].fillna(future['main_inflow'].mean(), inplace=True)
            future['change_pct'].fillna(future['change_pct'].mean(), inplace=True)


            y_fundflow.append(future['main_inflow'].values) # Use renamed column names
            y_direction.append((future['change_pct'] > 0).astype(int).values) # Use renamed column names
            y_pct.append(future['change_pct'].values) # Use renamed column names

    print(f"成功构建样本数量: {len(X)}")
    # Check if input_size matches the number of features in the samples
    if len(X) > 0:
        expected_input_size = len(feature_columns)
        if X[0].shape[1] != expected_input_size:
             print(f"❌ 构建的样本特征数量 ({X[0].shape[1]}) 与预期数量 ({expected_input_size}) 不匹配. 请检查 build_lstm_samples 中的列选择.")


    return np.array(X), np.array(y_fundflow), np.array(y_direction), np.array(y_pct)


# ✅ 训练函数
def train_model(X, y1, y2, y3, config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    # input_size needs to be determined based on the number of columns in the aligned DataFrame, this needs to be obtained dynamically here

    if X.shape[0] == 0:
        raise ValueError(f"Insufficient data to build training samples. Need at least {config.window_size + config.forecast_horizon} rows of data, but only have {len(X)} samples.")

    # Split data into training and validation sets
    X_train, X_val, y1_train, y1_val, y2_train, y2_val, y3_train, y3_val = train_test_split(
        X, y1, y2, y3, test_size=0.2, random_state=42, shuffle=False # Time series data should not be shuffled
    )

    # Scale the input features using the scaler fitted on the training data
    scaler = StandardScaler()
    # Reshape X_train for scaling
    X_train_reshaped = X_train.reshape(-1, X_train.shape[-1])
    X_train_scaled_reshaped = scaler.fit_transform(X_train_reshaped)
    # Reshape back to the original shape
    X_train_scaled = X_train_scaled_reshaped.reshape(X_train.shape)

    # Reshape X_val for scaling and apply the same scaler
    X_val_reshaped = X_val.reshape(-1, X_val.shape[-1])
    X_val_scaled_reshaped = scaler.transform(X_val_reshaped)
    # Reshape back to the original shape
    X_val_scaled = X_val_scaled_reshaped.reshape(X_val.shape)


    current_input_size = X_train_scaled.shape[2]
    print(f"Model input feature quantity (input_size): {current_input_size}")

    # Update Config's input_size
    config.input_size = current_input_size

    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.lr)

    # Ensure y1, y2, y3 also have enough samples
    if len(X_train_scaled) != len(y1_train) or len(X_train_scaled) != len(y2_train) or len(X_train_scaled) != len(y3_train):
         print(f"❌ Number of training samples ({len(X_train_scaled)}) does not match the number of target samples ({len(y1_train)}, {len(y2_train)}, {len(y3_train)}). Please check the build_lstm_samples function and train_test_split.")
         raise ValueError("Number of training samples does not match the number of target samples")

    if len(X_val_scaled) != len(y1_val) or len(X_val_scaled) != len(y2_val) or len(X_val_scaled) != len(y3_val):
         print(f"❌ Number of validation samples ({len(X_val_scaled)}) does not match the number of target samples ({len(y1_val)}, {len(y2_val)}, {len(y3_val)}). Please check the build_lstm_samples function and train_test_split.")
         raise ValueError("Number of validation samples does not match the number of target samples")


    # Use TensorDataset and DataLoader for mini-batch training and validation
    train_dataset = TensorDataset(torch.tensor(X_train_scaled).float(), torch.tensor(y1_train).float(), torch.tensor(y2_train).float(), torch.tensor(y3_train).float())
    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=False) # Keep shuffle=False for time series data

    val_dataset = TensorDataset(torch.tensor(X_val_scaled).float(), torch.tensor(y1_val).float(), torch.tensor(y2_val).float(), torch.tensor(y3_val).float())
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False)


    best_val_loss = float('inf')
    for epoch in range(config.epochs):
        model.train()
        total_train_loss = 0
        for batch_X, batch_y1, batch_y2, batch_y3 in train_loader:
            batch_X = batch_X.to(device)
            batch_y1 = batch_y1.to(device)
            batch_y2 = batch_y2.to(device)
            batch_y3 = batch_y3.to(device)

            pred1, pred2, pred3 = model(batch_X)
            loss1 = nn.MSELoss()(pred1, batch_y1)
            loss2 = nn.BCEWithLogitsLoss()(pred2, batch_y2.float()) # BCEWithLogitsLoss requires float type target
            loss3 = nn.MSELoss()(pred3, batch_y3)
            loss = loss1 + loss2 + loss3

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item()

        # Evaluate on validation set
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for batch_X_val, batch_y1_val, batch_y2_val, batch_y3_val in val_loader:
                batch_X_val = batch_X_val.to(device)
                batch_y1_val = batch_y1_val.to(device)
                batch_y2_val = batch_y2_val.to(device)
                batch_y3_val = batch_y3_val.to(device)

                pred1_val, pred2_val, pred3_val = model(batch_X_val)
                loss1_val = nn.MSELoss()(pred1_val, batch_y1_val)
                loss2_val = nn.BCEWithLogitsLoss()(pred2_val, batch_y2_val.float())
                loss3_val = nn.MSELoss()(pred3_val, batch_y3_val)
                val_loss = loss1_val + loss2_val + loss3_val
                total_val_loss += val_loss.item()

        avg_train_loss = total_train_loss / len(train_loader)
        avg_val_loss = total_val_loss / len(val_loader)
        print(f"Epoch {epoch}: TrainLoss={avg_train_loss:.4f}, ValLoss={avg_val_loss:.4f}")


        # Save the best model based on validation loss
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), config.model_path)
            print(f"✅ Saved best model at epoch {epoch} with ValLoss={best_val_loss:.4f}")


    # Return the scaler
    return scaler

# ✅ Load Model
def load_model(config):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
     # input_size needs to be determined based on the number of columns in the aligned DataFrame

    # Check if config.input_size is set (it should be updated in train_model)
    if not hasattr(config, 'input_size') or config.input_size is None:
        print("❌ config.input_size is not set. Cannot load model.")
        return None

    model = MultiTargetLSTM(config.input_size, config.hidden_size, config.num_layers, config.forecast_horizon, config.dropout).to(device)
    # Check if model file exists
    if not os.path.exists(config.model_path):
        print(f"❌ Model file does not exist: {config.model_path}. Please train the model first.")
        return None # Return None to indicate loading failed

    model.load_state_dict(torch.load(config.model_path, map_location=device))
    model.eval()
    return model

# ✅ Build Latest Sample (last 50 days)
def prepare_latest_sample(df, window_size=50, scaler=None):
    if df.empty or len(df) < window_size:
        print(f"⚠️ Insufficient data to build the latest prediction sample. Need at least {window_size} rows of data, but only have {len(df)} rows.")
        return None # Return None to indicate sample could not be built

    # Determine features to use dynamically - Use the same valid features determined during training
    # This requires the list of valid features to be passed or determined consistently

    # For now, let's assume feature_columns here should be the same as used in build_lstm_samples
    # A better way is to return the list of valid features from build_lstm_samples or align_stock_data
    # For this fix, we will re-determine valid features based on the *entire* df here for simplicity,
    # but ideally, this should be a consistent list passed from the training phase.

    cols_to_drop_for_features = ['date', 'time_minute', 'time_fundflow', 'datetime', 'time']
    cols_to_drop_for_features = [col for col in cols_to_drop_for_features if col in df.columns]
    all_potential_features = [col for col in df.columns if col not in cols_to_drop_for_features]

    # ✅ Identify and filter out columns with all NaN values in the entire DataFrame
    # This ensures consistency between training and prediction features
    valid_feature_columns = [col for col in all_potential_features if df[col].notnull().any()]

    if not valid_feature_columns:
        print("❌ 在整个数据集中没有可用的数值特征列来构建最新样本.")
        return None

    # ✅ Find the latest valid date from daily or fundflow data in the aligned DataFrame
    latest_daily_date = df['date'].dropna().max()
    latest_fundflow_date = df['date'].dropna().max() # Assuming fundflow dates are also in the 'date' column after merge

    # Use the earliest of the latest dates from reliable sources
    latest_valid_date = None
    if pd.notnull(latest_daily_date) and pd.notnull(latest_fundflow_date):
        latest_valid_date = max(latest_daily_date, latest_fundflow_date)
    elif pd.notnull(latest_daily_date):
        latest_valid_date = latest_daily_date
    elif pd.notnull(latest_fundflow_date):
        latest_valid_date = latest_fundflow_date

    if latest_valid_date is None:
         print("❌ 无法找到资金流或日线数据中最晚的有效日期，无法构建预测样本。")
         return None

    print(f"ℹ️ Using the latest valid date from daily/fundflow data to anchor the prediction sample: {latest_valid_date}")

    # Find the index of the latest entry in df_aligned that matches the latest_valid_date
    # We need to find the last row for this date to get the most recent data point
    latest_valid_index = df[df['date'] == latest_valid_date].index.max()

    if pd.isnull(latest_valid_index) or latest_valid_index < window_size -1:
         print(f"⚠️ 在对齐数据中未找到最新有效日期 ({latest_valid_date}) 的足够历史数据来构建样本 (需要 {window_size} 行). 找到的最新有效索引: {latest_valid_index}. DataFrame 长度: {len(df)}")
         return None


    # Extract the window_size rows ending at the latest_valid_index, using only valid features
    start_index = latest_valid_index - window_size + 1
    if start_index < 0:
        print(f"⚠️ 数据不足，无法构建预测样本. 需要至少 {window_size} 行数据，但从最新有效日期回溯只有 {latest_valid_index + 1} 行.")
        return None

    latest = df.iloc[start_index:latest_valid_index + 1][valid_feature_columns].copy()


    # ✅ Print shape before cleaning
    print(f"Raw latest sample shape before cleaning: {latest.shape}")


    # ✅ Add explicit cleaning for NaN and Inf values before building the sample
    print("\n--- Cleaning latest sample data ---")
    initial_latest_shape = latest.shape
    latest = latest.replace([np.inf, -np.inf], np.nan)

    # ✅ Identify and filter out columns with all NaN values in the latest sample
    # This step is now less critical as we filtered based on the entire df, but kept for robustness
    invalid_cols_latest = [col for col in latest.columns if latest[col].isnull().all()]
    if invalid_cols_latest:
        print(f"⚠️ 以下字段在最新样本中整列缺失，将从样本中剔除：{invalid_cols_latest}")
        latest = latest.drop(columns=invalid_cols_latest)

    # ✅ Fill remaining NaN values using ffill and then bfill
    latest.fillna(method='ffill', inplace=True)
    latest.fillna(method='bfill', inplace=True)


    print(f"Latest sample data shape after cleaning: {latest.shape}")
    print("NaN counts in latest sample after cleaning:")
    print(latest.isnull().sum())
    print("Inf counts in latest sample after cleaning:")
    print(np.isinf(latest).sum())
    print("--- Cleaning latest sample data finished ---")

    # ✅ Print latest sample data before cleaning if it becomes empty after cleaning
    # This condition is now less likely with ffill/bfill and pre-filtering, but kept for robustness
    if latest.empty and not initial_latest_shape[0] == 0:
        print("\n⚠️ Latest sample data became empty after cleaning. Printing raw latest sample data before cleaning:")
        display(df.iloc[start_index:latest_valid_index + 1][valid_feature_columns]) # Print the raw data before cleaning


    # Check if it contains non-numeric columns
    if not np.issubdtype(latest.values.dtype, np.number):
        print(f"⚠️ Latest sample data contains non-numeric types. Please check data cleaning and column selection: {latest.values.dtype}")
        non_numeric_cols = latest.select_dtypes(exclude=np.number).columns.tolist()
        print(f"Non-numeric columns: {non_numeric_cols}")
        return None # Return None to indicate sample could not be built

    # Ensure the sample still has enough data after dropping columns with NaNs and filling
    if len(latest) != window_size:
         print(f"⚠️ Latest sample data has incorrect number of rows ({len(latest)}) after cleaning to build a sample of size {window_size}. Cannot build sample.")
         return None


    # ✅ Check for NaN/Inf values after cleaning and before scaling/tensor conversion
    if np.isnan(latest.values).any() or np.isinf(latest.values).any():
         print(f"❌ Latest sample data still contains NaN or Inf values after cleaning. Cannot build sample.")
         return None


    X = latest.values.reshape(1, window_size, -1)

    # ✅ Update config.input_size based on the number of features in the latest sample after cleaning
    # This should now be consistent with the features used for training
    config.input_size = latest.shape[1]
    print(f"ℹ️ Updating config.input_size based on the latest sample after cleaning: {config.input_size}")


    # Scale the latest sample using the fitted scaler
    if scaler is not None:
        X_reshaped = X.reshape(-1, X.shape[-1])
        # Ensure the scaler has the correct number of features
        # This check is important because columns might have been dropped
        if scaler.n_features_in_ != X_reshaped.shape[1]:
             print(f"❌ Scaler was fitted on {scaler.n_features_in_} features, but the latest sample has {X_reshaped.shape[1]} features after cleaning. Cannot scale. This indicates an inconsistency in the features used during training and prediction.")
             # Instead of returning None, try to adjust the scaler if possible (less ideal)
             # Or, ideally, the build_lstm_samples should return the valid_feature_columns to train_model
             # For now, let's print a clear error and return None, emphasizing retraining.
             return None # Cannot scale if feature counts don't match

        X_scaled_reshaped = scaler.transform(X_reshaped)
        # Reshape back to the original shape
        X_scaled = X_scaled_reshaped.reshape(X.shape)
        return torch.tensor(X_scaled).float()
    else:
        print("⚠️ Scaler not provided for prediction. Using unscaled data.")
        return torch.tensor(X).float()


# ✅ Predict Future 8 days
def predict_future(model, X, config):
    if model is None or X is None: # Check if model and sample were built successfully
        print("❌ Unable to perform prediction. Model or sample building failed.")
        return None, None, None

    # ✅ Check if model input size matches the sample size
    if X.shape[-1] != model.lstm.input_size:
         print(f"❌ Model input size ({model.lstm.input_size}) does not match the sample feature count ({X.shape[-1]}). Cannot predict.")
         return None, None, None


    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    X = X.to(device)
    with torch.no_grad():
        pred_fundflow, pred_direction, pred_pct = model(X)
        pred_fundflow = pred_fundflow.cpu().numpy().flatten()
        pred_direction = (torch.sigmoid(pred_direction) > 0.5).int().cpu().numpy().flatten()
        pred_pct = pred_pct.cpu().numpy().flatten()
    return pred_fundflow, pred_direction, pred_pct

# ✅ Print Prediction Results
def print_predictions(pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # Check if prediction was successful
        print("❌ Unable to perform prediction. Please check data and sample building.")
        return
    print("📊 Future 8-day Prediction Results:")
    for i in range(8):
        direction = "📈 Up" if pred_direction[i] == 1 else "📉 Down"
        print(f"Day {i+1}: Main Inflow={pred_fundflow[i]:.2f}, Direction={direction}, Change %={pred_pct[i]:.2%}")

# ✅ Save Prediction Results to CSV
def save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct):
    if pred_fundflow is None: # Check if prediction was successful
        print("❌ Unable to save prediction results because prediction failed.")
        return
    df_pred = pd.DataFrame({
        'Day': [f'Day {i+1}' for i in range(8)],
        '主力净流入': pred_fundflow,
        '涨跌方向': ['涨' if d == 1 else '跌' for d in pred_direction],
        '涨跌幅': pred_pct
    })
    save_path = f'/content/drive/MyDrive/Colab Notebooks/ant_data/{prefix}_预测结果.csv'
    try:
        # Ensure the directory exists before saving
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        df_pred.to_csv(save_path, index=False)
        print(f"✅ Prediction completed, results saved to CSV file: {save_path}")
    except Exception as e:
        print(f"❌ Failed to save prediction results to CSV file: {e}")


# ✅ Batch process multiple stocks
stock_prefixes = ['002269_美邦服饰'] # Updated stock prefix

# ✅ Minute data (002269_美邦服饰_合并.csv)
minute_rename = {
    '日期': 'date',
    '时间': 'time',
    '主力净流入': 'main_inflow',
    '超大单净流入': 'super_inflow',
    '大单净流入': 'large_inflow',
    '总成交额': 'total_amount'
}
# ✅ Fund flow data (sz002269.csv)
fundflow_rename = {
    'opendate': 'date',
    'trade': 'time', # Note: The 'time' column in fundflow_df might represent something else, need to confirm its use or drop it after merging
    'changeratio': 'change_ratio',
    'turnover': 'turnover',
    'netamount': 'net_amount',
    'ratioamount': 'ratio_amount'
    # 'r0': 'r0',
    # 'r1': 'r1',
    # 'r2': 'r2', 'r3': 'r3',
    # 'r0_net': 'r0_net',
    # 'r1_net': 'r1_net',
    # 'r2_net': 'r2_net',
    # 'r3_net': 'r3_net'
}
# ✅ Daily data
daily_rename = {
    'Date': 'date',
    '股票代码': 'stock_code',
    'Open': 'open',
    'Close': 'close',
    'High': 'high',
    'Low': 'low',
    'Volume': 'volume',
    'Amount': 'amount',
    'Amplitude': 'amplitude',
    'ChangePct': 'change_pct',
    'ChangeAmt': 'change_amt',
    'Turnover': 'turnover_rate',
    'Code': 'code'
}


# Main processing loop
for prefix in stock_prefixes:
    try:
        print(f"🚀 Starting processing: {prefix}")

       # Build file paths - ✅ Ensure data roles are correct
        # Updated file paths based on user input
        daily_path = f"/content/drive/MyDrive/Colab Notebooks/ant_data/kline/{prefix}_daily.csv"
        minute_path = f"/content/drive/MyDrive/Colab Notebooks/ant_data/to_fns/{prefix}_合并_clipped_clipped_clipped_clipped.csv"
        fundflow_path = f"/content/drive/MyDrive/Colab Notebooks/ant_data/jiqi_do/sz{prefix[0:6]}.csv"


        print(f"Loading file: {minute_path}")
        minute_df_raw = pd.read_csv(minute_path)
        print(f"Loading file: {fundflow_path}")
        fundflow_df_raw = pd.read_csv(fundflow_path)
        print(f"Loading file: {daily_path}")
        daily_df_raw = pd.read_csv(daily_path)

        # ✅ Apply renaming
        minute_df = minute_df_raw.rename(columns=minute_rename)
        fundflow_df = fundflow_df_raw.rename(columns=fundflow_rename)
        daily_df = daily_df_raw.rename(columns=daily_rename)

        # ✅ Add numeric cleaning before alignment
        print("\n--- Starting numeric data cleaning ---")
        minute_df = clean_numeric_columns(minute_df)
        fundflow_df = clean_numeric_columns(fundflow_df)
        daily_df = clean_numeric_columns(daily_df)
        print("--- Numeric data cleaning finished ---")

        # ✅ Apply logarithmic compression to fund flow related columns
        print("\n--- Starting logarithmic compression ---")
        minute_df = compress_fundflow(minute_df, cols=['main_inflow', 'super_inflow', 'large_inflow', 'total_amount'])
        # Update compress_fundflow call for fundflow_df based on the new columns
        fundflow_df = compress_fundflow(fundflow_df, cols=['net_amount', 'ratio_amount'])
        # Also apply to daily data columns that might benefit, e.g., volume and amount
        daily_df = compress_fundflow(daily_df, cols=['volume', 'amount'])

        print("--- Logarithmic compression finished ---")


        # ✅ Print first few rows and columns after renaming and compression (for debugging)
        print("\n✅ Minute data sample after renaming, cleaning and compression:")
        display(minute_df.head())
        print("\n✅ Fund flow data sample after renaming, cleaning and compression:")
        display(fundflow_df.head())
        print("\n✅ Daily data sample after renaming, cleaning and compression:")
        display(daily_df.head())
        print("-" * 30) # Separator


        # # ✅ Filter out '2025-09-29' from minute_df
        # # Keeping this filter as the user indicated this data is erroneous
        # date_to_exclude = dt.date(2025, 9, 29)
        # initial_minute_rows = len(minute_df)
        # minute_df = minute_df[minute_df['date'] != date_to_exclude].copy() # Add .copy()
        # filtered_minute_rows = len(minute_df)
        # if initial_minute_rows > filtered_minute_rows:
        #     print(f"✅ Filtered out data for {date_to_exclude} from minute_df. Removed {initial_minute_rows - filtered_minute_rows} rows.")
        # else:
        #     print(f"ℹ️ No data found for {date_to_exclude} in minute_df to filter.")

        # # Column check (after renaming and filtering)
        # print("Minute data columns after renaming and filtering:", minute_df.columns.tolist())
        # print("Fund flow data columns after renaming:", fundflow_df.columns.tolist())
        # print("Daily data columns after renaming:", daily_df.columns.tolist())


        # Data alignment and sample building
        print("\n--- Starting data alignment ---")
        # ✅ Pass the renamed and cleaned DataFrames to align_stock_data
        df_aligned = align_stock_data(minute_df, fundflow_df, daily_df)
        print("--- Data alignment finished ---")


        # Check aligned data
        if df_aligned.empty:
            print(f"❌ Data alignment failed, skipping subsequent steps for {prefix}")
            continue # Skip processing for the current stock

        required_rows = Config.window_size + Config.forecast_horizon # Get parameters from Config
        if len(df_aligned) < required_rows:
            print(f"❌ Insufficient data to build training samples. Need at least {required_rows} rows of data, but only have {len(df_aligned)} rows. Skipping processing for {prefix}")
            continue # Skip processing for the current stock


        print("\n✅ Building training samples...")
        # input_size will be determined in build_lstm_samples based on the shape of df_aligned
        # build_lstm_samples will now also filter out columns with all NaNs in the entire df
        X, y1, y2, y3 = build_lstm_samples(df_aligned, window_size=Config.window_size, forecast_horizon=Config.forecast_horizon)

        if X.shape[0] == 0:
            print(f"❌ Sample building failed, skipping subsequent steps for {prefix}")
            continue # Skip processing for the current stock

        # Train model
        config = Config() # Create a new Config instance
        # input_size will be updated in train_model based on X's shape (which now only contains valid features)
        config.model_path = f"/content/drive/MyDrive/Colab Notebooks/ant_data/{prefix}_best_model.pt"

        print("✅ Starting model training...")
        # Pass the config instance
        # train_model will now receive X with potentially fewer features and fit the scaler accordingly
        scaler = train_model(X, y1, y2, y3, config) # Capture the fitted scaler
        print("✅ Model training completed, starting prediction...")


        # Predict and output
        print("\n✅ Loading model and performing prediction...")

        # The config.input_size is updated in train_model based on the actual features used for training.
        # We need to ensure the model is loaded with this correct input size.
        # A better approach is to store the list of valid features or the input_size determined during training
        # and use it here for loading the model. For now, we'll rely on config.input_size being updated.

        # Check if config.input_size was updated in train_model
        if not hasattr(config, 'input_size') or config.input_size is None or config.input_size <= 0:
             print("❌ Model input size was not correctly determined during training. Cannot load model.")
             continue


        model = load_model(config) # Load model with the updated config.input_size

        if model is None: # Check if model was loaded successfully
            print(f"❌ Model loading failed, skipping prediction for {prefix}")
            continue


        # prepare_latest_sample will now also filter out columns with all NaNs in the entire df
        # and use the same logic as build_lstm_samples for feature selection.
        X_latest = prepare_latest_sample(df_aligned, window_size=Config.window_size, scaler=scaler) # Pass the scaler to prepare_latest_sample

        if X_latest is None: # Check if latest sample was built successfully
             print(f"❌ Failed to build latest prediction sample for {prefix}. Skipping prediction.")
             continue


        pred_fundflow, pred_direction, pred_pct = predict_future(model, X_latest, config)

        print_predictions(pred_fundflow, pred_direction, pred_pct)

        # Save results
        save_predictions_to_csv(prefix, pred_fundflow, pred_direction, pred_pct)


    except Exception as e:
        print(f"❌ Error processing {prefix}: {e}")
```
