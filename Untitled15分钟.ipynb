### 重新生成数据加载和预处理代码

根据错误信息和已有的变量，我猜测您可能删除了用于加载、预处理股票分钟数据并进行特征工程的代码。以下是一个常见的代码框架，旨在帮助您重新构建这部分逻辑。请特别注意在加载数据或选择特征时，将 `Amount` 列名替换为正确的 `Amount_Fundflow`（如果您的数据中是这个名称的话）。

````python
import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import zipfile

# 设置你的数据目录 (保持不变，用于保存模型和图表)
base_dir = '/content/drive/MyDrive/Colab Notebooks/ant_data'

# 股票代码列表：先测试 2 支股票
stock_codes = [ "601012","601138"]

# 股票名称映射
stock_name_mapping = {
    "601138": "工业富联",
    "601012": "隆基绿能",
}

print(f"🔥 已加载股票代码列表，共 {len(stock_codes)} 支股票")

# 时间范围
start_date = "20251110"
end_date = datetime.today().strftime("%Y%m%d")

# 保存路径
save_path = "./data/"
os.makedirs(save_path, exist_ok=True)

# Minutely parameters
minutes_per_trading_day = 240
input_window_minutes = int(8 * minutes_per_trading_day)   # 输入窗口：2.5天 = 600 分钟
prediction_window_minutes = int(1.5 * minutes_per_trading_day)  # 预测窗口：0.5天 = 120 分钟

print(f"分钟级输入窗口大小: {input_window_minutes} 分钟 ({input_window_minutes / minutes_per_trading_day} 天)")
print(f"分钟级预测窗口大小: {prediction_window_minutes} 分钟 ({prediction_window_minutes / minutes_per_trading_day} 天)")

# 技术指标窗口
MA_WINDOW_5_MINUTES = 5 * minutes_per_trading_day
MA_WINDOW_10_MINUTES = 10 * minutes_per_trading_day
EMA_WINDOW_12_MINUTES = 12 * minutes_per_trading_day
EMA_WINDOW_26_MINUTES = 26 * minutes_per_trading_day
RSI_WINDOW_14_MINUTES = 14 * minutes_per_trading_day
VOLATILITY_WINDOW_5_MINUTES = 5 * minutes_per_trading_day

# 字段标准化函数
def standardize_columns(df, type='kline'):
    if type == 'kline':
        column_mapping = {
            "day": "Date", "open": "Open", "close": "Close",
            "high": "High", "low": "Low", "volume": "Volume",
            "amount": "Amount",
        }
    elif type == 'fundflow':
        column_mapping = {
            "主力净流入": "MainNetInflow",
            "总成交额": "Amount_Fundflow",
        }
    else:
        column_mapping = {}
    valid_columns = {k: v for k, v in column_mapping.items() if k in df.columns}
    df.rename(columns=valid_columns, inplace=True)
    return df

# RSI计算函数
def compute_rsi(series, window):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window, min_periods=1).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window, min_periods=1).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# 数据加载与特征工程
def load_and_prepare_data(code, start_date_str, end_date_str, minute_interval="1"):
    stock_name = stock_name_mapping.get(code)
    if not stock_name:
        print(f"⚠️ 未找到股票代码 {code} 对应的股票名称，跳过。")
        return None, None, None, None

    filename = f"{code}_{stock_name}_合并.csv"
    fundflow_file_path = os.path.join(base_dir, "to_fns", filename)
    kline_file_path = os.path.join(base_dir, "ESC", filename)

    try:
        df_kline_raw = pd.read_csv(kline_file_path)
        df_fundflow_raw = pd.read_csv(fundflow_file_path)

        if df_kline_raw.empty or df_fundflow_raw.empty:
            print(f"⚠️ 未能加载 {code} 的部分或全部数据，跳过。")
            return None, None, None, None

        df_kline = standardize_columns(df_kline_raw.copy(), type='kline')
        df_kline["Date"] = pd.to_datetime(df_kline["Date"])

        df_fundflow = df_fundflow_raw.copy()
        df_fundflow['Date'] = pd.to_datetime(df_fundflow['日期'].astype(str) + ' ' + df_fundflow['时间'])
        df_fundflow = standardize_columns(df_fundflow, type='fundflow')

        df_merged = pd.merge(df_kline, df_fundflow[['Date', 'MainNetInflow', 'Amount_Fundflow']], on='Date', how='inner')
        df_merged.sort_values("Date", inplace=True)
        df_merged.set_index("Date", inplace=True)

        # 技术指标
        df_merged["MA5"] = df_merged["Close"].rolling(window=MA_WINDOW_5_MINUTES, min_periods=1).mean()
        df_merged["MA10"] = df_merged["Close"].rolling(window=MA_WINDOW_10_MINUTES, min_periods=1).mean()
        df_merged["EMA12"] = df_merged["Close"].ewm(span=EMA_WINDOW_12_MINUTES, adjust=False).mean()
        df_merged["EMA26"] = df_merged["Close"].ewm(span=EMA_WINDOW_26_MINUTES, adjust=False).mean()
        df_merged["MACD"] = df_merged["EMA12"] - df_merged["EMA26"]
        df_merged["RSI"] = compute_rsi(df_merged["Close"], window=RSI_WINDOW_14_MINUTES)
        df_merged["Return_1m"] = df_merged["Close"].pct_change()
        df_merged["Volatility_5m"] = df_merged["Close"].rolling(window=VOLATILITY_WINDOW_5_MINUTES, min_periods=1).std()

        if 'MainNetInflow' in df_merged.columns:
            df_merged['MainNetInflowChange'] = df_merged['MainNetInflow'].diff()

        df_merged["Hour"] = df_merged.index.hour
        df_merged["Minute"] = df_merged.index.minute
        df_merged["DayOfWeek"] = df_merged.index.dayofweek

        df_merged.dropna(inplace=True)

        features = [
            "Open","High","Low","Close","Volume",
            "MA5","MA10","EMA12","EMA26","MACD","RSI",
            "Return_1m","Volatility_5m","Hour","Minute","DayOfWeek",
            "MainNetInflow","Amount_Fundflow","MainNetInflowChange"
        ]
        actual_features = [f for f in features if f in df_merged.columns]

        if 'Close' not in actual_features:
            return None, None, None, None

        scaler = MinMaxScaler()
        df_scaled = scaler.fit_transform(df_merged[actual_features])

        return df_scaled, df_merged["Close"].values, scaler, actual_features
    except Exception as e:
        print(f"❌ 获取或处理 {code} 的分钟数据失败：{e}")
        return None, None, None, None

# 构造时间序列样本
def create_sequences(data_scaled, close_prices, window_size, pred_size):
    X, y = [], []
    # Ensure window_size and pred_size are integers for range() function and indexing
    window_size = int(window_size)
    pred_size = int(pred_size)
    if len(data_scaled) < window_size + pred_size:
        return np.array(X), np.array(y)
    for i in range(len(data_scaled) - window_size - pred_size + 1):
        X.append(data_scaled[i:i+window_size])
        y.append(close_prices[i+window_size:i+window_size+pred_size])
    return np.array(X), np.array(y)

# 构建LSTM模型
def build_lstm_model(input_shape, output_size):
    model = Sequential()
    # Recommended Keras API usage: Add Input layer explicitly
    model.add(Input(shape=input_shape))
    model.add(LSTM(64, return_sequences=True))
    model.add(LSTM(32))
    model.add(Dense(output_size))
    model.compile(optimizer='adam', loss='mse')
    return model

# 模型训练
def train_lstm_model(X, y, input_shape, output_size, model_path):
    model = build_lstm_model(input_shape, output_size=output_size)
    checkpoint = ModelCheckpoint(model_path.replace(".h5",".keras"),
                                 monitor='loss', save_best_only=True, verbose=1)
    history = model.fit(X, y, epochs=30, batch_size=16, callbacks=[checkpoint], verbose=0)
    print(f"✨ 模型训练完成，最终MSE: {history.history['loss'][-1]:.4f}")
    return model

# 预测并绘图
def predict_and_plot(model, scaled_data, scaler, window_size, stock_code, output_dir="./charts", actual_features=None):
    if actual_features is None:
        print("Error: actual_features list not provided to predict_and_plot.")
        return None

    # Ensure window_size and prediction_window_minutes are integers
    window_size = int(window_size)
    prediction_window_minutes_int = int(prediction_window_minutes)

    if len(scaled_data) < window_size:
        print(f"Warning: Not enough data for prediction for {stock_code}. Need at least {window_size} data points, but got {len(scaled_data)}.")
        return None

    last_window = scaled_data[-window_size:]
    prediction = model.predict(np.expand_dims(last_window, axis=0))[0]

    # 创建一个虚拟数组用于逆缩放，其形状为 (prediction_window_minutes_int, len(actual_features))
    dummy = np.zeros((prediction_window_minutes_int, len(actual_features)))

    # 找到 "Close" 列在 actual_features 中的索引
    try:
        close_index = actual_features.index("Close")
    except ValueError:
        print("Error: 'Close' not found in actual_features list.")
        return None

    # 用最后已知的值填充非预测特征，用预测值填充 'Close' 列
    for i in range(prediction_window_minutes_int):
        dummy[i, :] = scaled_data[-1, :]  # 复制最后一个观测的特征集
        dummy[i, close_index] = prediction[i]  # 替换为预测的收盘价

    recovered = scaler.inverse_transform(dummy)[:, close_index]  # 逆缩放并获取收盘价

    # 创建输出目录
    os.makedirs(output_dir, exist_ok=True)
    chart_path = os.path.join(output_dir, f"{stock_code}_{datetime.today().strftime('%Y-%m-%d_%H%M%S')}_minute_forecast.png")

    plt.figure(figsize=(12, 6))
    plt.plot(range(1, prediction_window_minutes_int + 1), recovered, marker='o', linestyle='-', label="预测收盘价")
    plt.title(f"{stock_code} - 未来 {prediction_window_minutes_int} 分钟收盘价预测")
    plt.xlabel("未来分钟数 (每步1分钟)")
    plt.ylabel("收盘价")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path)
    plt.close()

    print(f"✅ 图表已保存：{chart_path}")
    return recovered

# 打包所有图表为 zip 文件
def zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip"):
    today_date_str = datetime.today().strftime("%Y-%m-%d")
    zip_file_path = os.path.join(base_dir, zip_name)
    with zipfile.ZipFile(zip_file_path, 'w') as zipf:
        for file in os.listdir(chart_dir):
            if file.endswith("minute_forecast.png") and today_date_str in file:
                zipf.write(os.path.join(chart_dir, file), arcname=file)
    print(f"📦 今日图表已打包为：{zip_file_path}")

# 主流程
scaled_stock_data = {}
for code in stock_codes:
    scaled_data, close_prices, scaler, actual_features = load_and_prepare_data(code, start_date, end_date, minute_interval="1")
    if scaled_data is not None:
        scaled_stock_data[code] = (scaled_data, close_prices, scaler, actual_features)

os.makedirs("./models", exist_ok=True)
chart_dir = "./charts"
os.makedirs(chart_dir, exist_ok=True)

for code in stock_codes:
    if code not in scaled_stock_data:
        continue

    scaled_data, close_prices, scaler, actual_features = scaled_stock_data[code]
    X, y = create_sequences(scaled_data, close_prices, window_size=input_window_minutes, pred_size=prediction_window_minutes)

    if X.shape[0] == 0:
        print(f"⚠️ 样本数量不足，跳过训练：{code}")
        continue

    model_path = f"./models/{code}_lstm_minute.h5"
    model = train_lstm_model(X, y, input_shape=X.shape[1:], output_size=prediction_window_minutes, model_path=model_path)
    predict_and_plot(model, scaled_data, scaler, window_size=input_window_minutes, stock_code=code, output_dir=chart_dir, actual_features=actual_features)

zip_today_charts(chart_dir)
````

## Update Granularity and Parameters

### Subtask:
Introduce a `GRANULARITY_FACTOR` of 15, adjust `minutes_per_trading_day` to reflect 15-minute intervals, and recalculate all window sizes for input, prediction, and technical indicators based on this new granularity.


**Reasoning**:
The subtask requires updating the granularity and recalculating various time-based parameters. I will modify the existing code cell to introduce `GRANULARITY_FACTOR`, `TRADING_DAY_IN_1_MIN_INTERVALS`, and update all relevant window sizes based on the new 15-minute granularity.



````python
import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import zipfile

# 设置你的数据目录 (保持不变，用于保存模型和图表)
base_dir = '/content/drive/MyDrive/Colab Notebooks'

# 股票代码列表：先测试 2 支股票
stock_codes = [ "601138"]

# 股票名称映射
stock_name_mapping = {
    "601138": "工业富联",
    # "601012": "隆基绿能",
}

print(f"🔥 已加载股票代码列表，共 {len(stock_codes)} 支股票")

# 时间范围
start_date = "20251110"
end_date = datetime.today().strftime("%Y%m%d")

# 保存路径
save_path = "./data/"
os.makedirs(save_path, exist_ok=True)

# --- 更新的分钟级参数 --- #
# 引入粒度因子，表示15分钟间隔
GRANULARITY_FACTOR = 15
# 1分钟间隔下，一个交易日总分钟数 (例如：A股上午2小时，下午2小时，共4小时 = 240分钟)
TRADING_DAY_IN_1_MIN_INTERVALS = 240

# Minutely parameters (现在指代15分钟为单位的'分钟'，即15分钟为一个数据点)
minutes_per_trading_day = TRADING_DAY_IN_1_MIN_INTERVALS / GRANULARITY_FACTOR

# 输入窗口：2.5天 = 600 分钟 / 15 = 40个15分钟数据点
input_window_minutes = int((2.5 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
# 预测窗口：0.5天 = 120 分钟 / 15 = 8个15分钟数据点
prediction_window_minutes = int((0.5 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)

print(f"粒度因子 (GRANULARITY_FACTOR): {GRANULARITY_FACTOR}")
print(f"每个交易日的'分钟'数 (15分钟单位): {minutes_per_trading_day} ('分钟')")
print(f"分钟级输入窗口大小: {input_window_minutes} ('分钟') ({input_window_minutes / minutes_per_trading_day} 天)")
print(f"分钟级预测窗口大小: {prediction_window_minutes} ('分钟') ({prediction_window_minutes / minutes_per_trading_day} 天)")

# 技术指标窗口 (全部除以 GRANULARITY_FACTOR)
MA_WINDOW_5_MINUTES = int((5 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
MA_WINDOW_10_MINUTES = int((10 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
EMA_WINDOW_12_MINUTES = int((12 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
EMA_WINDOW_26_MINUTES = int((26 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
RSI_WINDOW_14_MINUTES = int((14 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)
VOLATILITY_WINDOW_5_MINUTES = int((5 * TRADING_DAY_IN_1_MIN_INTERVALS) / GRANULARITY_FACTOR)

# 字段标准化函数
def standardize_columns(df, type='kline'):
    if type == 'kline':
        column_mapping = {
            "day": "Date", "open": "Open", "close": "Close",
            "high": "High", "low": "Low", "volume": "Volume",
            "amount": "Amount",
        }
    elif type == 'fundflow':
        column_mapping = {
            "主力净流入": "MainNetInflow",
            "总成交额": "Amount_Fundflow",
        }
    else:
        column_mapping = {}
    valid_columns = {k: v for k, v in column_mapping.items() if k in df.columns}
    df.rename(columns=valid_columns, inplace=True)
    return df

# RSI计算函数
def compute_rsi(series, window):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window, min_periods=1).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window, min_periods=1).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# 数据加载与特征工程
def load_and_prepare_data(code, start_date_str, end_date_str, minute_interval="1"):
    stock_name = stock_name_mapping.get(code)
    if not stock_name:
        print(f"⚠️ 未找到股票代码 {code} 对应的股票名称，跳过。")
        return None, None, None, None

    filename = f"{code}_{stock_name}_合并.csv"
    fundflow_file_path = os.path.join(base_dir, "to_fns", filename)
    kline_file_path = os.path.join(base_dir, "ESC", filename)

    try:
        df_kline_raw = pd.read_csv(kline_file_path)
        df_fundflow_raw = pd.read_csv(fundflow_file_path)

        if df_kline_raw.empty or df_fundflow_raw.empty:
            print(f"⚠️ 未能加载 {code} 的部分或全部数据，跳过。")
            return None, None, None, None

        df_kline = standardize_columns(df_kline_raw.copy(), type='kline')
        df_kline["Date"] = pd.to_datetime(df_kline["Date"])

        df_fundflow = df_fundflow_raw.copy()
        df_fundflow['Date'] = pd.to_datetime(df_fundflow['日期'].astype(str) + ' ' + df_fundflow['时间'])
        df_fundflow = standardize_columns(df_fundflow, type='fundflow')

        df_merged = pd.merge(df_kline, df_fundflow[['Date', 'MainNetInflow', 'Amount_Fundflow']], on='Date', how='inner')
        df_merged.sort_values("Date", inplace=True)
        df_merged.set_index("Date", inplace=True)

        # *** 资金流数据压缩：以万为单位四舍五入 ***
        if 'MainNetInflow' in df_merged.columns:
            df_merged['MainNetInflow'] = (df_merged['MainNetInflow'] / 10000).round(0)
        if 'Amount_Fundflow' in df_merged.columns:
            df_merged['Amount_Fundflow'] = (df_merged['Amount_Fundflow'] / 10000).round(0)
        # *****************************************

        # 技术指标
        df_merged["MA5"] = df_merged["Close"].rolling(window=MA_WINDOW_5_MINUTES, min_periods=1).mean()
        df_merged["MA10"] = df_merged["Close"].rolling(window=MA_WINDOW_10_MINUTES, min_periods=1).mean()
        df_merged["EMA12"] = df_merged["Close"].ewm(span=EMA_WINDOW_12_MINUTES, adjust=False).mean()
        df_merged["EMA26"] = df_merged["Close"].ewm(span=EMA_WINDOW_26_MINUTES, adjust=False).mean()
        df_merged["MACD"] = df_merged["EMA12"] - df_merged["EMA26"]
        df_merged["RSI"] = compute_rsi(df_merged["Close"], window=RSI_WINDOW_14_MINUTES)
        df_merged["Return_1m"] = df_merged["Close"].pct_change()
        df_merged["Volatility_5m"] = df_merged["Close"].rolling(window=VOLATILITY_WINDOW_5_MINUTES, min_periods=1).std()

        if 'MainNetInflow' in df_merged.columns:
            df_merged['MainNetInflowChange'] = df_merged['MainNetInflow'].diff()

        df_merged["Hour"] = df_merged.index.hour
        df_merged["Minute"] = df_merged.index.minute
        df_merged["DayOfWeek"] = df_merged.index.dayofweek

        df_merged.dropna(inplace=True)

        features = [
            "Open","High","Low","Close","Volume",
            "MA5","MA10","EMA12","EMA26","MACD","RSI",
            "Return_1m","Volatility_5m","Hour","Minute","DayOfWeek",
            "MainNetInflow","Amount_Fundflow","MainNetInflowChange"
        ]
        actual_features = [f for f in features if f in df_merged.columns]

        if 'Close' not in actual_features:
            return None, None, None, None

        scaler = MinMaxScaler()
        df_scaled = scaler.fit_transform(df_merged[actual_features])

        return df_scaled, df_merged["Close"].values, scaler, actual_features
    except Exception as e:
        print(f"❌ 获取或处理 {code} 的分钟数据失败：{e}")
        return None, None, None, None

# 构造时间序列样本
def create_sequences(data_scaled, close_prices, window_size, pred_size):
    X, y = [], []
    # Ensure window_size and pred_size are integers for range() function and indexing
    window_size = int(window_size)
    pred_size = int(pred_size)
    if len(data_scaled) < window_size + pred_size:
        return np.array(X), np.array(y)
    for i in range(len(data_scaled) - window_size - pred_size + 1):
        X.append(data_scaled[i:i+window_size])
        y.append(close_prices[i+window_size:i+window_size+pred_size])
    return np.array(X), np.array(y)

# 构建LSTM模型
def build_lstm_model(input_shape, output_size):
    model = Sequential()
    # Recommended Keras API usage: Add Input layer explicitly
    model.add(Input(shape=input_shape))
    model.add(LSTM(64, return_sequences=True))
    model.add(LSTM(32))
    model.add(Dense(output_size))
    model.compile(optimizer='adam', loss='mse')
    return model

# 模型训练
def train_lstm_model(X, y, input_shape, output_size, model_path):
    model = build_lstm_model(input_shape, output_size=output_size)
    checkpoint = ModelCheckpoint(model_path.replace(".h5",".keras"),
                                 monitor='loss', save_best_only=True, verbose=1)
    history = model.fit(X, y, epochs=30, batch_size=16, callbacks=[checkpoint], verbose=0)
    print(f"✨ 模型训练完成，最终MSE: {history.history['loss'][-1]:.4f}")
    return model

# 预测并绘图
def predict_and_plot(model, scaled_data, scaler, window_size, stock_code, output_dir="./charts", actual_features=None):
    if actual_features is None:
        print("Error: actual_features list not provided to predict_and_plot.")
        return None

    # Ensure window_size and prediction_window_minutes are integers
    window_size = int(window_size)
    prediction_window_minutes_int = int(prediction_window_minutes)

    if len(scaled_data) < window_size:
        print(f"Warning: Not enough data for prediction for {stock_code}. Need at least {window_size} data points, but got {len(scaled_data)}.")
        return None

    last_window = scaled_data[-window_size:]
    prediction = model.predict(np.expand_dims(last_window, axis=0))[0]

    # 创建一个虚拟数组用于逆缩放，其形状为 (prediction_window_minutes_int, len(actual_features))
    dummy = np.zeros((prediction_window_minutes_int, len(actual_features)))

    # 找到 "Close" 列在 actual_features 中的索引
    try:
        close_index = actual_features.index("Close")
    except ValueError:
        print("Error: 'Close' not found in actual_features list.")
        return None

    # 用最后已知的值填充非预测特征，用预测值填充 'Close' 列
    for i in range(prediction_window_minutes_int):
        dummy[i, :] = scaled_data[-1, :]  # 复制最后一个观测的特征集
        dummy[i, close_index] = prediction[i]  # 替换为预测的收盘价

    recovered = scaler.inverse_transform(dummy)[:, close_index]  # 逆缩放并获取收盘价

    # 创建输出目录
    os.makedirs(output_dir, exist_ok=True)
    chart_path = os.path.join(output_dir, f"{stock_code}_{datetime.today().strftime('%Y-%m-%d_%H%M%S')}_minute_forecast.png")

    plt.figure(figsize=(12, 6))
    plt.plot(range(1, prediction_window_minutes_int + 1), recovered, marker='o', linestyle='-', label="预测收盘价")
    plt.title(f"{stock_code} - 未来 {prediction_window_minutes_int} 分钟收盘价预测")
    plt.xlabel("未来分钟数 (每步15分钟)")
    plt.ylabel("收盘价")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(chart_path)
    plt.close()

    print(f"✅ 图表已保存：{chart_path}")
    return recovered

# 打包所有图表为 zip 文件
def zip_today_charts(chart_dir="./charts", zip_name="charts_output.zip"):
    today_date_str = datetime.today().strftime("%Y-%m-%d")
    zip_file_path = os.path.join(base_dir, zip_name)
    with zipfile.ZipFile(zip_file_path, 'w') as zipf:
        for file in os.listdir(chart_dir):
            if file.endswith("minute_forecast.png") and today_date_str in file:
                zipf.write(os.path.join(chart_dir, file), arcname=file)
    print(f"📦 今日图表已打包为：{zip_file_path}")

# 主流程
scaled_stock_data = {}
for code in stock_codes:
    scaled_data, close_prices, scaler, actual_features = load_and_prepare_data(code, start_date, end_date, minute_interval="1")
    if scaled_data is not None:
        scaled_stock_data[code] = (scaled_data, close_prices, scaler, actual_features)

os.makedirs("./models", exist_ok=True)
chart_dir = "./charts"
os.makedirs(chart_dir, exist_ok=True)

for code in stock_codes:
    if code not in scaled_stock_data:
        continue

    scaled_data, close_prices, scaler, actual_features = scaled_stock_data[code]
    X, y = create_sequences(scaled_data, close_prices, window_size=input_window_minutes, pred_size=prediction_window_minutes)

    if X.shape[0] == 0:
        print(f"⚠️ 样本数量不足，跳过训练：{code}")
        continue

    model_path = f"./models/{code}_lstm_minute.h5"
    model = train_lstm_model(X, y, input_shape=X.shape[1:], output_size=prediction_window_minutes, model_path=model_path)
    predict_and_plot(model, scaled_data, scaler, window_size=input_window_minutes, stock_code=code, output_dir=chart_dir, actual_features=actual_features)

zip_today_charts(chart_dir)
````
