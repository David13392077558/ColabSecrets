```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, MultiHeadAttention, LayerNormalization, Dense, Dropout, TimeDistributed, Lambda
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.metrics import MeanAbsoluteError, BinaryAccuracy
from keras.callbacks import ModelCheckpoint
from tqdm.keras import TqdmCallback
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib

# 设置matplotlib支持中文
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False


# 设置回测时间区间（最近90天） - 使用与之前一致的日期范围
backtest_end = datetime.today()
backtest_start = backtest_end - timedelta(days=120) # 与之前的设置一致
start_date = backtest_start.strftime("%Y-%m-%d")
end_date = backtest_end.strftime("%Y-%m-%d")

# 股票列表
stock_codes = ['601138'] # 使用与之前一致的股票代码

# 回测窗口组合
window_sizes_to_try = [15, 30, 45] # 使用与之前一致的窗口大小
pred_sizes_to_try = [5, 5, 5]   # 使用与之前一致的预测大小

# 辅助函数：标准化列名
def standardize_columns(df):
    column_mapping = {
        "日期": "Date", "股票代码": "Code", "开盘": "Open", "收盘": "Close",
        "最高": "High", "最低": "Low", "成交量": "Volume", "成交额": "Amount",
        "振幅": "Amplitude", "涨跌幅": "ChangePct", "涨跌额": "ChangeAmt", "换手率": "Turnover",
    }
    valid_column_mapping = {old: new for old, new in column_mapping.items() if old in df.columns}
    df.rename(columns=valid_column_mapping, inplace=True)
    return df

# 辅助函数：计算RSI
def compute_rsi(series, window=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0).rolling(window).mean()
    loss = -delta.where(delta < 0, 0).rolling(window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# 数据加载和特征处理函数（合并了之前的 load_daily_and_fundflow 和 process_features 的逻辑）
def load_and_process_data(code, start_date, end_date):
    prefix = code[-6:]
    market_prefix = "sz" if code.startswith("0") else "sh"
    daily_path = f"/content/drive/MyDrive/Colab Notebooks/kline/{prefix}_daily.csv"
    fundflow_path = f"/content/drive/MyDrive/Colab Notebooks/jiqi_do/jiqi_do/{market_prefix}{prefix}.csv"

    if not os.path.exists(daily_path):
        print(f"⚠️ 缺失日线数据文件：{daily_path}")
        return None, None, None, None, None, None

    try:
        df = pd.read_csv(daily_path)
        df = standardize_columns(df.copy())
        df["Date"] = pd.to_datetime(df["Date"])
        df.set_index("Date", inplace=True)
        df.sort_index(inplace=True)
        df = df[~df.index.duplicated(keep='last')]

        if os.path.exists(fundflow_path):
            df_fund = pd.read_csv(fundflow_path)
            df_fund.rename(columns={"opendate": "Date", "netamount": "主力净流入", "ratioamount": "主力净流入比率"}, inplace=True)
            df_fund.drop(columns=[col for col in ["turnover", "大单净流入", "超大单净流入"] if col in df_fund.columns], inplace=True)
            df_fund["Date"] = pd.to_datetime(df_fund["Date"])
            df_fund.set_index("Date", inplace=True)
            df = df.join(df_fund, how="left")
            print(f"💰 资金流数据合并成功：{code}")
        else:
            print(f"⚠️ 缺失资金流文件：{fundflow_path}")

        # 确保资金流列存在，即使文件缺失
        for col in ["主力净流入", "主力净流入比率"]:
            if col not in df.columns:
                df[col] = 0

        # 切片到回测时间区间
        df = df.loc[pd.to_datetime(start_date):pd.to_datetime(end_date)].copy()

        if df.empty:
            print(f"⚠️ 切片后数据为空：{code}")
            return None, None, None, None, None, None

        # 特征工程
        df["涨跌幅"] = df.get("ChangePct", 0) # 确保有涨跌幅列
        df["涨跌方向"] = (df["涨跌幅"] > 0).astype(int) # 计算涨跌方向

        # 计算技术指标，处理NaN
        df["MA5"] = df["Close"].rolling(5).mean()
        df["MA10"] = df["Close"].rolling(10).mean()
        df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
        df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
        df["MACD"] = df["EMA12"] - df["EMA26"]
        df["RSI"] = compute_rsi(df["Close"])
        df["Return_1d"] = df["Close"].pct_change()
        df["Volatility_5d"] = df["Close"].rolling(5).std()
        df["Weekday"] = df.index.weekday
        df["IsMonthEnd"] = df.index.is_month_end.astype(int)
        df["MoneyFlow"] = df.apply(lambda row: row["Amount"] / row["Volume"] if row["Volume"] != 0 else 0, axis=1)

        # 定义特征和目标列
        features = [
            "Open", "High", "Low", "Close", "Volume", "Amount", "Turnover",
            "MA5", "MA10", "EMA12", "EMA26", "MACD", "RSI",
            "Return_1d", "Volatility_5d", "Weekday", "IsMonthEnd", "MoneyFlow"
        ]
        fundflow_features = ["主力净流入", "主力净流入比率"]
        features.extend([col for col in fundflow_features if col in df.columns])

        targets_regression_names = [t for t in ["Close", "主力净流入", "涨跌幅", "ChangePct"] if t in df.columns]
        target_classification_name = "涨跌方向"

        # 移除包含NaN的行（在计算技术指标后）
        initial_rows = len(df)
        df.dropna(subset=features + targets_regression_names + [target_classification_name], inplace=True)
        if len(df) < initial_rows:
            print(f"⚠️ 由于NaN值移除了 {initial_rows - len(df)} 行数据：{code}")


        if df.empty:
            print(f"⚠️ 移除NaN后数据为空：{code}")
            return None, None, None, None, None, None

        # 确保所有必需的列都存在于df中
        required_cols = features + targets_regression_names + [target_classification_name]
        if not all(col in df.columns for col in required_cols):
            missing = [col for col in required_cols if col not in df.columns]
            print(f"❌ 数据处理后缺少必需的列：{missing}，跳过股票 {code}")
            return None, None, None, None, None, None

        # 标准化数据
        cols_to_scale = [col for col in features + targets_regression_names if col in df.columns and df[col].dtype in ['int64', 'float64']]
        scaler = MinMaxScaler()
        df_scaled = scaler.fit_transform(df[cols_to_scale])
        df_scaled = pd.DataFrame(df_scaled, index=df.index, columns=cols_to_scale) # 转回DataFrame保留索引和列名

        # 提取目标数据
        target_regression_data = df[targets_regression_names].values
        target_classification_data = df[target_classification_name].values

        print(f"✅ 股票 {code} 数据加载和处理完成。数据长度: {len(df)}")

        return df, df_scaled, target_regression_data, target_classification_data, scaler, targets_regression_names

    except Exception as e:
        print(f"❌ 数据加载或处理失败：{code} → {e}")
        return None, None, None, None, None, None

# 构建Transformer模型
def build_transformer_model(input_shape, num_regression_targets, pred_size=15):
    inputs = Input(shape=input_shape)
    projected_inputs = Dense(64)(inputs)
    attn = MultiHeadAttention(num_heads=4, key_dim=32)(projected_inputs, projected_inputs)
    attn_out = LayerNormalization()(attn + projected_inputs)
    x = Dense(64, activation='relu')(attn_out)
    x = Dropout(0.2)(x)
    x = LayerNormalization()(x + attn_out)

    # 回归输出层
    regression_output = TimeDistributed(Dense(num_regression_targets))(x)
    # 确保回归输出只包含预测窗口的部分
    regression_output = Lambda(lambda t: t[:, -pred_size:, :], name="regression_output")(regression_output)


    # 分类输出层
    classification_output = TimeDistributed(Dense(1, activation='sigmoid'))(x)
    # 确保分类输出只包含预测窗口的部分
    classification_output = Lambda(lambda t: t[:, -pred_size:, :], name="classification_output")(classification_output)


    model = Model(inputs=inputs, outputs=[regression_output, classification_output])
    model.compile(optimizer=Adam(),
                  loss={"regression_output": MeanSquaredError(), "classification_output": BinaryCrossentropy()},
                  metrics={"regression_output": MeanAbsoluteError(), "classification_output": BinaryAccuracy()})
    return model

# 滚动回测函数（重构逻辑）
def run_rolling_backtest(stock_code, df, df_scaled, target_regression_data, target_classification_data,
                         scaler, actual_regression_target_names, window_size, pred_size,
                         model_dir="./models", output_dir="./charts", epochs=20): # 减少epochs进行测试
    """
    Runs a true rolling window backtest for a given stock.

    Args:
        stock_code (str): The stock code.
        df (pd.DataFrame): Original DataFrame with dates as index.
        df_scaled (pd.DataFrame): Scaled DataFrame with dates as index.
        target_regression_data (np.ndarray): Array of regression targets.
        target_classification_data (np.ndarray): Array of classification targets.
        scaler (MinMaxScaler): The scaler used for standardization.
        actual_regression_target_names (list): List of regression target column names.
        window_size (int): The size of the look-back window for training.
        pred_size (int): The size of the prediction window.
        model_dir (str): Directory to save/load models.
        output_dir (str): Directory to save charts and results.
        epochs (int): Number of training epochs per step.

    Returns:
        pd.DataFrame: DataFrame containing collected predicted and actual values and dates for the backtest period.
                      Returns None if backtest cannot be performed.
    """
    # 确保数据长度足够进行第一次训练和预测
    if len(df_scaled) < window_size + pred_size:
        print(f"⚠️ 数据不足以进行回测 ({window_size} + {pred_size} = {window_size + pred_size} 条数据所需)，跳过回测：{stock_code}")
        return None

    # 回测从第一个可以预测的日期开始
    # 这个日期是训练窗口结束后的第一天
    # 例如，如果训练窗口是30天，预测5天，那么第一个预测开始日期是数据的第31天
    backtest_start_index = window_size
    # 回测在最后一个可以预测完整 pred_size 天的日期结束
    backtest_end_index = len(df_scaled) - pred_size

    if backtest_start_index > backtest_end_index:
         print(f"⚠️ 回测数据范围不足，无法形成完整的预测步。至少需要 {window_size + pred_size} 天数据。")
         return None


    print(f"🏃‍♂️ 开始对股票 {stock_code} 进行滚动回测...")
    print(f"回测数据总长度: {len(df_scaled)}")
    print(f"回测步数 (预测开始日期数量): {backtest_end_index - backtest_start_index + 1}")
    print(f"回测数据范围 (基于索引): {backtest_start_index} to {backtest_end_index}")
    print(f"回测数据日期范围: {df.index[backtest_start_index].strftime('%Y-%m-%d')} to {df.index[backtest_end_index].strftime('%Y-%m-%d')}")


    all_predicted_regression = []
    all_actual_regression = []
    all_predicted_classification = []
    all_actual_classification = []
    all_dates = []

    # 滚动窗口迭代
    # current_pred_start_index 是当前预测窗口的开始索引
    for current_pred_start_index in range(backtest_start_index, backtest_end_index + 1):
        # 训练窗口结束索引是预测开始索引的前一天
        train_end_index = current_pred_start_index - 1
        train_start_index = train_end_index - window_size + 1 # 训练窗口开始索引

        # 预测窗口结束索引
        pred_end_index = current_pred_start_index + pred_size

        # 获取当前训练窗口的数据和目标
        X_train_window = df_scaled.iloc[train_start_index : train_end_index + 1].values
        y_train_regression_window = target_regression_data[current_pred_start_index : pred_end_index]
        y_train_classification_window = target_classification_data[current_pred_start_index : pred_end_index]


        # 检查训练和目标数据形状是否正确
        if X_train_window.shape[0] != window_size or \
           y_train_regression_window.shape[0] != pred_size or \
           y_train_classification_window.shape[0] != pred_size:
             print(f"❌ 滚动窗口数据切片错误，跳过当前回测步：{df.index[current_pred_start_index].strftime('%Y-%m-%d')}")
             print(f"X_train_window shape: {X_train_window.shape}, Expected: ({window_size}, {df_scaled.shape[1]})")
             print(f"y_train_regression_window shape: {y_train_regression_window.shape}, Expected: ({pred_size}, {target_regression_data.shape[1]})")
             print(f"y_train_classification_window shape: {y_train_classification_window.shape}, Expected: ({pred_size},)")
             continue

        # 调整目标数据的维度以匹配模型输出
        y_train_regression_window = np.expand_dims(y_train_regression_window, axis=0) # 增加 batch 维度
        y_train_classification_window = np.expand_dims(y_train_classification_window, axis=0) # 增加 batch 维度
        if y_train_classification_window.ndim == 2: # 确保分类目标有最后一维
             y_train_classification_window = np.expand_dims(y_train_classification_window, axis=-1)


        # 构建或加载模型
        # 在滚动回测中，通常每个步都训练一个新模型或微调旧模型
        # 为了简化和测试，这里每个步都构建一个新模型并训练少量epochs
        model = build_transformer_model((window_size, df_scaled.shape[1]), target_regression_data.shape[1], pred_size)

        # 训练模型
        print(f"🔄 正在训练模型，回测步预测开始日期: {df.index[current_pred_start_index].strftime('%Y-%m-%d')}")
        try:
            # 训练数据形状: (1, window_size, num_features)
            # 训练目标形状: (1, pred_size, num_regression_targets), (1, pred_size, 1)
            history = model.fit(np.expand_dims(X_train_window, axis=0), # 增加 batch 维度
                                {"regression_output": y_train_regression_window, "classification_output": y_train_classification_window},
                                epochs=epochs,
                                batch_size=1, # 每个回测步使用一个样本（即当前窗口数据）进行训练
                                verbose=0)
            # print(f"✅ 模型训练完成，Loss: {history.history['loss'][-1]}")

        except Exception as e:
             print(f"❌ 模型训练失败，跳过当前回测步：{df.index[current_pred_start_index].strftime('%Y-%m-%d')} → {e}")
             continue

        # 进行预测
        # 预测输入是当前训练窗口的数据
        predict_data_scaled = X_train_window # 使用训练窗口的数据进行预测
        if predict_data_scaled.shape != (window_size, df_scaled.shape[1]):
             print(f"❌ 预测输入数据维度错误，跳过预测：{df.index[current_pred_start_index].strftime('%Y-%m-%d')}")
             continue

        try:
            predictions = model.predict(np.expand_dims(predict_data_scaled, axis=0), verbose=0) # Predict on a single sample
            regression_prediction = predictions[0][0] # Shape: (pred_size, num_regression_targets)
            classification_prediction = predictions[1][0] # Shape: (pred_size, 1)

            # 确保预测结果形状正确
            if regression_prediction.shape != (pred_size, target_regression_data.shape[1]) or \
               classification_prediction.shape != (pred_size, 1):
                print(f"❌ 预测输出形状错误，跳过当前回测步：{df.index[current_pred_start_index].strftime('%Y-%m-%d')}")
                print(f"Regression prediction shape: {regression_prediction.shape}, Expected: ({pred_size}, {target_regression_data.shape[1]})")
                print(f"Classification prediction shape: {classification_prediction.shape}, Expected: ({pred_size}, 1)")
                continue

        except Exception as e:
            print(f"❌ 模型预测失败，跳过当前回测步：{df.index[current_pred_start_index].strftime('%Y-%m-%d')} → {e}")
            continue

        # 获取当前预测窗口的实际数据和日期
        actual_regression_window = target_regression_data[current_pred_start_index : pred_end_index]
        actual_classification_window = target_classification_data[current_pred_start_index : pred_end_index]
        prediction_dates = df.index[current_pred_start_index : pred_end_index]

        # 存储当前预测步的结果
        # 这里我们存储的是当前预测窗口的完整预测结果和实际结果
        # 后续需要处理重叠日期的问题
        all_predicted_regression.append(regression_prediction)
        all_actual_regression.append(actual_regression_window)
        all_predicted_classification.append(classification_prediction.flatten()) # 存储概率或预测类别
        all_actual_classification.append(actual_classification_window.flatten())
        all_dates.append(prediction_dates.tolist()) # 存储日期列表

    # 将所有预测步的结果展平并合并，处理重叠日期
    # 由于是滚动回测，同一个日期可能被多个预测窗口预测到
    # 简单的处理方法是保留最后一次预测的结果
    combined_predicted_regression = []
    combined_actual_regression = []
    combined_predicted_classification = []
    combined_actual_classification = []
    combined_dates = []

    # 展平列表
    for i in range(len(all_dates)):
        combined_dates.extend(all_dates[i])
        combined_predicted_regression.extend(all_predicted_regression[i])
        combined_actual_regression.extend(all_actual_regression[i])
        combined_predicted_classification.extend(all_predicted_classification[i])
        combined_actual_classification.extend(all_actual_classification[i])

    if not combined_dates:
        print(f"⚠️ 未生成任何回测结果，跳过结果保存：{stock_code}")
        return None

    # 创建一个临时 DataFrame 来处理重叠日期，保留每个日期的最后一次预测
    temp_results_df = pd.DataFrame({
        "Date": combined_dates,
        "Predicted_Regression": list(combined_predicted_regression),
        "Actual_Regression": list(combined_actual_regression),
        "Predicted_Classification": combined_predicted_classification, # 存储概率
        "Actual_Classification": combined_actual_classification
    })

    temp_results_df["Date"] = pd.to_datetime(temp_results_df["Date"])
    temp_results_df.set_index("Date", inplace=True)
    temp_results_df.sort_index(inplace=True)

    # 处理重复日期，保留最后一次预测
    results_df = temp_results_df[~temp_results_df.index.duplicated(keep='last')].copy()

    # 将存储为列表的回归结果转换回 numpy 数组或适当的格式
    # 这里假设 Predicted_Regression 和 Actual_Regression 中的每个元素都是一个包含 num_regression_targets 个元素的列表
    num_regression_targets = target_regression_data.shape[-1]
    try:
        results_df["Predicted_Regression"] = results_df["Predicted_Regression"].apply(lambda x: np.array(x).flatten())
        results_df["Actual_Regression"] = results_df["Actual_Regression"].apply(lambda x: np.array(x).flatten())
    except Exception as e:
         print(f"❌ 回归结果格式转换失败：{e}")
         # 可以选择跳过保存或返回 None
         # return None # 如果转换失败则不返回结果

    # 添加预测类别列 (0或1)
    results_df["Predicted_Direction"] = (results_df["Predicted_Classification"] > 0.5).astype(int)


    print(f"✅ 滚动回测对股票 {stock_code} 完成。生成 {len(results_df)} 条回测结果。")

    return results_df

# ----- 主回测流程 -----
debug_mode = False # 保持与之前一致
all_backtest_results = {} # 存储所有股票和组合的回测结果

stocks_to_process = stock_codes if not debug_mode else [stock_codes[0]]

# 1. 加载和处理所有股票的数据
print("--- 正在加载和处理所有股票数据 ---")
processed_stocks_data = {}
for code in stocks_to_process:
    print(f"处理股票: {code}")
    df, df_scaled, target_regression_data, target_classification_data, scaler, actual_regression_target_names = load_and_process_data(
        code, start_date, end_date
    )

    # 检查数据是否成功加载和处理
    if df is not None and not df.empty and \
       df_scaled is not None and not df_scaled.empty and \
       target_regression_data is not None and target_regression_data.shape[0] > 0 and \
       target_classification_data is not None and target_classification_data.shape[0] > 0 and \
       scaler is not None and \
       actual_regression_target_names is not None and len(actual_regression_target_names) > 0:

        processed_stocks_data[code] = {
            "df": df,
            "df_scaled": df_scaled,
            "target_regression_data": target_regression_data,
            "target_classification_data": target_classification_data,
            "scaler": scaler,
            "actual_regression_target_names": actual_regression_target_names
        }
        print(f"✅ 股票 {code} 数据准备成功。数据长度: {len(df)}")
    else:
        print(f"❌ 股票 {code} 数据准备失败，跳过该股票。")


# 2. 对每个股票和窗口组合运行滚动回测
print("\n--- 正在对股票和窗口组合运行滚动回测 ---")

for code in processed_stocks_data:
    stock_data = processed_stocks_data[code]
    df = stock_data["df"]
    df_scaled = stock_data["df_scaled"]
    target_regression_data = stock_data["target_regression_data"]
    target_classification_data = stock_data["target_classification_data"]
    scaler = stock_data["scaler"]
    actual_regression_target_names = stock_data["actual_regression_target_names"]

    print(f"\n--- 对股票 {code} 进行窗口组合测试 ---")

    for current_window_size in window_sizes_to_try:
        for current_pred_size in pred_sizes_to_try:
            print(f"\n--- 测试组合: 回溯窗口={current_window_size}, 预测窗口={current_pred_size} ---")

            # 运行滚动回测
            backtest_results_df = run_rolling_backtest(
                stock_code=code,
                df=df,
                df_scaled=df_scaled,
                target_regression_data=target_regression_data,
                target_classification_data=target_classification_data,
                scaler=scaler,
                actual_regression_target_names=actual_regression_target_names,
                window_size=current_window_size,
                pred_size=current_pred_size,
                model_dir="./models",
                output_dir=f"./charts/window_{current_window_size}_pred_{current_pred_size}",
                epochs=20 # 可以根据需要调整epochs
            )

            if backtest_results_df is not None and not backtest_results_df.empty:
                # 存储回测结果
                combination_key = f"{code}_window_{current_window_size}_pred_{current_pred_size}"
                all_backtest_results[combination_key] = backtest_results_df

                # 保存原始回测结果到 CSV
                output_dir_comb = f"./charts/window_{current_window_size}_pred_{current_pred_size}"
                os.makedirs(output_dir_comb, exist_ok=True)
                results_path = os.path.join(output_dir_comb, f"{code}_backtest_results.csv")
                backtest_results_df.to_csv(results_path)
                print(f"✅ 回测结果CSV已保存：{results_path}")
            else:
                print(f"⚠️ 组合 回溯窗口={current_window_size}, 预测窗口={current_pred_size} 未生成有效回测结果。")


# 3. 回测任务完成，all_backtest_results 字典包含了所有成功的回测结果
print("\n--- 所有回测任务完成 ---")

# 后续步骤：分析 all_backtest_results 字典，计算指标，可视化等。
# 例如，您可以遍历 all_backtest_results 计算每个组合的 MSE, MAE, Accuracy 等指标。
# 您也可以为每个组合绘制预测值和实际值的对比图。
```
