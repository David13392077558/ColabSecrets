```python
import pandas as pd

# 加载 Excel 文件，并指定表头在第三行（索引为2）
excel_file_path = "/content/05汇总表(1).xls"
df = pd.read_excel(excel_file_path, header=2)

# 显示数据的前几行和列名，以帮助识别姓名列
display(df.head())
display(df.columns)
```

```python
# Identify the column containing names - based on previous inspection, it's '姓名'
# 根据之前的检查，确定包含姓名的列是 '姓名'
name_column = '姓名'

# Replace '谭剑仪' with '梁子羿' in the '姓名' column
# 在 '姓名' 列中，将 '谭剑仪' 替换为 '梁子羿'
df[name_column] = df[name_column].replace('谭剑仪', '梁子羿')

# Display the rows where the name was modified to confirm the change
# 显示被修改姓名的行，以确认修改成功
display(df[df[name_column] == '梁子羿'])
```

```python
# Save the modified DataFrame to a new Excel file
# 将修改后的 DataFrame 保存到新的 Excel 文件
output_file_path = "/content/05汇总表(1)_modified.xlsx"
df.to_excel(output_file_path, index=False)
display(df.head())
display(df.columns)
```
