```python
import yfinance as yf
import numpy as np

def get_latest_close_prices(ticker='600362.SS', days=60):
    """
    获取指定股票最近 N 天的收盘价，用于预测模型输入。
    ticker: 股票代码（如 '600362.SS' 表示江西铜业）
    days: 需要的收盘价天数
    """
    print(f"正在下载股票 {ticker} 的最近 {days+20} 天数据...")
    data = yf.download(ticker, period=f'{days+20}d')

    if data.empty:
        raise ValueError("下载失败，未获取到任何数据。请检查 ticker 是否正确。")

    print(f"成功获取数据，共 {len(data)} 条记录。")
    print("原始收盘价如下（最近几天）：")
    print(data['Close'].tail(5))

    close_prices = data['Close'].dropna().values[-days:]

    if len(close_prices) < days:
        raise ValueError(f"收盘价不足 {days} 天，仅获取到 {len(close_prices)} 天")

    print(f"已提取最近 {days} 天的收盘价，用于模型输入。")
    print("最终用于模型的收盘价数组（形状）：", close_prices.reshape(-1, 1).shape)
    print("收盘价预览：", close_prices[-5:])

    return close_prices.reshape(-1, 1)  # reshape 为 LSTM 输入格式

```

# Task
Predict the next 15 days of stock prices using the loaded model "best_model.h5" and visualize the results.

## Prepare input data

### Subtask:
Use the existing `get_latest_close_prices` function to get the most recent 60 days of closing prices. Scale this data using the `MinMaxScaler`.


**Reasoning**:
I will call the `get_latest_close_prices` function, instantiate `MinMaxScaler`, fit and transform the data, and reshape it for model input, following the instructions provided in the subtask.



```python
# Get the latest 60 days of closing prices for the specified ticker.
close_prices = get_latest_close_prices('600362.SS', days=60)

# Instantiate the MinMaxScaler.
scaler = MinMaxScaler()

# Fit the scaler to the data and transform the data.
scaled_prices = scaler.fit_transform(close_prices)

# Reshape the scaled data for model input: (number of samples, time steps, features).
X_input = scaled_prices.reshape(1, 60, 1)

print("Shape of the input data for the model:", X_input.shape)
```

## Load the model

### Subtask:
Load the trained LSTM model (`best_model.h5`). Note that based on previous errors, this model likely expects input with 4 features, so this step might require adjustment or using a different model if you only have closing prices.


**Reasoning**:
Load the trained LSTM model from the specified file path, including the custom object for the loss function.



```python
model = load_model('sample_data/best_model.h5', custom_objects={'mse': tf.keras.losses.MeanSquaredError()})
print("Model loaded successfully.")
```

## Predict future prices

### Subtask:
Use the loaded model to predict the next 15 days' prices. This will involve an iterative process where each prediction is used as part of the input for the next prediction.


**Reasoning**:
Implement the iterative prediction process as described in the instructions to predict the next 15 days of stock prices.



```python
# Initialize an empty list to store the predicted future scaled prices.
predicted_scaled_prices = []

# Use the scaled_prices array as the initial input sequence for the model.
# The input sequence needs to have the shape (batch_size, time_steps, features)
# Based on the error, the model seems to expect 4 features.
# We will duplicate the single feature across 4 dimensions as a workaround.
current_input_sequence = np.repeat(scaled_prices.reshape(1, 60, 1), 4, axis=2)


# Loop 15 times to predict each of the next 15 days.
for _ in range(15):
    # Predict the next day's scaled price.
    predicted_scaled_next_day_output = model.predict(current_input_sequence)

    # Assuming the model output is (1, 1) and represents a single feature prediction.
    # Extract the predicted scaled price.
    predicted_scaled_next_day = predicted_scaled_next_day_output[0][0]


    # Append the predicted scaled price to the list of predictions.
    predicted_scaled_prices.append(predicted_scaled_next_day)

    # Update the input sequence for the next iteration.
    # Remove the oldest data point and add the newly predicted scaled price.
    # Ensure the sequence maintains the shape (1, 60, 4).
    # Duplicate the predicted scaled price across the 4 feature dimensions when appending.
    new_data_point = np.repeat(predicted_scaled_next_day, 4).reshape(1, 1, 4)
    current_input_sequence = np.append(current_input_sequence[:, 1:, :], new_data_point, axis=1)


# The list predicted_scaled_prices now contains the 15 predicted scaled prices.
print("Predicted scaled prices for the next 15 days:", predicted_scaled_prices)
```

## Predict future prices

### Subtask:
Retry the prediction of future prices. Use the loaded model to predict the next 15 days' prices. This will involve an iterative process where each prediction is used as part of the input for the next prediction. Since the previous attempt failed due to a mismatch in the number of features the model expected (4) vs. the available data (1), we need to acknowledge this and make an assumption. Assuming the model is intended for single-feature time series forecasting, we will proceed by reshaping the input to match the model's expected feature dimension, even though the number of features is mismatched. **This is a workaround based on the available information and may not yield meaningful results if the model was genuinely trained on multiple features.**


**Reasoning**:
Initialize a list to store predictions and reshape the input to match the assumed model input shape, duplicating the single feature across the 4 required dimensions. Then, start the loop to predict the next 15 days, updating the input sequence in each iteration by removing the oldest data point and appending the new prediction, duplicated across the 4 features.



```python
# 1. 初始化一个空列表 predicted_scaled_prices 来存储缩放后的预测结果。
predicted_scaled_prices = []

# 2. 将初始的 scaled_prices 数据重塑以匹配模型假设的输入形状 (1, 60, 4)。
# 作为临时解决方案，将单一特征沿着第4维重复4次。
# 在重塑之前检查 scaled_prices 是否具有期望的形状 (60, 1)。
if scaled_prices.shape != (60, 1):
    raise ValueError(f"期望 scaled_prices 的形状为 (60, 1)，但得到的是 {scaled_prices.shape}")

# 通过沿最后一个轴将单一特征重复4次来重塑为 (1, 60, 4)。
current_input_sequence = np.repeat(scaled_prices.reshape(1, 60, 1), 4, axis=2)

print("重塑后初始输入序列的形状:", current_input_sequence.shape)

# 3. 启动一个循环，运行15次以预测接下来的15天。
for i in range(15):
    print(f"正在预测第 {i+1} 天...")
    # 4. 在循环内部，使用 model.predict() 方法并传入当前输入序列。
    predicted_scaled_next_day_output = model.predict(current_input_sequence)

    # 假设模型输出为 (1, 1)，表示单一特征预测。
    # 提取预测的缩放后的价格。
    predicted_scaled_next_day = predicted_scaled_next_day_output[0][0]

    # 5. 从预测输出中提取预测的缩放后的价格，并将其添加到 predicted_scaled_prices 列表中。
    predicted_scaled_prices.append(predicted_scaled_next_day)

    # 6. 更新下一轮迭代的输入序列：移除最旧的数据点并追加新预测的缩放后的价格。
    # 确保序列保持形状 (1, 60, 4)。
    # 在追加时将预测的缩放后的价格沿着4个特征维度重复。
    new_data_point = np.repeat(predicted_scaled_next_day, 4).reshape(1, 1, 4)
    current_input_sequence = np.append(current_input_sequence[:, 1:, :], new_data_point, axis=1)

    print(f"预测第 {i+1} 天后输入序列的形状:", current_input_sequence.shape)

# 7. 循环结束后，predicted_scaled_prices 列表将包含15个预测的缩放后的值。
print("\n接下来15天的预测缩放后价格:", predicted_scaled_prices)
print("预测缩放后价格的数量:", len(predicted_scaled_prices))
```

## Inverse transform predictions

### Subtask:
Inverse transform the scaled predictions back to the original price scale.


**Reasoning**:
Convert the list of scaled predictions into a NumPy array and reshape it for inverse transformation.



```python
# Create a NumPy array from the list of predicted scaled prices.
predicted_scaled_prices_array = np.array(predicted_scaled_prices)

# Reshape the array to (15, 1) for inverse transformation.
predicted_scaled_prices_reshaped = predicted_scaled_prices_array.reshape(-1, 1)

print("Shape of reshaped predicted scaled prices:", predicted_scaled_prices_reshaped.shape)
```

**Reasoning**:
Apply the inverse transform to the reshaped scaled predictions to get the prices in the original scale and print the results.



```python
# Use the fitted scaler to inverse transform the reshaped scaled prices.
predicted_prices = scaler.inverse_transform(predicted_scaled_prices_reshaped)

# Print the predicted prices in their original scale.
print("Predicted stock prices for the next 15 days (original scale):")
print(predicted_prices)
```
