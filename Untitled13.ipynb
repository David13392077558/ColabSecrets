```python
# 加载你的数据（假设 CSV 文件中包含 Date 和 Close 列）
df = pd.read_csv('cleaned_stock_data.csv', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# 计算 MACD 和信号线
def compute_macd(df, fast=12, slow=26, signal=9):
    df['EMA_fast'] = df['Close'].ewm(span=fast, adjust=False).mean()
    df['EMA_slow'] = df['Close'].ewm(span=slow, adjust=False).mean()
    df['MACD'] = df['EMA_fast'] - df['EMA_slow']
    df['MACD_signal'] = df['MACD'].ewm(span=signal, adjust=False).mean()
    return df

# 计算 RSI 指标
def compute_rsi(df, period=14):
    delta = df['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    return df

# 添加技术指标
def add_technical_indicators(df):
    df = compute_macd(df)
    df = compute_rsi(df)
    df.dropna(inplace=True)
    return df

# 应用技术指标
df = add_technical_indicators(df)
```

```python
# 添加周期性时间特征（例如：星期几、月份）
df['DayOfWeek'] = df['Date'].dt.dayofweek  # 周一=0，周日=6
df['Month'] = df['Date'].dt.month

# 转为周期性编码（sin/cos）
df['Day_sin'] = np.sin(2 * np.pi * df['DayOfWeek'] / 7)
df['Day_cos'] = np.cos(2 * np.pi * df['DayOfWeek'] / 7)
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
```

```python
# 设置窗口大小和预测偏移
window_size = 60         # 用过去 60 天的数据预测
target_shift = 1         # 预测未来第 1 天的收盘价

# 选择用于训练的特征（你已经更新过了）
features = ['Close', 'MACD', 'MACD_signal', 'RSI', 'Day_sin', 'Day_cos', 'Month_sin', 'Month_cos']

# 归一化处理
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features])

# 构建训练数据集
X, y = [], []
for i in range(window_size, len(scaled_data) - target_shift):
    X.append(scaled_data[i - window_size:i])
    y.append(scaled_data[i + target_shift][0])  # 预测收盘价

X, y = np.array(X), np.array(y)

# 划分训练集和验证集
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# 构建模型函数
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(32))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# 构建模型（确保 X_train 已定义）
model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

# 回调函数（可选）
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True)

# 模型训练
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop, checkpoint],
    verbose=1
)
```

```python
# 模型预测
y_pred = model.predict(X_val)

# 反归一化（只对 'Close' 一列）
y_val_rescaled = scaler.inverse_transform(
    np.concatenate([y_val.reshape(-1, 1), np.zeros((len(y_val), len(features)-1))], axis=1)
)[:, 0]

y_pred_rescaled = scaler.inverse_transform(
    np.concatenate([y_pred, np.zeros((len(y_pred), len(features)-1))], axis=1)
)[:, 0]
```

```python
import pandas as pd
import os

# Directory containing the individual ticker CSV files
output_dir = 'stock_data_by_ticker'

# List to hold DataFrames
all_dfs = []

# Read each CSV and append to the list
for filename in os.listdir(output_dir):
    if filename.endswith("_data.csv"):
        filepath = os.path.join(output_dir, filename)
        try:
            # Try reading assuming the first row is the header
            df = pd.read_csv(filepath)

            # Check if the first column name is 'Date' (standard) or 'Price' (seen in some outputs)
            # If it's 'Price', the actual header is likely the second row, and data starts from the third
            if df.columns[0] == 'Price':
                 # Read again, skipping the first two rows and setting the header to the third row
                df = pd.read_csv(filepath, skiprows=[0, 1], header=0)
                # Assuming the columns in this case are Date, Open, High, Low, Close, Volume
                df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
            elif df.columns[0] != 'Date':
                 # If the first column is neither 'Date' nor 'Price', something else is wrong, skip for now
                 print(f"Skipping {filename} due to unexpected column structure: {df.columns}")
                 continue

            # Extract ticker from filename (assuming format like "TICKER_data.csv")
            ticker = filename.replace("_data.csv", "")
            df['Ticker'] = ticker
            all_dfs.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {e}")

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(all_dfs, ignore_index=True)

# Convert 'Date' column to datetime and sort
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df.sort_values(by=['Ticker', 'Date'], inplace=True)

print("Combined DataFrame preview:")
print(combined_df.head())
print("\nCombined DataFrame info:")
combined_df.info()
```

```python
import numpy as np
# Rename columns for easier access
# Assuming the columns are in the order: Date, Open, High, Low, Close, Volume, Ticker, and then the intermediate columns from technical indicator calculations
# We will only rename the first 7 columns that we know the names for
combined_df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Ticker'] + list(combined_df.columns[7:])


# Convert necessary columns to numeric, coercing errors
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

# Drop rows with missing values that might have resulted from coercion or technical indicator calculations
combined_df.dropna(inplace=True)


# Calculate MACD and Signal line
def compute_macd(df, fast=12, slow=26, signal=9):
    df['EMA_fast'] = df.groupby('Ticker')['Close'].transform(lambda x: x.ewm(span=fast, adjust=False).mean())
    df['EMA_slow'] = df.groupby('Ticker')['Close'].transform(lambda x: x.ewm(span=slow, adjust=False).mean())
    df['MACD'] = df['EMA_fast'] - df['EMA_slow']
    df['MACD_signal'] = df.groupby('Ticker')['MACD'].transform(lambda x: x.ewm(span=signal, adjust=False).mean())
    return df

# Calculate RSI indicator
def compute_rsi(df, period=14):
    df['Close_diff'] = df.groupby('Ticker')['Close'].diff()
    df['gain'] = df['Close_diff'].where(df['Close_diff'] > 0, 0)
    df['loss'] = -df['Close_diff'].where(df['Close_diff'] < 0, 0)
    df['avg_gain'] = df.groupby('Ticker')['gain'].transform(lambda x: x.rolling(window=period).mean())
    df['avg_loss'] = df.groupby('Ticker')['loss'].transform(lambda x: x.rolling(window=period).mean())
    # Avoid division by zero
    df['rs'] = df['avg_gain'] / df['avg_loss'].replace(0, np.nan)
    df['RSI'] = 100 - (100 / (1 + df['rs']))
    # Drop intermediate columns
    df = df.drop(columns=['Close_diff', 'gain', 'loss', 'avg_gain', 'avg_loss', 'rs'])
    return df


# Add technical indicators
def add_technical_indicators(df):
    df = compute_macd(df)
    df = compute_rsi(df)
    df.dropna(inplace=True) # Drop rows with NaN created by rolling/ewm
    return df

# Apply technical indicators
combined_df = add_technical_indicators(combined_df)

# Add cyclical time features
combined_df['DayOfWeek'] = combined_df['Date'].dt.dayofweek  # Monday=0, Sunday=6
combined_df['Month'] = combined_df['Date'].dt.month

# Convert to cyclical features (sin/cos)
combined_df['Day_sin'] = np.sin(2 * np.pi * combined_df['DayOfWeek'] / 7)
combined_df['Day_cos'] = np.cos(2 * np.pi * combined_df['DayOfWeek'] / 7)
combined_df['Month_sin'] = np.sin(2 * np.pi * combined_df['Month'] / 12)
combined_df['Month_cos'] = np.cos(2 * np.pi * combined_df['Month'] / 12)

print("\nDataFrame with technical indicators and cyclical features:")
print(combined_df.head())
print("\nDataFrame info after preprocessing:")
combined_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define features to use for scaling and model input
features = ['Close', 'MACD', 'MACD_signal', 'RSI', 'Day_sin', 'Day_cos', 'Month_sin', 'Month_cos', 'Volume', 'Open', 'High', 'Low'] # Include all numerical features


# Prepare data for LSTM (creating sequences per ticker)
window_size = 60
target_shift = 1

X, y, X_dates, y_dates, tickers_list = [], [], [], [], []

# Process data for each ticker separately
for ticker in combined_df['Ticker'].unique():
    ticker_df = combined_df[combined_df['Ticker'] == ticker].copy()

    # Select and scale features for the current ticker
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(ticker_df[features])

    # Create sequences
    for i in range(window_size, len(scaled_data) - target_shift):
        X.append(scaled_data[i - window_size:i])
        y.append(scaled_data[i + target_shift, features.index('Close')])  # Predict the 'Close' price
        X_dates.append(ticker_df['Date'].iloc[i - window_size:i].tolist())
        y_dates.append(ticker_df['Date'].iloc[i + target_shift])
        tickers_list.append(ticker)


X, y = np.array(X), np.array(y)
X_dates, y_dates, tickers_list = np.array(X_dates), np.array(y_dates), np.array(tickers_list)


print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
print("Shape of X_dates:", X_dates.shape)
print("Shape of y_dates:", y_dates.shape)
print("Shape of tickers_list:", tickers_list.shape)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Split data while keeping sequences from the same ticker together
# We can't use a simple train_test_split with shuffle=True here
# as it would mix sequences from different tickers.
# A manual split or using GroupShuffleSplit would be more appropriate for a robust evaluation
# However, for simplicity and demonstrating the combined training, we'll use a time-based split within each ticker or a simplified split if the order is already by ticker and date.
# Assuming the data is sorted by Ticker and then Date, a simple split will keep initial tickers/dates in train and later ones in validation.

# Determine split point (e.g., 80% for training)
split_ratio = 0.8
split_index = int(len(X) * split_ratio)

X_train, X_val = X[:split_index], X[split_index:]
y_train, y_val = y[:split_index], y[split_index:]
tickers_train, tickers_val = tickers_list[:split_index], tickers_list[split_index:]
y_dates_train, y_dates_val = y_dates[:split_index], y_dates[split_index:]


print("Shape of X_train:", X_train.shape)
print("Shape of X_val:", X_val.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_val:", y_val.shape)


# Build the LSTM model
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(32))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# Build the model
model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

# Callbacks (optional)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_combined_model.h5', save_best_only=True)

# Model training
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop, checkpoint],
    verbose=1
)
```

```python
# Model prediction
y_pred = model.predict(X_val)

# Inverse scaling for evaluation
# We need a scaler that was fit on the 'Close' price specifically, or use the original scaler and extract the 'Close' column after inverse transform
# Let's refit a scaler on the 'Close' price of the entire combined_df for accurate inverse scaling
close_scaler = MinMaxScaler()
close_scaler.fit(combined_df['Close'].values.reshape(-1, 1))


y_val_rescaled = close_scaler.inverse_transform(y_val.reshape(-1, 1))
y_pred_rescaled = close_scaler.inverse_transform(y_pred)

print("Shape of y_val_rescaled:", y_val_rescaled.shape)
print("Shape of y_pred_rescaled:", y_pred_rescaled.shape)
```

```python
import numpy as np
# Rename columns for easier access
# Assuming the columns are in the order: Date, Open, High, Low, Close, Volume, Ticker, and then the intermediate columns from technical indicator calculations
# We will only rename the first 7 columns that we know the names for
combined_df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Ticker'] + list(combined_df.columns[7:])


# Convert necessary columns to numeric, coercing errors
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

# Drop rows with missing values that might have resulted from coercion or technical indicator calculations
combined_df.dropna(inplace=True)


# Calculate MACD and Signal line
def compute_macd(df, fast=12, slow=26, signal=9):
    df['EMA_fast'] = df.groupby('Ticker')['Close'].transform(lambda x: x.ewm(span=fast, adjust=False).mean())
    df['EMA_slow'] = df.groupby('Ticker')['Close'].transform(lambda x: x.ewm(span=slow, adjust=False).mean())
    df['MACD'] = df['EMA_fast'] - df['EMA_slow']
    df['MACD_signal'] = df.groupby('Ticker')['MACD'].transform(lambda x: x.ewm(span=signal, adjust=False).mean())
    return df

# Calculate RSI indicator
def compute_rsi(df, period=14):
    df['Close_diff'] = df.groupby('Ticker')['Close'].diff()
    df['gain'] = df['Close_diff'].where(df['Close_diff'] > 0, 0)
    df['loss'] = -df['Close_diff'].where(df['Close_diff'] < 0, 0)
    df['avg_gain'] = df.groupby('Ticker')['gain'].transform(lambda x: x.rolling(window=period).mean())
    df['avg_loss'] = df.groupby('Ticker')['loss'].transform(lambda x: x.rolling(window=period).mean())
    # Avoid division by zero
    df['rs'] = df['avg_gain'] / df['avg_loss'].replace(0, np.nan)
    df['RSI'] = 100 - (100 / (1 + df['rs']))
    # Drop intermediate columns
    df = df.drop(columns=['Close_diff', 'gain', 'loss', 'avg_gain', 'avg_loss', 'rs'])
    return df


# Add technical indicators
def add_technical_indicators(df):
    df = compute_macd(df)
    df = compute_rsi(df)
    df.dropna(inplace=True) # Drop rows with NaN created by rolling/ewm
    return df

# Apply technical indicators
combined_df = add_technical_indicators(combined_df)

# Add cyclical time features
combined_df['DayOfWeek'] = combined_df['Date'].dt.dayofweek  # Monday=0，周日=6
combined_df['Month'] = combined_df['Date'].dt.month

# Convert to cyclical features (sin/cos)
combined_df['Day_sin'] = np.sin(2 * np.pi * combined_df['DayOfWeek'] / 7)
combined_df['Day_cos'] = np.cos(2 * np.pi * combined_df['DayOfWeek'] / 7)
combined_df['Month_sin'] = np.sin(2 * np.pi * combined_df['Month'] / 12)
combined_df['Month_cos'] = np.cos(2 * np.pi * combined_df['Month'] / 12)

print("\nDataFrame with technical indicators and cyclical features:")
print(combined_df.head())
print("\nDataFrame info after preprocessing:")
combined_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define features to use for scaling and model input
features = ['Close', 'MACD', 'MACD_signal', 'RSI', 'Day_sin', 'Day_cos', 'Month_sin', 'Month_cos', 'Volume', 'Open', 'High', 'Low'] # Include all numerical features


# Prepare data for LSTM (creating sequences per ticker)
window_size = 60
target_shift = 1

X, y, X_dates, y_dates, tickers_list = [], [], [], [], []

# Process data for each ticker separately
for ticker in combined_df['Ticker'].unique():
    ticker_df = combined_df[combined_df['Ticker'] == ticker].copy()

    # Select and scale features for the current ticker
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(ticker_df[features])

    # Create sequences
    for i in range(window_size, len(scaled_data) - target_shift):
        X.append(scaled_data[i - window_size:i])
        y.append(scaled_data[i + target_shift, features.index('Close')])  # Predict the 'Close' price
        X_dates.append(ticker_df['Date'].iloc[i - window_size:i].tolist())
        y_dates.append(ticker_df['Date'].iloc[i + target_shift])
        tickers_list.append(ticker)


X, y = np.array(X), np.array(y)
X_dates, y_dates, tickers_list = np.array(X_dates), np.array(y_dates), np.array(tickers_list)


print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
print("Shape of X_dates:", X_dates.shape)
print("Shape of y_dates:", y_dates.shape)
print("Shape of tickers_list:", tickers_list.shape)
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Split data while keeping sequences from the same ticker together
# We can't use a simple train_test_split with shuffle=True here
# as it would mix sequences from different tickers.
# A manual split or using GroupShuffleSplit would be more appropriate for a robust evaluation
# However, for simplicity and demonstrating the combined training, we'll use a time-based split within each ticker or a simplified split if the order is already by ticker and date.
# Assuming the data is sorted by Ticker and then Date, a simple split will keep initial tickers/dates in train and later ones in validation.

# Determine split point (e.g., 80% for training)
split_ratio = 0.8
split_index = int(len(X) * split_ratio)

X_train, X_val = X[:split_index], X[split_index:]
y_train, y_val = y[:split_index], y[split_index:]
tickers_train, tickers_val = tickers_list[:split_index], tickers_list[split_index:]
y_dates_train, y_dates_val = y_dates[:split_index], y_dates[split_index:]


print("Shape of X_train:", X_train.shape)
print("Shape of X_val:", X_val.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_val:", y_val.shape)


# Build the LSTM model
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(32))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# Build the model
model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

# Callbacks (optional)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_combined_model.h5', save_best_only=True)

# Model training
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop, checkpoint],
    verbose=1
)
```
